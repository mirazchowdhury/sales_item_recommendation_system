```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import uuid
import os

# ==========================================
# STEP 1: LOAD RAW DATA
# ==========================================
print("Loading and Preparing Data...")

# আপনার সম্পূর্ণ Item লিস্টটি এখানে পেস্ট করবেন
raw_items = """
Beef Bone Full
Akher Sugar - 1 kg
Green Chilli (kg)
Onion Deshi (New) - kg
Shrimp Fish Bagda
Teer Sugar - 1 kg
ACI Pure Salt - 1 kg
Prince Lacha Semai - 200 g (Pkt)
Sweet Pumpkin (kg)
Lassa Special Semai - 500 g (Pkt)
Pointed Gourd (Potol) - kg
Lady's Finger (Vendi) - kg
Potato Deshi New - 1 kg
Carrot (China) - kg
Cow Liver - kg
Wheel 2-in-1 Clean & Fresh Powder - 1 kg
Milk Vita Liquid Milk - 1 L
Snickers Chocolate - 50 g
Vermicelli Semai - 200 g
Coca-Cola - 2.25 L (Pet)
Malta (Yellow) - kg
Tomato LC - kg
Capsicum Green
Garlic China Big - kg
Prince Bread (Special) - 400 g
Green Papaya - kg
Diploma Milk Powder - 500 g
Chicken (Skinless)
Special Chinigura Rice - 1 kg
Cucumber (Hybrid) - kg
Vim Bar Lemon - 300 g
Apple Green
Musur Pulse Kangaroo (Special) - 1 kg
Prince Sour Curd - 500 g
Bottle Gourd Big (Lau) - pc
Rui Fish Medium (MNK)
Pabda Fish Small (HB)
Snake Gourd (Chichinga) - kg
Bitter Gourd (Usta) - kg
Apple (Gala)
Chicken (Deshi)
Ginger Burmese - kg
Grapes (Red)
Wheel Laundry Soap - 125 g
Kachki Fish (THR)
Buldak Noodles (All)
Akher Sugar - 2 kg
Chicken Deshi (Pakistani) Cock
Cow Stomach - kg
Cucumber Deshi - kg
Bashundhara Paper Napkin - 100 pcs
Brinjal Oval (Tal Begun) - kg
Vim Bar Lemon - 125 g
Apple Fuji
Cinnamon (Daruchini) - 100 g
Radhuni Biriyani Masala - 40 g
Moong Pulse (Sona) - 1 kg
Korean Ramen 3X Spicy Noodles - 248 g
Yardlong Bean (Borboti) - kg
Brinjal Long (Thin) - kg
Fresh Refined Sugar - 1 kg
Thai Papaya Ripe Red Lady
Pear (Naspati) - kg
Pomegranate (Mitulla) - Box
Rin Advanced - 1 kg
Pusti Soyabean Oil - 5 L
Chashi Aromatic Chinigura Rice - 1 kg
Radhuni Spices Cumin - 100 g
Pran Hot Tomato Sauce - 500 g
Coriander Leaf - 100 g
Kurkure Cream & Onion - 37 g
Aarong Butter Salted - 100 g
Shrimp Fish Small (THR)
Cardamom (Elachi) - 50 g
Kolson Vermicelli - 200 g
Net Pumpkin (Jali Kumra) - pc
Golden Raisin (Kismis) - 100 g
Prince Borhani - 1.5 L
Grape White Small (Indian)
Mutton (Special) - kg
Kolson Macaroni All Flavors - 200 g
China Jhilmil Scrubber (4/23)
Mola Fish (THR)
Rui Fish S1 (MNK)
Pusti Atta - 2 kg
Beef Without Bone Special
Prince Bread (Sugar-Free) - 400 g
Kaveri Mehendi Cone 2 pcs - 25 g
Aarong Sour Curd Yogurt - 500 g
Cock Roast (Deshi) - pc
Musur Pulse Deshi - 1 kg
Teer Atta - 2 kg
Lemon Elachi - pc
Detos CW Chips - 30 g
Hilsha Fish Big (MNK) - pcs
Prince Borhani - 1 L
Aarong Sour Yogurt - 500 g
Nescafe Classic Jar - 25 g
Chia Seed New - 100 g
Queens Burger Bun - 250 g
Dragon Fruit (B)
Prince Square Toast - 250 g
Vim Dishwash Liquid - 250 ml
Belleame Digestive Biscuit - 135 g
Bashundhara Atta - 2 kg
Shing Fish (THR)
China Steel Scrubber Single (4/23)
Lati Thik - kg
Cabbage - pc
Prince Dubli Pulse - 1 kg
Golden Raisin (Kismis) - 200 g
Cassia Leaf (Tejpata) - 100 g
Brinjal Oval Big - kg
Sponge Gourd (Jhinga) - kg
Loitta Fish
Pran Aromatic Rice - 2 kg
Gulsha Fish Small (HB)
Funtastic Oreo Cream Pineapple
Snickers (Buy 2 Get 1 Free) - 50 g
Bisk Club Potato Biscuit - 100 g
Chocolate Digestive - 145 g
Easy Mix Roast Masala - 35 g
Bashundhara Toilet Tissue (White)
Knorr Thai Soup - 28 g
Aarong Pasteurized Liquid Milk Regular - 1 L
Teasel Gourd (Kakrol) - kg
Bashundhara Fortified Soyabean Oil - 5 L
Funtastic Oreo Cream Biscuit - 43 g
Sprite Pet - 2.25 L
Teer Soyabean Oil - 1 L
Bele Fish Small
Cauliflower (B) - pc
Mr. Cookie Small - 39 g
Mini Burger Bun 4 pcs - 160 g
Bashundhara Paper Towel - 250 pcs
Kurkure Naughty Tomato - 37 g
Tengra Fish
Roja Deshi Classic Paratha - 1300 g
Fresh Salt Premium - 1 kg
Bounce Choco Cream Biscuits - 32 g
Kurkure Masala Munch - 37 g
Bombay Chilli - 5 pcs
Calendar Pineapple
Prince Cream Bread Bun
Poa Fish Small (Deshi)
UHT Chocolate Milk - 200 ml
Knorr Soup - 24 g
Puti Fish Deshi (HB)
Teer Premium Iodized Salt - 1 kg
Orange (China) Net - kg
Maggi Masala Noodles - 496 g
Chicken (With Skin)
Queens Butter Bun - 100 g
White Kaju Badam Special - 100 g
Kurkure Spicy Chicken - 37 g
Kurkure Chilli Chatka - 37 g
Mango Rupali Special - kg
Boot Pulse Broken - 1 kg
Confidence Salt - 1 kg
Cocola Champion Biscuits Choco - 65 g
Hilsha Fish Medium - pcs
Cumin - 200 g
Radhuni Spices Turmeric - 200 g
Chira - 500 g
Black Pepper Powder - 50 g
Radhuni Spices Coriander - 100 g
China Steel Scrubber 4 pcs Set (4/23)
Radhuni Spices Chilli - 100 g
Queens Egg Bread - 125 g
Banana Green Small - kg
Rupchanda Small (THR)
Onion LC New
Magur Fish (THR)
Prince Tiffin Bun
Bitter Gourd (Korolla) Big - kg
Rupchanda Chinigura Rice - 1 kg
Radhuni Chilli Powder - 200 g
Coca-Cola Pet - 600 ml
Doodles Stick Noodles - 300 g
CBL Munchee Cream Chocolate Wafer - 200 g
Prince Dragon Egg Noodles - 300 g
Lemon (Colombo) - pc
Radhuni Garam Masala - 40 g
Clove (Lobongo) - 50 g
Prince Fresh Eggs Red - 12 pcs Box
Sammam Fruit - kg
Bacha Boot Pulse - 1 kg
Prince Sour Curd - 1 kg
Prince Coconut - pc
Fresh Paper Napkin - 100 pcs
Thai Guava
Fresh Soyabean Oil - 1 L
Marks Cookies & Cream Flavor - 200 ml
Hilsha Fish Small (Sajah) - pcs
Sprite Pet - 1.25 L
Rin Advanced - 500 g
Pepsodent Germicheck+ Toothpaste - 100 g
Maggi Saad-E-Magic Seasoning Mix - 48 g
Saffron (Jafran)
Mirzapore Best Leaf Tea - 400 g
Rupchanda Soyabean Oil - 5 L
Diploma Instant Milk Powder - 1 kg
Radhuni Spices Turmeric - 100 g
Maggi 2-Min Masala Jumbo - 248 g
Rin Advanced - 2 kg
Potato Bogura - 1 kg
Chicken Breast
Arla Dano Daily Pushti Pack - 200 g
Bashundhara Kitchen Towel - 1 Roll
Aarong Pasteurized Full Cream Milk - 1 L
Rupchanda Soyabean Oil Pet - 2 L
Harpoon Toilet Cleaner - 750 ml
Starship Chocolate Milk - 200 ml
Tomato Sour Deshi - kg
Shoal Fish Big (THR)
Harpic All-In-One Powder - 400 g
Potato Yam
Bashundhara Paper Towel 1-Ply - 250 pcs
Moong Pulse (Sona) - 500 g
Orange (South Africa) - kg
Prince Cone Mehendi
Sun Chips Mix Masala - 35 g
Blackberry (Kalo Jam)
Marks Belgian Chocolate Flavor - 200 ml
Ruchi Chanachur Hot - 150 g
Starship Condensed Milk - 397 g
Prince Special Chanachur - 150 g
Fresh Fun Fill Chocolate Biscuits - 44 g
Maggi Masala Blast Noodles 8 Pack - 504 g
Sweet Potato - kg
Mixed Vegetable - kg
Taro Root (Kachurmukhi) Tall - kg
Mirzapore Best Leaf - 200 g
First Choice - 100 g
Plum (Alu Bukhara) - 100 g
Custard Bread - 275 g
Capsicum Red/Yellow - kg
Milk Vita Butter - 200 g
Energy Plus Biscuit - 62 g
Kewra / Rose Water
White Kaju Badam - 50 g
Mango Himsagor - kg
Bele Fish Big (MNK)
Poppers Corn Coco Chips - 25 g
Pepsodent Sensitive Expert Toothpaste - 80 g
Maggi Masala Noodles 12 pcs - 744 g
Golden Harvest Premium Deshi Paratha - 1300 g
China PVC Ball (5/23)
Sagu - 250 g
Belleame Choco Little Biscuit - 45 g
Motor Bhaja - 15 g
Prince Plain Cake (B)
Good Life Mozzarella Cheese - 200 g
Teer Suji - 500 g
Piali Fish (Pachmishali)
Prince Bread Special - 800 g
Hog Plum (Amra) - kg
China Hair Clip (5/23)
Cumin - 100 g
Raisin Sesame Bread - 320 g
Fresh Atta - 2 kg
Marks Full Cream Milk Powder - 500 g
Sun Chips Tomato Tango - 38 g
Mango Himsagor Special - kg
Fresh Hand Towel - 250 pcs
Panch Phoron - 100 g
Radhuni Panch Phoron - 50 g
Mint Leaf (Pudina) - pc
Olympic Energy Plus Malai Biscuit - 45 g
Teer Masur Dal - 1 kg
Chicken Deshi Cock (Live)
Drumstick (Sajna) - kg
Bath Sponge (Dhundul) - kg
Mango Langra
Barley For Eating - 200 g
Vaseline Mosquito Defence Lotion - 100 ml
Belleame Cremo Biscuit - 90 g
Yum Muffin Cake
Jodi Chicken Sausage - 300 g
Prince Ovaltine Cake (B)
Prince Bread Big - 800 g
Olympic Salcoti Elachi Cookies - 80 g
Knorr Chicken Hot & Sour Soup - 28 g
Mr. White Detergent Powder - 2 kg
Prince Egg Noodles - 300 g
Loitta Shutki - 200 g
Banana Flower (Mocha) - pc
Prince Vinegar - 600 ml
Farm Fresh Sour Curd - 500 g
Prince Borhani - 2 L
Green Mango - kg
Maggi Masala Noodles Family Pack - 992 g
Bashundhara Toilet Tissue Extra Saving Pack - 4 Rolls
Yeast - 100 g
Radhuni Chatpati Masala - 50 g
Freedom Heavy Flow Wings - 16 pads
Farm Fresh Pasteurized Milk - 1 L
Chicken Nuggets (Box & Poly) - 250 g
Plum (Alu Bukhara) - 200 g
Cocola Egg & Chicken Noodles - 500 g
Prince Fruit Cake (B)
Chicken Sausage - 350 g
Brooke Bond Taaza Tea - 400 g
CP Chicken Hot Sausage - 400 g
Mixed Masala - 100 g
Grape Black
Jhok Mok Dish Wash Bar (Buy 3 Get 1) - 100 g
Golden Harvest Chicken Nuggets - 300 g
Rupchanda Big (THR)
Arla Dano Daily Pushti - 500 g
Prince Vinegar - 400 ml
Teer Soyabean Oil Bottle - 2 L
Nag Fruit - kg
Shrimp Fish Golda Medium (STR)
China Foam Scrubber (4/23)
Pandan Kaya Bread - 225 g
Nutmeg (Joy Fall) - 50 g
Aquafina Drinking Water - 500 ml
Pigeon - 1 pc
Chicken Meatball - 250 g
Telapia Fish Big (THR)
CBL Munchee Stix Choco Wafer
Foster Clarks Vanilla Essence - 28 ml
Radhuni Vermicelli Semai - 200 g
Cardamom (Elachi) - 100 g
Katol Fish Small (MNK)
Odonil Air Freshener Combo (40 OFF) - 50 g
Golden Harvest Chicken Sausage - 340 g
Chola Pulse - 1 kg
Black Pepper Powder - 50 g
Sensodyne Toothbrush (Buy 2 Get 1)
Danish Florida Orange Biscuits - 70 g
Bashundhara Facial Tissue - 240 Sheets
Radhuni Mustard Oil - 500 ml
Radhuni Coriander Powder - 200 g
Mr. Noodles Korean Super Spicy - 248 g
Sajeeb Kolson Vermicelli Semai (Buy 2 Get 1)
Coca-Cola Original - 250 ml
Quality Plain Paratha 10 pcs - 600 g
Chicken Samosa - 250 g
Topashi Fish (MNK)
China Spoon 6 pcs Set (05/23)
Chicken Spring Roll - 250 g
Teer Maida - 2 kg
Madina VIP Morium Dates - 500 g
Bakemans Choco Mate - 100 g
Good Life Mozzarella Cheese - 370 g
Surf Excel Aloe Vera Wash - 500 g
Aarong Ghee - 200 g
Pran Up - 2 L
Green Pumpkin - pc
Ifad Muri Toast - 350 g
Brooke Bond Taaza Black Tea - 200 g
Nocilla Cream Biscuit FG
Pistachio (Pasta Dana) - 50 g
Fresh Instant Full Cream Milk Powder - 500 g
Pusti Maida - 2 kg
Sunshine Masur Dal - 1 kg
Omega 3 Enriched Egg 12 pcs - T03
Fanta Orange - 1 L
Ifad Kaju+Choco Delight (Buy 3 Get 1 Orange Free) - 220 g
RD Chocolate Milk - 200 ml
Teer Whole Wheat Atta - 2 kg
Butter Halua - 300 g
Cocola Wafer Roll Jar - 325 g
Pusti Deshi Red Lentils (Masur Dal) - 1 kg
Yokozuna Twisty Goodness Crackers - 60 g
Oreo Choco Creme Biscuits - 43.75 g
Prince Betel Spice - 100 g
Sprite Pet - 600 ml
Pusti Mustard Oil - 250 ml
Madina VIP Morium Dates Loose - kg
Dust Boot (Besan) - 500 g
Sensodyne Deep Clean Toothbrush (Buy 2 Get 1)
Mr. Noodles Chow Mein Stick Noodles (Buy 2 Get 1)
Fanta Orange Drink - 250 ml
Shero Cream Biscuit - 85 g
Lux Rose Soap - 150 g
Fun Chocolate Cake
Danish Condensed Milk - 397 g
Prince Noli Toast - 250 g
Oreo Original Biscuits - 46.3 g
UHT Mango Milk - 200 ml
Amon Rice Loose - kg
Kath Badam (Almond) - 50 g
Queens Naan Bread - 250 g
Kinley Club Soda - 600 ml
Parachute Coconut Hair Oil - 100 ml
Sprite Original - 250 ml
Doughnut Sugar Bread - 200 g
Mr. Noodles Easy Masala - 8 pcs
Lux Jasmine & Vitamin E Soap - 150 g
Bournvita Biscuits - 41.85 g
Chanachur Special - 300 g
Chaka Advanced Ball Soap - 125 g
NJ Fresh Sweet Potato - 500 g
Lifebuoy Total Liquid Hand Wash - 170 ml
China Floor Mat (4/23)
Radhuni Chanachur Classic - 100 g
Maggi Thai Soup 4 pcs Combo - 35 g
Wood Apple Small (Bell)
Melon (Bangi) Small - kg
Prince Chicken Leg
Kachki Shutki - 100 g
Hide & Seek Black Bourbon Biscuits - 100 g
Prince Orange Jelly - 250 g
Sun Chips Garlic & Chili - 38 g
Bakemans Saltice Biscuit - 50 g
Jodi Chicken Frankfurter - 300 g
Popcorn (Khai Bhutta) - 500 g
Prince Borhani - 500 ml
Kath Badam (Almond) - 100 g
Quality Omega 3 Enriched Egg - 1 Dozen
Frutika Red Grape Drink - 260 ml
Sun Chips Wasabi - 35 g
Mrigel Fish Medium
Black Cumin (Kalijira) - 100 g
Dried Chilli - 200 g
Milk Vita Labang - 250 ml
China Water Pot Medium (4/23)
Bashundhara Maida - 2 kg
Radhuni Chanachur Bar-B-Q - 300 g
Ahmed Soya Sauce - 500 ml
Muslim Masala Badam - 100 g
Radhuni Chanachur Premium Mix - 200 g
Bashundhara Brown Atta - 1 kg
White Bread Small - 300 g
Dark Fantasy Choco Fills - 75 g
Oregano Bottle - 10 g
China Hair Band (5/23)
Prince Super Vermicelli Semai - 150 g
Sun Prawn Crackers Chips - 20 g
Burger Bun 2 pcs Pack
Cocola Egg & Chicken Noodles Family Pack
Dark Chocolate Digestive - 130 g
Starship Mango Juice - 250 ml
Harpic Power Toilet Cleaner Original - 750 ml
Dettol Original Soap - 125 g
Fit Crackers Small Pack
RD Coffee Milk (Buy 2 Get 1) - 200 ml
Beef Keema
Chashi Aromatic Chinigura Rice - 2 kg
Lifebuoy Lemon Fresh Liquid Hand Wash - 170 ml
Europe Chewing Gum - 13 g
Cocola Salty Salted Biscuits - 100 g
Chicken Boneless - kg
Pesta Nut - 50 g
Fresh Chinigura Rice - 1 kg
White Bread Big - 500 g
Tamarind Sora Special Wrapping
Pangas Fish Small (River)
Radhuni Chilli Flakes - 40 g
Green Jackfruit (Echor)
Amaranth (Data Shak) - pc
Egg & Chicken Masala Noodles (Egg)
Golden Harvest Spicy Chicken Sausage - 340 g
Vim Liquid Dish Wash - 500 ml
Sajeeb Soya Sauce - 300 ml
China Water Pot (05/23) - 1000 ml
Meril Milk & Rose Soap - 100 g
CBL Munchee Choco Kalo - 100 g
Odomos Mosquito Repellent Cream - 100 g
Katol Fish Big
Ready Koi Fish Deshi Medium
Mixed Masala Powder - 100 g
Kolson Special Lachcha - 200 g
Oreo Choco Cream Biscuits - 120 g
Funtastic Toni Toast - 210 g
Toi Moi Choco Wafer Biscuits - 1 pc
Chicken Strips - 250 g
Radhuni Tehari Masala - 45 g
Lemon Kagoji - pc
Capsicum Yellow - kg
Fresh Toilet Tissue White Roll
Quality Premium Paratha 20 pcs - 1300 g
Sensodyne Repair & Protect Toothpaste - 100 g
Dry Shrimp Shutki Small - 100 g
Deshi Paratha Mega Family Pack - 1600 g
Lemu Pet - 250 ml
Radhuni Beef Curry Masala - 100 g
Sunshine Atta - 2 kg
Lifebuoy Care Liquid Hand Wash - 170 ml
Golden Harvest Deshi Paratha Family Pack - 1300 g
Fay Napkin 4-Fold - 100 pcs
Cocola Egg Noodles - 170 g
Fun Vanilla Cake
Radhuni Kabab Masala - 50 g
Dust Rice (Atop) - 1 kg
CBL Munchee Chocolate Wafer - 100 g
Magica Choco Fills Cookies Biscuit - 220 g
Coke Zero - 250 ml
Cheesy Bites - 40 g
Nescafe Creamy Latte 6 pcs - 18 g
China Knife (5/23)
Nutella - 180 g
Aarong Sliced Cheese Classic - 204 g
Ramisa Nut White - 200 g
Hilsha Fish Large - 1 kg+
Tibet Ball Soap - 130 g
Jackfruit Seeds - kg
Olympic Digestive Biscuits - 110 g
Bhaja Maskalai Pulse - 1 kg
Fresh Fantasy Chocolate Biscuits - 75 g
Funtastic Milk Mate Biscuit - 45 g
Lux Fragrant Skin Body Wash - 245 ml
Real Chocolate Wafer Roll - 50 g
Sun Chips Tomato Tango - 75 g
Sunsilk Black Shine Shampoo - 330 ml
Ifad Nutri Bites (Buy 2 Get 1 Cheesy Bites) - 180 g
China Peanut Packet - 200 g
Bakemans Lexus Biscuit - 58 g
Prince Soya Sauce - 500 ml
CBL Munchee Lemon Wafer - 100 g
Watermelon Banglalink - kg
Marks Milk Shake - 125 ml
CP Chicken Nugget - 320 g
Isubgol Pack - 100 g
Fruit Jam (Morobba) - 200 g
Odonil Air Freshener (Buy 2 Get 1) - 75 g
Omega 3 Vitamin-E Egg 12 pcs - T27
Farm Fresh Butter - 200 g
Pringles Original - 134 g
Brooke Bond Taaza Black Tea - 400 g
Olympic Saltes Classic Biscuits - 50 g
Special Sugarcane Molasses
Olympic Nutty Biscuits - 80 g
Fresh Beetroot - kg
Quaker Oats Packet - 450 g
RD Strawberry Milk - 200 ml
Mola Shutki - 100 g
CBL Munchee Choco-Mo White Chocolate - 100 g
Harpic Power Plus Toilet Cleaner - 1000 ml
Cow Liver Original - kg
Black Salt (Bit Lobon) - 200 g
Olympic Lexus Vegetable Cracker Biscuit - 216 g
Cardamom Big (Elachi) - 50 g
Cocola Chicken Noodles - 180 g
Pran Hot Tomato Sauce Jar - 1000 g
Kazi & Kazi Green Tea - 60 g
Fit Biscuit - 130 g
Sandalina Sandal Soap - 125 g
CP Spicy Chicken Ball - 500 g
Ultra Probiotic Yogurt Drink Strawberry - 200 ml
Rupchanda Soyabean Oil Pet - 1 L
Tibet 570 Soap - 130 g
Radhuni Kheer Mix - 150 g
Golden Harvest Chicken Meatball - 400 g
Hide & Seek Biscuits - 120 g
China Body Scrubber (05/23)
Radhuni Falooda Mix Vanilla - 250 g
CBL Munchee Nice Biscuit Premium - 230 g
Maggi Hot Cup Instant Soup 4 pcs Combo
Radhuni Kala Bhuna Masala - 80 g
China Baby Sticker (4/23)
Ultra Probiotic Yogurt Drink Mango - 200 ml
Biscuit Choco Star - 45 g
Radhuni Chanachur Bar-B-Q - 150 g
Fortune Basmati Rice - 1 kg
Shrimp Fish River - kg
Prince Mixed Pulse - 1 kg
Lux Velvet Glow Jasmine Soap - 100 g
Eggy Masala Noodles (Buy 2 Get 1) - 400 g
Mutton (1)
Safe Hand Liquid Hand Wash (Buy 2 Get 1) - 180 ml
Bakemans Choco Boost - 250 g
Mango Fazli
Mango Gobindo Vog
Puck Cream - 160 g
Saad Sweet Chilli Sauce - 340 g
Surf Excel Aloe Vera - 1 kg
Wheel Soap - 100 g
Mace (Jotrik) - 50 g
Big Babol Gum - 20 g
Dano Instant Milk Powder Pack - 500 g
Ginger Powder - 100 g
Pran Hot Tomato Sauce - 340 g
Single Beef Chicken Samosa - 250 g
Prince Sweet Toast - 400 g
CBL Munchee Premium Nice Biscuits - 124 g
Sepnil Fruity Hand Wash Orange - 180 ml
Pusti Suji - 500 g
China Basin Brush (4/23)
Bisk Club Salty Salted Biscuit - 95 g
China Hook (05/23)
Oreo Chocolatey Sandwich Biscuits - 113.75 g
Queens Custard Bread - 90 g
Pepsi - 600 ml
Quality Egg Red - 1 Dozen
Red Chira - 500 g
Mr. Noodles Cup - 40 g
Fresh Deshi Premium Masur Dal - 1 kg
Nocilla Cocoa Spread - 190 g
Pran Sour Curd Yogurt - 500 g
Katol Fish XXL (Kata)
Fay Facial Tissue 150 pcs - 300 Sheets
China Car Toy (05/23)
Lemon Bright Dish Wash (Buy 2 Get 1) - 250 ml
NJ Fresh Sweet Potato - 1 kg
Tulardandy Fish Big (MNK)
Lux Velvet Glow Soap - 100 g
RD Banana Milk (Buy 2 Get 1) - 200 ml
Swiss Roll Strawberry Cake - 375 g
Gahana Baby Clip - 72
Prince Jhal Toast - 350 g
Black Salt Powder Bottle - 80 g
Fun Crack - 50 g
Muslim Dal Bhaja - 200 g
Golap Jam
Cinnamon (Daruchini) - 200 g
Vim Dishwash Liquid - 1 L
Phul Pitha Packet
Dabur Red Paste - 200 g
Flat Noodles Chicken Masala - 180 g
Tatkini Fish River (HB)
CP Chicken Nugget Chili - 320 g
Savory Molasses Brown Sugar - 454 g
Brooke Bond Taaza Tea - 100 g
Baby Corn Fresh
Latkon Fruit
Parachute Coconut Oil - 350 ml
Austagram Cheese
Koko Krunch - 70 g
Nutrilife Orange Fruit Magic - 160 ml
Golden Harvest French Fries - 500 g
Fresh Healthy Choice Biscuits - 80 g
Radhuni Kasundi - 285 ml
UHT Chocolate Milk Drink - 125 ml
Muslim White Peanut - 100 g
Tip Biscuit - 60 g
Funtastic Bakemans Glaze Biscuit - 40 g
Quality Brown Box Egg - 1 Dozen
Jhatpat Deshi Paratha 20 pcs - 1300 g
Pran Chocolate Milk - 200 ml
Milk Bread Small - 300 g
Plain Sugar Bread - 150 g
Vim Dishwashing Liquid - 500 ml
Katol Fish Medium (MNK)
Chepa Shutki - 100 g
Radhuni Mustard Oil - 250 ml
Fresh Energy Cookies Biscuits - 80 g
Queens Kaya Bread - 90 g
Pran Drinko Float Litchi Drink - 250 ml
Radhuni Kachchi Biriyani Masala - 40 g
Ponds Super Light Gel - 49 g
Kath Badam (Almond) - 200 g
Aloe Vera
Belleame Coconoir - 240 g
Close Up Cool Mint Toothpaste - 145 g
Kaalar Steam Basmati Rice - 1 kg
Ifad Orange Delight - 40 g
Lifebuoy Lemon Fresh Handwash - 200 ml
Prince Soya Sauce - 250 g
China PVC Ball - 9 cm
Kazi Chicken Nuggets Original - 250 g
Horlicks Classic Malt Jar - 250 g
Dettol Fresh Odour Protection Lemon - 125 g
Dan Cake Chocolate Muffin - 50 g
Aarong Sliced Cheese Pizza - 204 g
Shrimp Fish Golda (TR)
Al-Ansar Safavi Special Dates - 500 g
C.N Plain Paratha 20 pcs - 1300 g
Amloki
Fresh Instant Noodles Masala (Buy 2 Get 1) - 496 g
C.N Chicken Sausage - 300 g
Lux Soft Glow 4 pcs Combo - 35 g
Farm Fresh Butter Box - 200 g
Pangas Fish Big River (Kata)
Olympic Premium Toast Biscuits - 275 g
Jimmy Lollipop Chocolate - 11 g
Perfumed Facial Tissue - 200 Sheets
Quality Dal Puri 10 pcs - 450 g
Saad Hot Tomato Sauce - 500 g
Sepnil Hand Wash Refill - 170 ml
Pure It Micro Fiber Mesh
Pran Pasteurized Milk - 1000 ml
Bashundhara Facial Tissue - 240 Sheets
Kishwan Horlicks Biscuit - 250 g
Nocilla Cocoa Spread Two Flavor - 190 g
Quality Smoked Sausage - 350 g
Funtastic Glucoz Biscuit - 45 g
Belleame Toast Biscuit - 250 g
Chui Jhal - 100 g
Foli Fish
Cumin - 400 g
Nescafe Classic Jar - 100 g
Memory Masala Badam - 200 g
Puffed Rice (Muri) - 250 g
Ifad Choco Kaju Butter Free Orange - 820 g
Harpic Power Toilet Cleaner - 500 ml
Frutika Mango Fruit Drink - 200 ml
Fast Wash Detergent Powder - 1000 g
Pran Tomato Ketchup - 340 g
Lays Large ASCO - 61.5 g
Saad Savory Breadcrumbs Jar - 180 g
UHT Strawberry Milk - 200 ml
Kellogg's Chocos Variety Pack (Buy 1 Get 1)
Clear Men Cool Menthol Shampoo - 170 ml
Pabda Fish Big
Seylon Family Blend Tea - 200 g
Hilsha Fish XL - kg
Hilsha Fish XL (MNK) - pcs
Pomelo (Jumbura) Medium
Dettol Cool Soap - 125 g
Kinder Joy Chocolate Girls - 20 g
Pran Mango Junior - 125 ml
Meril Baby Toothpaste + Brush - 45 g
Star Masala - 50 g
White Chocolate Digestive - 125 g
Musur Pulse Deshi Loose - 1 kg
Mithai Soan Papdi - 200 g
Fresh Toilet Tissue Value Pack - 4 pcs
Pusti Aromatic Chinigura Rice - 1000 g
Mirzapore Tea Bags - 100 g
Red Amaranth (Lal Shak) - pc
Sweetened Condensed Milk No.1
Ifad Eggy Instant Noodles (Buy 2 Get 1)
Trix Lemon Soap - 300 g
Mr. White Detergent Powder (Buy 3 Get 1) - 500 g
Italiano Sweet Chilli Sauce - 340 g
Ispahani Butterful Biscuit - 75 g
Close-Up Green - 100 g
Cumin Powder - 100 g
Close-Up Menthol Fresh - 160 g
Mediplus Toothpaste DS - 140 g
Alfa Mayonnaise - 236.5 ml
MP Baby Hair Clip - 52
Sagu - 500 g
White Pepper Powder - 50 g
Chicken Spring Roll 15 pcs Pack - 300 g
Teer Fortified Soyabean Oil - 3 L
Aarong Ghee - 100 g
Pran Jhatpot Plain Paratha - 1600 g
Pran Dhaka Cheese - 200 g
Harpic 1 L + Lizol 1 L - 60 TK Off
Kinder Joy Chocolate Boys - 20 g
Bashundhara Gold Toilet Tissue
Queen Soft Jelly - 20 g
Dettol Soap (Buy 4 Get 50 TK Off) - 125 g
Maggi Masala Blast Noodles 4 Pack - 252 g
Lux Soft Glow Soap - 100 g
Belleame Energy GO Biscuit - 85 g
Morium Special Dates Loose - kg
Fenugreek (Methi) - 100 g
Bashundhara Paper Napkin Perfumed - 80 pcs
Mr. Cookie Family Pack - 275 g
Rok Liquid Dishwash (Buy 2 Get 1 Free) - 250 ml
Unik Complete Custard Powder - 400 g
Savory Roast Masala - 35 g
Naples Mayonnaise - 170 ml
Tibet Laundry Soap Green - 130 g
Funtastic Nimok Halal Biscuit - 86 g
Cream Crackers Sugar Free - 125 g
Radhuni Meat Curry Masala - 100 g
High Yeast Big Bottle - 20 g
Green Park Mushrooms Whole - 425 g
Fresh Kitchen Towel
Dettol Soap (Buy 3 Get 30 TK Off) - 75 g
Radhuni Haleem Mix - 200 g
Kurkure Style Spicy Chicken - 80 g
Organikare Pink Hand Wash + Refill Combo
Haiko Baking Soda - 40 g
Sweet Fennel (Misti Shop) - 100 g
Mixed Pulse - 500 g
Xpel Mosquito Repellent Spray - 30 ml
Chicken Burger Patty - 240 g
Polar Choco Crombo Ice Cream - 1 L
Kinder Bueno Chocolate - 43 g
Koko Krunch - 170 g
Mutton Head
Sajeeb ChocoMoo Chocolate - 135 g
Quality Premium Paratha 10 pcs - 650 g
Funtastic Pineapple 200 g + Choco 167 g (Buy 3 Get 1)
Kishwan Chocolate Cookies - 250 g
Butter Bite - 100 g
Carrot Deshi - kg
Poa Fish Small (MNK)
Mutton Liver
Hit Chalk - 1 pc
Lifebuoy Lemon Fresh Health Soap - 100 g
Aftab Chicken Samosa - 8 pcs
Meril Baby Gel Toothpaste Orange - 45 g
Cocola Strawberry Wafer Roll Jar
Lux Soft Glow Jasmine Soap - 75 g
Toblerone Chocolate - 100 g
Kalibaush Fish Big (HB)
China Toilet Brush (4/23)
Kazi Spicy Chicken Nuggets - 250 g
Dettol Aloe Vera Soap Bar - 125 g
Rupchanda Soyabean Oil Pet - 3 L
RD Mango Milk - 200 ml
Belleame Golden Cremo Biscuit - 90 g
Saad Tasty Black Salt - 100 g
Bakemans Ovaltine Biscuit - 80 g
Kurkure Naughty Tomato - 80 g
ACI Salt Bottle - 750 g
Rin Liquid - 400 ml
Energy Plus Biscuit - 185 g
Mountain Dew Pet - 600 ml
Shrimp Fish Golda Small (TR)
ACI Aerosol - 475 ml
Pran Lachcha Semai - 200 g
Dry Jujube (Boroi) - 1 kg
Aarong Sweet Yogurt - 500 g
Kurkure Chilli Chatka - 80 g
Fun Vanilla Pound Cake - 80 g
Fay Toilet Tissue White - 180 Sheets
Aarong Labang - 500 ml
Mojo Cola Pet - 250 ml
Arla Dano Daily Pushti - 1 kg
Aarong UHT Full Cream Liquid Milk Finopack - 500 ml
Mr. Energy Biscuits - 85 g
Maggi Healthy Thai Soup - 35 g
Nestle Koko Krunch - 15 g
Mango Bari-4 - kg
Milk Vita Ghee - 200 g
Good Knight Advance Power Active + Refill Pack - 45 ml
C.N Chicken Spicy Nuggets - 250 g
Vim Dishwashing Powder - 500 g
Muslim Moog Dal Papor - 100 g
Kumarika Herbal Soap (Buy 3 Get 1) - 100 g
Foxs Crystal Clear Fruits Chocolate - 90 g
ACI Pure Atta - 2 kg
White Pepper Powder - 50 g
Palletoss Gol Gappa - 500 g
Moza Crispy Wafer Roll Cream Choco Nut - 50 g
White Plus Toothpaste - 80 g
Lifebuoy Lemon Fresh Soap - 150 g
Puffed Rice (Muri) - 500 g
Pepsodent (Buy 2 Get 30 TK Off)
Nescafe 3-in-1 6 pcs - 15 g
Coriander Leaves - 200 g
Milk Soft Chewy Candy (Buy 2 Get 1 Free) - 50 g
Hide & Seek Choco Chip Creme Sandwiches - 120 g
Mini Oreo Original Biscuits - 61.3 g
Sun Papor Chips - 18 g
Special Butter Toast - 300 g
Jhatpat Low Fat Paratha - 10 pcs
Bata Fish Small
ACI Salt Bottle - 100 g
Prince Tandoori Chicken Noodles - 180 g
Cadbury Perk Chocolate - 13 g
Sunsilk Perfect Straight Conditioner - 80 ml
Fun Chocolate Pound Cake - 80 g
Meril Splash Milk & Beli Soap - 100 g
Crab (Kakra) - kg
Mediplus DS Toothpaste - 90 g
Red Cow Butter Oil - 200 g x 24
Happy Pop Lollipop
Fresh Maida - 2 kg
Good Knight Advance Cartridge
Motor Bhaja - 30 g
Ultra Fruit & Yogurt Strawberry - 75 ml
White Sesame Seeds (Til) - 200 g
Sunshine Muri - 500 g
Funtastic Milk Mate Biscuit (Buy 3 Get 1) - 225 g
Mama Hot Noodles - 496 g
Meril Milk Beli Soap - 150 g
KitKat (Buy 3 Get 5 TK Off) - 17.7 g
Mama Creamy Chocolate Noodles - 110 g
Aftab Chicken Nuggets - 250 g
Dabli Bhaja - 200 g
MP Hair Kakra - 54
Pran UHT Milk - 200 ml
Cadbury Dairy Milk Silk - 13.2 g
Dove Pink Beauty Bar - 135 g
Ramisa Nut Oil - 200 g
Pran Mango Bar - 10 pcs
CBL Munchee Chit Chat White Bar - 60 g
China Lunch Box (05/23)
First Choice Salted Biscuits - 180 g
China Clay Toy (4/23)
Dano Sterilized Cream - 170 g
Olympic Metallic Battery UM-3 - 1 pc
Prince Baking Powder - 150 g
Toast (Muri + Stick + Plain) Combo - 40 TK Off
Toren Classic Chocolate - 52 g
China Spoon Set 6 pcs (5/23)
Trix Liquid Dish Wash Lemon - 250 ml
Organikare Orange Hand Wash + Refill Combo
Mango Bari-4
Polar Kheer Ice Cream - 1 L
Surf Excel 1 kg (Buy 2 Get Vim 250 ml Free)
Hide & Seek Black Bourbon Biscuit All - 100 g
Sun Chips Salt & Pepper - 22 g
Lux Soft Skin Body Wash - 245 ml
Garlic Powder - 100 g
Harmony Soap - 60 g
Ramisa Papor - 10 pcs
C.N Chicken Kids Nuggets - 250 g
Pusti HT Noodles Medium (Buy 2 Get 1) - 220 g
Haiko Agar Agar - 30 g
Jhok Mok Dishwash Bar 300 g (Buy 2 Get 100 g Free)
RD Cold Coffee Milk - 200 ml
Swiss Roll Chocolate Cake - 375 g
Ramisa Moong Dal Bhaja - 200 g
Yupi Gummi Pizza - 56 g
Maggi Soup 4 pcs Combo All
Pringles Sour Cream & Onion - 134 g
Ifad Stick Toast - 350 g
Seylon Family Blend Tea - 400 g
Pran Orange Jelly - 500 g
Sensodyne Soft Toothbrush
Savlon Liquid - 1 L
Pran Chilli Sauce - 340 g
Chewa Fish
Chopstick Deshi Yummy Masala Noodles - 4 pack
Fenugreek (Methi) Powder - 100 g
Agar-Agar - 25 g
Pran Premium Ghee - 200 g
Facial Soft Pack Tissue - 400 Sheets
Sapodilla (Sofeda)
Digestive Plain Biscuits - 120 g
Sensodyne Daily Care Toothbrush
Kazi Spicy Chicken Sausages - 340 g
Jhatpat Chicken Nuggets - 300 g
ROK Glass Cleaner Spray - 350 ml
Kazi Chicken Samosa - 250 g
White Bean - kg
Food Grade Color - 50 g
Funtastic Hebby Energy Biscuits (Buy 3 Get 1) - 200 g
Vangon Fish Big
Yupi Fruit Cocktails - 30 g
Salmon Fire Box - 1 Dozen
EM Spaghetti Pasta - 400 g
Kazi Plain Chicken Sausages - 340 g
Knorr Classic Thai Soup - 12 g
RD Frooti Plus (Buy 2 Get 1) - 200 ml
Savoy Mawa Kulfi Ice Cream - 120 ml
Chupa Chups Lollipops - 12 g
Ahmed Hot Tomato Sauce - 340 g
Rok Fruity Handwash (Buy 2 Get 1 Free) - 200 ml
Jamrul Fruit - kg
Jordana Lip Liner - 0.85 g
Mutton
CBL Munchee Vanilla Wafer - 50 g
Frutika Mango Juice - 250 ml
Golda Shrimp XL
Sun Chips Mix Masala - 80 g
Refill Pen Craze Black - 6 pcs
Special Quality Biscuit Pot - 400 g
Danish Plain Toast - 350 g
Amul Dark Chocolate - 150 g
Partex Cleen Super Soft Toilet Tissue
Dabur Gulabari Rose Water - 120 ml
Savlon Liquid - 500 ml
Pumpkin Seeds - 100 g
CP Chicken Stick - 250 g
Savoy Butter Scotch Ice Cream - 120 ml
Kazi Paratha 20 pcs - 1300 g
Sugar Free Bread Special - 300 g
Alfa Mayonnaise - 473 ml
RD Banana Milk - 200 ml
Colgate Active Salt Paste - 200 g
Black Fighter Max Mosquito Coil - 10 pcs
Dream Lollipop - 1 pc
Teer Chinigura Rice - 1 kg
Mrigel Fish XL River
Odonil Lavender Meadows Air Freshener - 48 g
Energy Biscuits - 80 g
Kaikka Fish Sea
Sajeeb Hot Tomato Sauce - 340 g
Aarong Labang - 250 ml
Savoy Coffee Ice Cream - 120 ml
Pran Puffed Rice - 500 g
Sepnil Tea Tree Oil Hand Wash Refill - 170 ml
Meril Nail Polish Remover - 40 ml
Hubba Bubba Mega Gum - 56 g
China Mobile Toy (05/23)
Nestle KitKat Chocolate - 41.5 g
Polar Vanilla Ice Cream Tub - 1 L
Kabir Kashmiri Mirch - 100 g
Godrej No.1 Sandal Soap (Buy 3 Get 1 Free) - 100 g
MP Hair Rubber Band - 07
China Scissors (5/23)
Radhuni Chicken Masala - 100 g
Kazi Chicken Spring Roll - 250 g
Dove White Beauty Bar - 135 g
ACI Aerosol - 800 ml
ACI Pure Brown Atta - 2 kg
Dettol Handwash 170 ml (Buy 2 Get Tiffin Box Free)
Ramisa Dal Bhaja - 200 g
Cocola Chicken Noodles - 300 g
Brown Bread - 300 g
Prince Corn Flour - 150 g
Olympic Tim Tim Pineapple Cream Biscuits - 272 g
Milk Merry Biscuits - 240 g
Natural Loofah Sponge Medium - 533
Chaka Washing Powder - 1000 g
Radhuni Chanachur Bar-B-Q - 500 g
Super Refined Naphthalene - 200 g
Al-Ansar Safavi Special Dates Loose - kg
Katol Fish Big (MNK)
Revival Karak Masala Chai - 200 g
Chicken Keema - kg
Radhuni Borhani Masala - 50 g
Sunsilk Hairfall Solution Shampoo - 330 ml
Bashundhara Suji - 500 g
Austagram Mozzarella Cheese - 250 g
Johnson's Baby Powder - 100 g
Minister Bright Wash Detergent - 2 kg
China Jar Small (4/23)
Golden Harvest Chicken Shami Kabab - 250 g
Cucumber Big - kg
China Carpet Brush (4/23)
Aarong Cheezarella Cheese - 200 g
Ultra Fruit & Yogurt Mango - 75 ml
Rok Apple Handwash - 350 ml
Bashundhara Toilet Tissue Lemon
Chicken Lollipop - 02
BD Garam Masala - 50 g
Toi Moi Vanilla Wafer Biscuits - 98 g
Chocolate Cream Wafer Biscuit - 70 g
Center Fruit Gum - 23.6 g
Lemon White Detergent Powder (Buy 3 Get 1) - 500 g
Dry Puti Shutki - 100 g
Muslim Soft Nimki - 200 g
Prince Sweet Toast Biscuit - 400 g
AMA 3-in-1 Coffee 6 pcs Combo
Sunshine Maida - 2 kg
Pran Puffed Rice - 250 g
MP Thai Kakra Clip - 022
Pran Mango Milk - 200 ml
Basil Seeds (Tokma) - 100 g
Imperial Leather Soap - 200 g
Chamak Fabric Brightener - 100 ml
China Green Scrubber (4/23)
Sunsilk Thick & Long Shampoo - 330 ml
Fanta Soft Drink - 600 ml
China Soft Ball (4/23)
7Up Lite Pet - 600 ml
China Coffee Mug (4/23)
Brown Bread - 500 g
Pure It GKK Filter - 3000 L
Dania Lexus Vegetable - 240 g
Chupa Chups Lollipop
China Hook 2 pcs Set (05/23)
Fresh Facial White Tissue 120 pcs - 240 Sheets
Mirinda Orange - 600 ml
Kishwan Toast - 250 g
Savlon Lemon Burst Antiseptic Handwash - 180 ml
Peanut Butter Sandwich - 125 g
Chicken Spring Roll - 10 pcs
Aarong Sliced Cheese Peri Peri - 204 g
KitKat 2 Finger Dubai 3-in-1
Sajeeb Kolson Macaroni 400 g (Buy 2 Get 1 Noodles)
Pizza Bread - 225 g
Ajwa Dates - 500 g
Nutty Biscuits - 250 g
Radhuni Firni Mix - 150 g
Fruit Bread - 270 g
Maggi Healthy Corn Soup 4 pcs Combo
Sunlight Battery - 2 pcs
Nutella - 350 g
China Cloth Clip 16 pcs Set (5/23)
Radhuni Beef Masala - 25 g
Amul Milk Chocolate - 40 g
Aarong Austagram Cheese - 200 g
Haribo Happy Cola Candy - 160 g
Discovery Rigata Pasta (Buy 1 Get 1) - 400 g
Royal Club Soda - 500 ml
Dove Hair Fall Rescue Shampoo - 330 ml
Meril Splash Milk & Kiwi Soap - 100 g
Sunlight Heavy Duty AA Battery - 1 pc
Ruchi Jhal Muri - 30 g
Tripti Fuchka - 100 g
Polar Chocolate Ice Cream Tub - 1 L
Oreo Choco Cream Biscuits - 43.75 g
Kodomo Baby Powder - 160 g
Prince Apple Jelly - 250 g
Marks Milk Shake - 200 ml
Surf Excel Aloe Vera Wash - 1000 g
Pro-Dental Floss Pick - 36 pcs
Eon Plain Paratha 10 pcs - 650 g
Vita Juice - 250 ml
Olympic Dry Cake - 290 g
Rin Liquid Detergent - 800 ml
Shrimp Fish Golda Big (STR)
Ramisa Nimki Packet - 200 g
Savoy Caramel Ice Cream - 120 ml
Aarong Pasteurized Premium Milk - 1 L
Odonil Jasmine Mist Air Freshener - 48 g
Aarong Matha - 200 ml
Orbit Chewing Gum - 4.4 g
Mojo Cola Can - 250 ml
Dove White Beauty Cream Bar - 90 g
Shahi Cumin - 50 g
Belleame Chocolate Fondue - 180 g
Rui Fish XL River
Ramisa Nut Oil - 100 g
Pringles Kellogg's Chocos - 385 g
Prince Cocoa Powder - 100 g
Chicken Soup Deshi
Wrapping Paper
Funtastic Oreo Cream - 200 g
Kazi French Fries - 500 g
Prabhuji Soan Papdi - 200 g
Dry Cake Box - 400 g
Quality Chicken Crispy Meatball - 360 g
Radhuni - 50 g
Bacha Boot Pulse - 500 g
Deshi Paratha Family 20 pcs - 1200 g
Hilux Vegetable Cracker Biscuit - 216 g
Lifebuoy Germ Protection Soap - 150 g
Nestle KitKat 4 Finger Chocolate - 36.5 g
Fresh Premium Tea - 400 g
Isubgol Pack - 50 g
Mirzapore Best Leaf Tea - 100 g
CBL Munchee 3-in-1 Biscuits (Buy 2 Get 1) - 150 g
Faysal Birthday Balloons - 04
Kazi Chicken Strips - 250 g
Lal Atta Roti - 800 g
Chaat Masala - 40 g
Nutri Rich Diabetic Biscuits
Dettol Cool Soap - 75 g
Bashundhara Soft Toilet Tissue
Mr. Noodles Easy - 12 pcs
CBL Munchee Dark Chocolate Cookies - 100 g
Jhatpat Low Fat Paratha 20 pcs - 1200 g
Hit Anti Roach Gel
Starship Full Cream Milk Powder Pack - 500 g
Tibet Ice Cool Talc - 100 g
Lux Soft Glow Rose Soap - 100 g
China Green Scrubber 5 pcs Set (05/23)
Chicken Spring Roll
Cadbury Bournvita Biscuit - 46.5 g
Harpic Total Powder - 400 g
C.N Rice Roti 10 pcs - 500 g
Ruchi Red Chilli Sauce - 360 g
Good Knight Advance Power Active + Combi Pack - 45 ml
Mr. Brasso Spray - 500 ml
Ayir Fish Small (HB)
Fresh Facial Tissue White 150 pcs - 300 Sheets
Madhuban Oval Paan Masala
Fresh Lexus Crackers (Buy 2 Get 1) - 204 g
Godrej No.1 Jasmine Soap (Buy 3 Get 1 Free) - 100 g
Fortune Rice Bran Oil - 2 L
Ramisa Fried Chira Oil Packet - 200 g
Marks Full Cream Milk Powder Pack - 1 kg
Mr. Noodles Korean Super Spicy - 496 g
Aarong Ghee - 400 g
Big Babol Gum - 22.5 g
Dragon Durable Coil Lemon Flavor - 10 pcs
Aafi Puffed Rice (Muri) - 250 g
Savlon Freedom Heavy Flow - 8 Pads
Tresemme Keratin Smooth Shampoo - 580 ml
Godrej Aer Car Freshener Sea Breeze (Buy 3 Get 1 Free) - 10 g
Savlon Soap 3 pcs All (15 TK Save) - 100 g
Lifebuoy Mild Care Health Soap - 100 g
China Knife Set (5/23)
Muslim Muruli - 200 g
Godrej No.1 Rose Soap (Buy 3 Get 1 Free) - 100 g
Sensodyne Rapid Relief Toothpaste - 80 g
Savlon Ocean Blue Hand Wash - 200 ml
Pran Vermicelli Semai - 200 g
Radhuni Coriander Powder Jar - 200 g
Chanachur Special - 350 g
Pran UHT Liquid Milk - 1000 ml
Chicken Strips - 300 g
China Flower (4/23)
Dettol Fresh Soap - 75 g
China Yang Tong Stickers (05/23)
Center Fresh Gum - 20 g
Harpic Manik Jore Cleaner - 20 TK Off
Alam Baby Dress - 031
Pran Toast - 250 g
Nutrilife Apple Fruit Magic - 160 ml
Lamia Number Candle 53 (4/23)
Saad Hot Tomato Sauce - 1 kg
Trident Gum - 14 Sticks
Muslim Jhal Nimki - 200 g
Muslim Fry Chickpeas - 200 g
Cinthol Health+ Soap (Buy 3 Get 1 Free) - 100 g
Belleame Peanut Bite Biscuit - 240 g
Bounty Chocolate - 57 g
Sugar Free Crackers Biscuit - 180 g
Tibet Prickly Heat Powder - 100 g
Coriander Leaves - pcs
Pran Haleem Mix (Buy 2 Get 1) - 200 g
Savory Chatpati Masala - 50 g
Nido Growing Up Milk 1+ BIB - 350 g
Erko Mallow Stick - 20 g
Dolphin Nata De Coco Cup All Flavors - 270 ml
HRC Clevedon Tea Poly - 400 g
Godrej Aer Power Pocket Floral Delight - 10 g
Funtastic Hebby Energy Biscuits - 44 g
Enchanteur Powder All - 250 g
Lizol Citrus Surface Cleaner - 500 ml
Golden Harvest French Fries - 500 g
Radhuni Mezbani Beef Masala - 68 g
Dettol Skincare Soap - 125 g
Fresh Beetroot - pc
Oreo Mini Cup Biscuits - 20.4 g
Pran Drinko Float Coconut Jelly - 250 ml
Good Knight Gold Flash Refill - 45 ml
Malabar Spinach (Pui Shak) - pc
Strawberry Yogurt Drink UHT - 125 ml
Aarong Instant Full Cream Milk Powder - 500 g
Muslim Jhuri Bhaja - 200 g
Apple Fuji Big
Fa Soap - 175 g
Farm Fresh UHT Milk - 200 ml
Nescafe Hazelnut Latte 4 pcs Combo - 25 g
Lifebuoy Total Liquid Handwash - 170 ml
Pangas Fish Deshi
Lexus Vegetable Cracker Biscuit - 240 g
Sensodyne Fresh Mint Toothpaste - 75 g
Quality Biscuits Pot - 400 g
Safe Hand Wash Purple - 300 ml
Fresh Kitchen Towel - 2 pcs
Chopstick Masala Instant Noodles - 496 g
Grapes White (Jhora)
Khondoker Flower Broom - 56
Horlicks Jar - 500 g
Fresh Perfumed Facial Tissue 150 pcs - 300 Sheets
Chopstick Deshi Masala Noodles - 12 pcs
Pink Salt - 200 g
Calypso Mayonnaise - 250 ml
Lexus Vegetable Biscuits (Buy 2 Get 1)
Chaad Lemon Dish Washing Liquid - 500 ml
Parachute Beliful Coconut Hair Oil - 100 ml
Pran Hot Tomato Sauce - 1000 g
Almas Prayer Tupi - 56
Milk Vita Matha - 250 ml
Radhuni Meat Curry Masala - 20 g
Mediplus Advanced Gum Care Toothbrush
Koral Fish Small (STR)
Dove Intense Repair Conditioner - 170 ml
Nutkao Cocoa & Hazelnuts Spread - 200 g
Radhuni Custard Powder
Tal Misri (Sugar Candy) - 200 g
Tresemme Keratin Smooth Shampoo - 190 ml
CBL Munchee Choco-Mo Biscuits - 90 g
Godrej Aer Pocket Air Freshener All (Buy 3 Get 1) - 10 g
Sunsilk Volume Shampoo - 195 ml
Colgate Strong Toothpaste - 175 g
Hazrat Mozzarella Cheese - 250 g
Hilsha Fish Extra Small - pcs
Cumin Powder - 200 g
Vixol Bathroom Cleaner All - 450 ml
Flaxseed (Tishi) - 250 g
CP Spicy Chicken Burger Chilled - 150 g
Hazrat Ali Paneer Cheese - 200 g
Mediplus Gel Paste - 100 g
MP Baby Hair Clip - 63
Palmolive Soap - 170 g
Savlon Wet Wipes - 10 pcs
Lettuce Leaves - kg
Sepnil Sanitizing Hand Wash (Buy 3 Get 1) - 170 ml
Karfu Fish Big
Aquafina Drinking Water - 1 L
Garnier Vitamin-C Face Wash - 100 g
Shammi Juice Drink All - 250 ml
Gahana Hair Rubber Band - 54
Olympic Biscoti Cookies - 80 g
Sepnil Apple Hand Wash Refill - 180 ml
Meril Milk Soap - 100 g
Pesta Nut - 100 g
ZeroCal - 75s Box
Savoy Shahi Kheer Ice Cream - 100 ml
Dan Cake Vanilla Muffin - 50 g
Radhuni Tandoori Masala - 50 g
Skittles Fruits Candy - 45 g
Shan Biryani Masala - 50 g
Cadbury Dairy Milk Silk Bubbly - 17 g
Isubgol Pack - 200 g
Maggi Vegetable Soup 4 pcs - 25 g
Dolphin Nata Drinks All Flavors - 350 ml
C.N Chicken Garlic Meatball - 250 g
Rexona Roll-On - 50 ml
Dairy Milk Crispello Chocolate - 34 g
Safe Hand Wash Green - 300 ml
White Tone Face Powder - 30 g
Ramisa Jhuri Bhaja Packet - 200 g
Mango Gopal Bhog
ACI Pure Suji - 500 g
Ifad Orange Delight - 55 g
Shrimp Fish Dima Big (HB)
Odonil Orchid Dew Air Freshener - 48 g
Revive Sun Block Powder - 200 g
EM Penne Pasta - 400 g
Radhuni Spices Turmeric - 500 g
Pepsodent Gum Care - 140 g
MP Thai Hair Kakra - 03
China Lead Knob (5/23)
Radhuni Spices Chilli - 500 g
Mama Hot Noodles - 248 g
KitKat 2 Finger - 17.7 g
Morium Special Dates - 500 g
Jhatpat Mini Chicken Samosa - 250 g
Pran Jhatpat French Fries - 500 g
C.N Plain Paratha Jumbo Pack 20 pcs - 1600 g
Dabur Hajmola Tablet Bottle (Buy 2 Get 1) - 100 pcs
Discovery Vite Pasta (Buy 1 Get 1) - 400 g
Shezan Tomato Ketchup - 340 g
Sub Bread - 210 g
Polar Mango Mellow Ice Cream - 1 L
CP Chicken Italian Chilli Bologna - 340 g
Aarong UHT Full Cream Milk Fino - 200 ml
Lifebuoy Total Health Soap - 100 g
Lifebuoy Total 10 Soap - 150 g
Raw Turmeric (Kacha Holud) - kg
Real Strawberry Wafer Roll - 50 g
Ramisa Badam Papri Packet - 100 g
Faysal Yo-Yo 081 (4/23)
Nescafe 3-in-1 Original - 150 g
Sunsilk Green Tea & White Lily Shampoo - 195 ml
Aarong Liquid Milk - 500 ml
Butter Toast Biscuit - 200 g
Gillette Blue 3 Comfort Razor Single
Ifad Milk Biscuits - 45 g
Cotton Buds - 100 Sticks
Knorr Chefs Chicken Flavor Cubes - 75 g
HRC Clone Tea Poly - 400 g
Baqarah Pasteurized Milk - 1 L
Olympic Remote Battery AAA
Ramisa Murali Packet - 200 g
China Ice Cream Mold (05/23)
Golden Harvest Chicken Samosa - 250 g
Florida Orange Biscuit - 165 g
Sweet Green Mango
Ruchi Puffed Rice - 400 g
Muslim Dry Dates (Khurma) - 200 g
Koko Krunch - 300 g
BFL Foil Container 12 pcs - 400 ml
Belleame Saltish Cookies Biscuit - 240 g
Bashundhara Toilet Tissue Pink
China Kitchen Peeler - 04
Golden Harvest Chicken Spicy Nuggets - 250 g
Bakemans Lexus Vegetable Biscuit - 200 g
Pusti HT BBQ Chanachur - 250 g
Toren Rocco Chocolate - 20 g
Miss U Marshmallow Chocolate - 30 g
Chicken Tikka Kabab - 10 pcs
Ruchi Garlic Sauce - 350 g
Rupchanda Deshi Lentil - 1 kg
ACI Pure Chinigura Rice - 1 kg
Savlon Aloe Vera Hand Wash - 200 ml
RC Cola Pet - 250 ml
Mediplus MC Toothbrush
Taaqa Apple Juice - 160 ml
Custard Powder - 150 g
Pop Chicken - 250 g
China Floor Mat (4/23)
Haiko Cocoa Powder - 100 g
Senora Confidence - 15 Pads
Vixol Bathroom Cleaner All - 900 ml
Pepsi Diet - 600 ml
Onion LC New - 1 kg
Dr. Rhazes Lemon Oil Air Freshener - 150 g
Muslim Chira Bhaja - 200 g
Madina VIP Morium Dates - 1 kg
Tatkini Fish Deshi (HB)
Cinnamon Powder (Daruchini) - 100 g
Meril Baby Gel Toothpaste Strawberry - 45 g
Gahana Rubber Jhuti - 51
Quality Chicken Lollipop - 350 g
Fortune Mustard Oil - 250 ml
Ruchi Tomato Sauce - 350 g
China Vegetable/Fruit Peeler (5/23)
Pure It GKK Filter - 1500 L
Mama Hot & Spicy Cup Noodles - 62 g
Ramisa Murir Mowa Packet - 20 g
Basil Seeds (Tokma) - 200 g
Mr. Magic Handwash Powder (Buy 3 Get 1 Free) - 9 g
Prince Labang - 1 L
Red Mustard Seeds - 200 g
Whisper Ultra Nights - 15 pcs
Polo Mint Small - 15 g
Ponds White Beauty Skin Cream - 50 g
White Mustard Seeds - 200 g
Parachute Coconut Hair Oil Jar - 200 ml
C.N Chicken Nuggets - 250 g
Dove Intense Repair Shampoo - 170 ml
Ramisa Til Papri Packet - 100 g
Cosco Soap - 80 g
Melbourne Sliced Cheese - 200 g
Champion Mushroom - 400 g
Matador i-teen Pen - 6 pcs
CP Chicken Frank Sausage - 400 g
Maggi 2-Min Masala Sumo - 620 g
Taro Root (Man Kanchu Deshi) - 1 pc
China Ice Tray (05/23)
China Kitchen Peeler - 15
RC Jeera Pani - 250 ml
Angelic Fresh Citrus - 300 ml
Amul Peanut Butter - 200 g
BFL Foil Container 6 pcs - 510 ml
Sweet Rainbow Umbrella - 32 g
BD Butter Toast Biscuit - 250 g
Dabli Bhaja - 200 g
Colgate Herbal Toothpaste - 100 g
Sensodyne Expert Toothbrush (Buy 2 Get 1)
Hibiscus Mushrooms - 425 g
Twix Chocolate - 50 g
Bakemans Ovaltine Cookies - 80 g
Aftab Chicken Mini Samosa - 250 g
Cosco Soap - 30 g
Harpic Toilet Cleaner (20 TK Off)
Chicken Popcorn - 250 g
China Doll Toy (05/23)
Harpic Toilet Cleaner Powder - 200 g
Head & Shoulders Anti Hairfall Shampoo - 180 ml
Prince Khesari Pulse - 1 kg
Fresh Oyster Mushroom - 150 g
Golden Harvest Atta Roti - 1200 g
Haldiram Kacha Fuchka - 200 g
Kazi Kids Chicken Nuggets - 250 g
Rooh Afza Sharbat - 300 ml
Godrej No.1 Lime Soap (Buy 3 Get 1) - 100 g
Bactrol Soap All - 100 g
Sajeeb Bar-B-Q Noodles - 180 g
Chicken Roll - 5 pcs
Lux Skin Detox Body Wash - 245 ml
Roja Fish Nugget 16 pcs - 240 g
Muslim Jhinuk Pitha - 200 g
Dove Shampoo 340 ml + Conditioner 170 ml Combo
Ahmed White Vinegar - 250 ml
Meril Milk Bar Soap - 150 g
Ispahani Premium Ghee Rusk Toast - 180 g
Kishwan Sugar Free Cookies Biscuit - 300 g
ACI Pure Thai Noodles (Buy 3 Get 1) - 150 g
BD Food Hot/Jhal Chanachur - 300 g
Mini Oreo Chocolate - 61.3 g
Memory Jhuri Bhaja Packet - 200 g
Prince Labang - 500 ml
Danish Dry Cake Biscuits - 350 g
Farm Fresh Ghee - 200 g
Olympic Daily Toast Biscuits - 210 g
Dettol Skin Care Soap - 75 g
M&M Gems Chocolate - 45 g
Koko Krunch Duo - 330 g
Premio Diamond Mug All - 2 L
Lizol Citrus - 1000 ml
Dabur Herbal Toothpaste (Buy 1 Get 1)
Sufia Comb - 07
Improved Jet Poly Pack - 200 g
Golden Harvest Vegetable Spring Roll - 400 g
Dr. Rhazes Lavender Hand Wash Bottle - 300 ml
China Floor Brush (4/23)
Godrej Aer Car Freshener Lavender Bloom - 10 g
Frutika Mango Juice - 1 L
Oreo Original Biscuits - 120 g
Memory Salt Monakka Packet - 200 g
Bournvita Biscuits - 41.85 g
Ahmed Tomato Sauce - 1000 g
Pepsi Pet - 2 L
China Pedicure Set (4/23)
All Time Croissant - 55 g
Pran Chilli Powder - 100 g
Ferrero Rocher - 3 pcs
China Cutting Board (5/23)
Boost Standard Jar - 400 g
Fresh Facial Tissue 120 pcs - 240 Sheets
C.N Chicken Finger - 250 g
Meril Baby Shampoo - 110 ml
Hershey's Cookies N Creme - 40 g
Chicken Tikka Kabab
China Glass Jar (5/23)
Koi Fish Big (HB)
Tastie Breadcrumbs All - 200 g
Little Frutika Mango Juice - 125 ml
Prince Kabli Boot - 500 g
Water Spinach (Kolmi Shak) - pc
Sensodyne Sensitivity & Gum Toothbrush
BD Food Chanachur - 280 g
Morium Special Date - 1 kg
Revive SunBlock - 100 ml
Kellogg's Corn Flakes With Iron Shakti - 250 g
Mango Arshini
Jui Coconut Oil Plastic - 350 ml
Yupi Big Burger - 28 g
Cinthol Lime Soap (Buy 3 Get 1 Free) - 100 g
Kazi Spicy Chicken Meatballs - 250 g
Haiko Corn Flour Pot - 160 g
Dettol Cool Hand Wash Refill - 170 ml
Papor (Dried Ethnic Snack) - 100 g
Pringles Smoky BBQ - 134 g
Haribo Goldbears Candy - 30 g
Godrej Aer Car Freshener Berry Rush - 10 g
Kings Sunflower Oil Pet - 2 L
Lifebuoy Total Liquid Hand Wash - 1 L
All Time Milk Bread - 300 g
Mixed Nut & Fruits - 200 g
CBL Munchee Chunky Choco - 90 g
Til Papri - 50 g
Foster Clarks Corn Flour - 200 g
Savoy Kheer Malai Ice Cream - 120 ml
Savlon Lavender Hand Wash - 200 ml
Diamond Foil Paper - 200/60 SQ
Rok Bleaching Powder - 500 g
Odonil Mystic Rose Air Freshener - 48 g
Foster Clarks Mango - 250 g
Taaqa Orange Juice - 160 ml
Eon Dal Puri - 500 g
Sepnil Hand Sanitizer Pump - 40 ml
ACI Aerosol - 350 ml
Hide & Seek Creme Sandwiches - 120 g
Mama Shrimp Noodles - 496 g
Maskalai Pulse - 1 kg
Ruchi Sweet Chilli Sauce - 350 g
Pepsodent Strawberry Toothpaste - 45 g
Sensodyne Expert Toothbrush
Dettol Skincare Hand Wash Refill - 170 ml
Farm Fresh Standard Milk - 1 L
Kazi Chicken Burgers - 200 g
Special Horlicks Biscuits - 400 g
Ramisa Fried Chira Masala Packet - 200 g
Savlon Iris Hand Wash - 200 ml
Baking Soda - 200 g
Foster Clarks Orange - 250 g
Yupi Twin Burger - 53 g
KSF Mozzarella Cheese - 250 g
Amul Kool Kesar Flavor - 180 ml
Wrapping Paper - 50
Elephant Apple (Chalta) - 1 pc
Pusti HT Orange Biscuit - 50 g
Pro-Dental Pixie Toothbrush
China Baby Rabbit Toy (5/23)
Quanta Mushroom - 425 g
Diamond Naphthalene - 200 g
Vaseline Deep Restore Lotion - 300 ml
Foster Clarks Food Color Green - 28 ml
Kellogg's Oats - 400 g
7Up - 2 L
Fruit Cake - 380 g
Mentos Rainbow - 37.8 g
Gurer Shahi Khorma - 200 g
Chicken Breakfast Sausage - 180 g
Mr. Noodles Easy Instant 16 Pkt - 62 g
Boal Fish Small (HB)
Pepsodent Fresh - 140 g
Fair & Lovely Multivitamin - 50 g
Glow & Lovely Advanced Multivitamin - 80 g
Suree Fish Sauce - 150 ml
Lamia Hot Wheels Car Toy - 551
Chicken Nuggets Kids - 250 g
Mr. Energy Instant Biscuit - 240 g
Cosco Chocolate - 85 g
Big Bom Lollipop - 02
Vaseline Petroleum Jelly - 50 g
Trix Lemon Soap - 125 g
BD Tomato Sauce - 340 g
Baby Cherie Cotton Buds - 100 pcs
Paprika Powder - 20 g
Milk Vita Sour Curd - 500 g
Fruit Cake - 270 g
Sunsilk Hair Fall Solution Shampoo - 170 ml
KSF Cooking Paneer - 250 g
Sunfeast Bounce Biscuits - 32 g
C.N Chicken Breaded Crispy Strips - 250 g
Ramisa Diabetes Murali - 200 g
Fresh Fortified Soyabean Oil - 3 L
Pepsodent Charcoal White Toothpaste - 140 g
Pastiano Maja Penne Pasta Poly - 250 g
Oreo Chocolate Creme - 133 g
Fresh Turmeric Powder - 200 g
Tibet Luxury Talcum Powder - 100 g
RD Mango Milk - 200 ml
Jhatpat Premium Deshi Paratha 10 pcs - 650 g
Golden Harvest Plain Paratha Family Pack 20 pcs - 1300 g
Dalos Juice All Flavors - 250 ml
Mirinda - 250 ml
Face Powder Puff
MP Hair Kakra Clip - 081
Sweet Curd - 500 g
Super Refined Naphthalene - 50 pcs
Cadbury Dairy Milk Silk Oreo - 15 g
Dr. Rhazes Orchid Hand Wash Pack - 300 ml
Radhuni Chicken Masala
Marshmallow Chocolate - 26 g
Kazi Chicken Lollipop - 300 g
Sunsilk Shampoo All - 300 ml
Plain Paratha 20 pcs - 1300 g
Pringles Potato Crisps All - 158 g
KitKat 2 Finger Dubai 2-in-1 (10 TK Off)
Harpic 750 ml (Buy 2 Get 40 TK Off)
Raffaello Chocolate - 3 pcs
Radhuni Fish Curry Masala - 20 g
Vegetable Spring Roll
Ispahani Premium Toast - 170 g
Miss & Mrs Velvet Matte Lipstick - 2.8 g
Ifad Orange Delight Biscuit - 160 g
Rok Lemon Floor Cleaner - 900 ml
Mango Hari Vanga - kg
Quaker Oats Jar - 450 g
Cardamom Powder - 50 g
Eel Fish (Baim Star) Small
Muslim Coconut Naru - 1 Box
Prince Mixed Fruits - 100 g
Royal Elegance Choice Whole Mushrooms All - 69
Nestle KitKat White Chocolate - 41.5 g
Ponds White Beauty Face Wash - 100 g
Colgate Max Fresh Blue Gel - 150 g
Umami Chilli Sauce - 300 ml
Muslim Chui Pitha - 200 g
Aafi Puffed Rice - 500 g
Vinut Red Grape Juice Can - 240 ml
Muslim Soft Monakka - 200 g
Radhuni Cumin Powder - 500 g
Dettol Re-energize Handwash - 170 ml
Dettol Fresh Hand Wash Refill - 170 ml
Polar Mango Ice Cream Tub - 1 L
Umami BBQ Sauce - 300 ml
Coriander - 200 g
Gillette Blue II Razor
China Cloth Brush (4/23)
Red Cow Instant Milk Powder - 500 g
Big Top Lollipop - 1 pc
Haldiram Black Pepper Powder - 30 g
White Bean (Mau) - kg
Ramisa Fried Chickpeas - 200 g
Marks Full Cream Milk Powder Tin - 1 kg
China Balloon 20 pcs Packet (5/23)
Radhuni Vinegar - 280 g
Maltova BIB Pack - 400 g
Churi Shutki Big - 200 g
Savlon Fresh Soap - 125 g
Fuchka (Kacha) - 200 g
Saad Monosodium Glutamate Foil Pack - 20 g
Pusti Black Tea - 400 g
China Kitchen Khunti - 38 (5/23)
Jet Classic Detergent Poly - 1000 g
Haribo Goldbears Candy - 80 g
Nutmeg Powder - 20 g
KitKat Chunky Chocolate - 57 g
Quality Vegetable Singara 20 pcs - 400 g
All Time White Bread - 300 g
Sunshine Suji - 500 g
Pringles Salt & Vinegar - 158 g
Olympic Marie Gold Biscuits - 190 g
Sajeeb Orange Soft Drink Powder - 500 g
Quality Chilli Cheese Chicken Salami - 200 g
Shan Karachi Beef Biryani - 60 g
Kazi Chicken Shami Kabab - 200 g
Beef Tikka Kabab
Sensodyne Sensitivity & Gum Toothbrush (Buy 2 Get 1)
Savoy Choco Fudge Ice Cream - 120 ml
Dabur Honey - 100 g
China Rain Coat R100 (4/23)
No-Kroach - 8 g
Fry Pan Without Lid - 22 cm
Aloo Puri - 454 g
Bismillah Safety Pin - 01
Clean & Clear Foaming Face Wash - 150 ml
Radhuni Fish Curry Masala - 100 g
BD Chilli Sauce - 360 g
CloseUp Toothpaste - 100 g
Lizol Floral - 1000 ml
Act II Butter Delite - 70 g
Lactogen 3 BIB - 350 g
Golden Rose Lipstick - 4.2 g
Mediker Safe Life Hand Wash 4 pcs Combo - 170 ml
Capital Food Essence All - 28 ml
Sunsilk Black Shine Shampoo - 170 ml
Memory Muruli Packet - 200 g
Ramisa Til Naru Packet - 100 g
Imperial Leather Soap - 115 g
Tic Tac Orange - 7.2 g
Polar Strawberry Smile Ice Cream - 1 L
Faisha Fish (STR)
Himalaya Neem Face Wash - 100 ml
Foster Clark Mango - 450 g
Chopstick Bar-B-Q Noodles - 8 Packs
Pusti Mustard Oil - 1 L
7Up - 250 ml
EM Shell Pasta Big - 400 g
Polar Raffinato Mocha Ice Cream - 1 L
Pran White Chicken Meatball - 250 g
Dano Full Cream Milk Powder - 1 kg
Harpic Manik Jor 750 ml + 500 ml
Haiko Mix Herbs - 25 g
Shipon Ear Ring - 041
Nestle KitKat Chocolate - 28.5 g
Parachute Advanced Oil - 275 ml
CBL Munchee Chocolate Fingers - 90 g
Mama Creamy Strawberry Noodles - 110 g
Chaad Orange Dish Washing Liquid - 500 ml
Muslim Shapla Papor - 250 g
Horlicks Classic Malt Jar - 1 kg
Mama Chicken Noodles - 496 g
ACI Pure Maida - 2 kg
Shezan Hot Tomato Sauce - 340 g
Saad Custard Powder Tin Can - 150 g
Haiko Corn Flour - 150 g
Golden Harvest Mini Chicken Singara - 300 g
Brinjal Oval Small Red - kg
Pusti HT Aloo Bhujia - 20 g
Shipon Ornaments - 001
Pran Jhatpot Aloo Paratha - 625 g
EM Vite Pasta - 400 g
Ponds Daily Face Wash - 50 g
Cocola Mughal Ghee Toast - 200 g
Shan Malay Chicken Biryani - 60 g
Bismillah Hair Comb (Chiruni) - 02
Badam Papri - 50 g
Radhuni Spices Chilli Jar - 200 g
Pran Ready Fuchka - 200 g
Cocola Bowl Noodles - 70 g
Prince Special Ovaltine Biscuit - 400 g
Red Chili Flakes - 50 g
Aloe Vera Hair Oil - 250 ml
Eon Aloo Puri - 500 g
Parachute Skin Pure Aloe Vera Gel - 100 ml
Milky Bar Mould B-3 (10 TK Off) - 12.5 g
Koi Koral Fish
Biscuit Pizza - 40 g
Chicken Samosa
China Water Bottle (4/23)
Roja Fish Sticks 10 pcs - 240 g
Meril Baby Toothbrush Panda
Radhuni Korma Masala - 30 g
China Face Seal 9303 (4/23)
Godrej Aer Car Freshener Sea Breeze - 10 g
Bragg Apple Cider Vinegar - 473 ml
Master Hair Band - 021
Cheese Puffs - 18 g
Pepsi Pet - 1 L
Haiko Cake Sprinkles Mix - 40 g
Clinic Plus Shampoo - 340 ml
Mojo Naga Can - 250 ml
Balachaw Hot & Spicy Fry Shrimp - 250 g
Dalos Mango Can Juice - 250 ml
Fay Multi Fold Towels - 150 pcs
Bajaj Almond Drops Hair Oil BD - 100 ml
Godrej Aer Power Pocket Fresh Blossom - 10 g
Falooda - 250 g
Doodles Masala Noodles - 8 Packs
Pepsodent Orange Toothpaste - 45 g
BD Chilli Powder - 100 g
Pran Oyster Sauce - 280 ml
Meril Baby Toothbrush Giraffe
Coconut Powder Cake Decoration - 50 g
Munchy's Oat Krunch - 208 g
Gem Biscuits - 150 g
SKB Plate Both Side Mirror - 24 cm
Coffee Mate BIB - 450 g
Dal Puri - 454 g
American Harvest Milk Drink All - 290 ml
Cotton Buds - 200 pcs
Kazi Chicken Meatballs - 250 g
Taslima Towel - 021
Discovery Corni Pasta (Buy 1 Get 1) - 400 g
Farmland Full Cream Milk Powder - 500 g
Muslim Fish Papor - 100 g
Gelatine Powder - 50 g
Ponds Bright Beauty Face Wash - 50 g
Tip Biscuit Super Saver Pack - 200 g
Air Wick Citrus Splash - 50 g
Chicken Keema
Noodles Masala - 35 g
Amul Kool Kesar - 180 ml
Clariss Pomace Olive Oil - 1 L
Suree Oyster Sauce - 150 ml
Select Plus Anti Dandruff Shampoo - 75 ml
Partex Cleen Kitchen Towel - 1 Roll
China Cleaning Brush (05/23)
Sanjida Wooden Khunti - 51
Quaker Whole Oats - 700 g
Kellogg's Chocos - 110 g
Tibet Detergent Powder - 1 kg
Olympic Saltes Biscuits - 200 g
Saad Tomato Sauce - 340 g
ACI Nutrilife Fortified Rice Bran Oil - 5 L
Savlon Baby Wipes - 80 pcs
Spinach (Palong Shak) - pc
Ligion Active Cone Mehendi - 35 g
Shezan Mango Fruit Drink - 1000 ml
BD Soya Sauce - 250 ml
Madina Azoa VIP Dates - 500 g
Prince Special Saltest Biscuit - 400 g
Almarai Cheese 10 Slices - 200 g
China Cloth Clip (5/23)
Memory Jhal Nimki - 200 g
Zipbag Sandwich 165mm X 150mm
Stute Cranberry Juice - 1.5 L
Ifad Biscuit 4 pcs Combo - 820 g
Golden Rose Wow Nail Color - 6 ml
Act II Golden Sizzle - 50 g
Faysal Color Balloon - 04
Nutrilife Guava Fruit Magic - 160 ml
Hat Tortolino Macaroni - 500 g
Britannia Bourbon Biscuits - 150 g
Chira Bhaja Packet - 200 g
Swiss Garden Mushroom Whole - 425 g
Banana (Sabri)
Ramisa Chirar Laddu Packet - 20 pcs
Purnava Folate Enriched Egg 12 pcs - TE4
Chicken Quib
Tresemme Botanique Nourish & Replenish Shampoo - 580 ml
China Silicone Khunti (5/23)
C.N Chicken Burger Patty - 300 g
Pringles All - 149 g
RFL Flat Lid
Sandalina Sandal & Rose Soap - 150 g
C.N Chicken Spicy Lollipop - 400 g
Sticker M-81
Nautilus Lite Tuna - 165 g
Jordana / Pearl Lip Liner - 0.85 g
Dettol Aloe Vera Soap - 75 g
China Doll Key Ring (05/23)
Dr. Rhazes Lavender Hand Wash Pack - 300 ml
Haiko Mace Powder - 20 g
Colgate Max Fresh Red Gel Toothpaste - 150 g
Corn Flakes - 475 g
Grapes American - kg
Eon Vegetable Singara - 400 g
Vita Multifruit Juice - 250 ml
Dark Fantasy Choco Fills - 300 g
Rui Fish XL (HB)
Prince Mango Pickle - 400 g
Ayir Fish Medium (HB)
Bakemans Ovaltine Biscuits - 250 g
Onion Powder - 100 g
Pringles Cheesy Cheese IDC - 134 g
Koral Fish Big
Kishwan Ovaltine Biscuits - 350 g
Olympic Classic Marie Biscuits - 240 g
Silver Carp Fish
Tresemme Botanique Conditioner - 190 ml
China Mini Fans (4/23)
Diamond Naphthalene - 12 pcs
Sajeeb Tomato Ketchup - 340 g
Foxs Crystal Clear Berries - 90 g
Cadbury Bournvita Biscuits - 120 g
Tibet Laundry Soap Blue - 125 g
Hershey's Choco Syrup - 680 g
BD Green Chilli Sauce - 380 g
Shipon Stone Ear Ring - 051
Cerelac Wheat With 3 Fruits - 400 g
Polar Raffinato Natural Vanilla Ice Cream - 1 L
Aafi Mango Drink - 500 ml
CBL Munchee Choco Finger - 192 g
Bombay Mixed Misti Masala
Fa Roll-On All - 50 ml
Mr. Noodles Chicken 8 pcs - 496 g
Radhuni BBQ Masala - 50 g
Muslim Moog Dal Bhaja - 200 g
Garnier Naturals Burgundy - 20 ml
Mug Cover - 202A
Horlicks Chocolate - 500 g
Godrej Hair Color Easy 5-Min Shampoo Based Natural Black - 20 ml
Savlon Antiseptic Liquid - 112 ml
Diamond Naphthalene Color - 200 g
BD Bar-B-Q Chanachur - 280 g
Uncle Barns Bread Crumbs - 200 g
Rose Water - 180 ml
Gillette Blue II Plus Razor
Sunsilk Coconut & Aloe Vera Shampoo - 375 ml
Mixed Fruit (Kata)
Colgate Cavity Protection Toothpaste - 100 g
Ruchi Tomato Ketchup - 350 g
Latina Basil Seed Drink Cocktail All Flavors - 290 ml
Harpic 1 L (Buy 2 Get 50 TK Off)
Pepsodent Charcoal & Lemon Toothpowder - 100 g
Kishwan Special Toast - 260 g
Muslim Chirar Moya - 20 pcs
Oregano Leaves - 10 g
Godrej Good Knight Gold Flash - 45 ml
Dove Hair Fall Rescue Conditioner - 170 ml
Glucon-D - 400 g
Golden Harvest Mini Chicken Samosa - 250 g
China Paposh (4/23)
Pran Jhatpot Chicken Dumpling - 250 g
Diploma Instant Milk Powder - 400 g
Olympic Tim Tim Orange Biscuit - 272 g
Starship Mango Fruit Drink - 125 ml
ACI Aerosol - 250 ml
Sensodyne Fresh Gel Paste (Buy 2 Get 1 Free)
Savlon Fresh Bar Soap - 100 g
Ramisa Fried Coconut - 100 g
Deli Cup Granulated Coffee - 50 g
Pusti HT Salty Sooper Biscuit - 50 g
Ragu Pizza Sauce - 396 g
Aussiebee Pure Natural Honey - 80 g
Clemon Pet - 250 ml
China Punch Hair Clip (05/23)
Pran Butter - 100 g
Gahana Kakra Hair Clip - 5
Black Rupchanda Fish
Mars Chocolate - 51 g
Kellogg's Chocos - 250 g
Memory Salt Nimki Packet
Nautilus Lite Tuna Steak - 165 g
Bakemans Horlicks Biscuits - 250 g
China Knife Sharpener (05/23)
Khondoker Flower Broom - 02
Tang Orange Packet - 200 g
Jhatpat Mini Chicken Spring Roll - 200 g
Fresh Chocolate Chip Biscuits - 250 g
Ujala Liquid Blue - 100 ml
Ruchi Tamarind Sauce - 370 g
Funtastic Cake Carnival Box - 363 g
Wonder Dish Wash Bar - 100 g
Parachute Just For Baby Powder - 200 g
Umami Spring Roll Sauce - 300 ml
Soya Sauce - 500 ml
Barley Arrowroot - 400 g
Mixed Dry Fruits - 200 g
Diamond Alu Foil Paper - 37.5 SQFT
Maggi Healthy Soup - 25 g
Faysal Candle - 71
Italiano BBQ Hot & Spicy Sauce - 340 g
Godrej Expert Rich Hair Color Black
Kikkoman Soy Sauce - 500 ml
Children Toothbrush 3-6
Pran Jhatpot Chicken Momo - 250 g
CP Chicken Frank Cocktail - 340 g
Mustard Powder - 50 g
Memory Diabetics Murali - 200 g
Jui Coconut Oil Tin - 200 ml
RFL Pure Rectangular Container Assorted - 400 ml
China Baby Football (5/23)
Mug Cover - 205A
Beef Samosa
Kamal Rice Stick - 500 g
Savlon Antiseptic Liquid - 56 ml
Tibet Luxury Soap - 100 g
Dettol Aloe Vera Hand Wash - 170 ml
Black Watermelon - kg
Hubba Bubba Bubble Gum - 35 g
Pusti HT Milk Biscuit - 50 g
Pusti HT Pops Biscuit - 45 g
Flame Box Match - 1 Dozen
Ifad Milk Biscuits
Prince Peanut Pack - 500 g
Kalpana Korai
Butter Bite - 170 g
Mountain Dew Can - 250 ml
Dragon Mosquito Coil
China Flowers & Garden (5/23)
Eon Chicken Samosa - 300 g
Sensodyne Fresh Mint Toothpaste - 130 g
Marks Diet Milk Powder - 400 g
Pringles Original - 40 g
Hat Vite Pasta - 500 g
Ramisa Jhal Nimki Packet - 200 g
Mama Chicken Flavor Noodles - 248 g
Supermom Baby Wipes - 20 pcs
Angelic Fresh Orchid Breeze - 300 ml
Nescafe Creamy Latte 3-in-1 - 18 g
Ponds Men Oil Clear Face Wash - 100 g
Lafz Body Spray - 120 ml
ZeroCal - 150 g
Pearl Fashion Eyebrow Pencil - 1.7 g
Italiano BBQ Sauce - 340 g
Sharpener - 22
Gahana Hair Khopa Band - 09
Ramisa Murali Box - 500 g
China Basin Brush (4/23)
Pusti HT Pin Biscuit - 45 g
Memory Chanachur - 400 g
Pringles Hot & Spicy - 134 g
Pusti HT Flakes Potato Chips - 22 g
Ripon Capsule Umbrella - 043
Sensodyne Fresh Gel Toothpaste - 40 g
Arrow Toothbrush
Chicken Sausage - 300 g
Golden Harvest Vegetable Samosa 10 pcs - 400 g
HIT Cockroaches Spray - 200 ml
Dairy Milk Fruit & Nut Chocolate - 36 g
Good Luck Forever Refill Pen
Haiko Custard Powder Pot - 160 g
Munch Chocolate Wafer - 21 g
Sweet Ball Cake Decoration - 50 g
Alfa Mayonnaise - 500 ml
Kodomo Baby Shampoo - 200 ml
Quality Chicken Gizzard Singara
Almarai Cream Cheese - 200 g
Savory Haleem Mix - 200 g
Golden Harvest French Fries - 1000 g
Dettol Soap 75 g (Buy 3 Get 30 TK OFF)
Radhuni Vanilla Flavoring Essence - 28 ml
Sajeeb Orange Soft Drink Powder - 500 g
Vegetable Spring Roll - 10 pcs
China Mini Fans CS109 (4/23)
Sajeeb Tandoori Chicken Noodles - 180 g
China Hook 5 pcs Set (05/23)
Pepsi Pet Bottle - 250 ml
Odonil Air Freshener Hanger All (Buy 2 Get 1) - 48 g
Marshmallow Chocolate - 28 g
ZeroCal - 25 Sachets
Fish Head
Polar Zafran Malai Ice Cream - 1 L
Pusti HT Ghee Toast Biscuit - 250 g
Aarong Probiotic Yogurt - 500 g
Shipon Finger Ring - 051
G.G Sticker - 3
Colgate Visible White - 100 g
Royal Mushrooms - 400 g
Moong Dal Bhaja - 200 g
Ramisa Coconut Naru - 25 pcs
Dolphin Nata de Coco Mango Drink - 350 g
Pringles Cheesy Cheese - 42 g
Roja Shrimp Tempura - 224 g
Ramisa Monakka - 200 g
Fresh Cumin Powder - 100 g
Hat Elbow Pasta - 500 g
Mediplus Toothpaste - 140 g
Dove Pink Beauty Bar - 90 g
Pozzy Wet Wipes - 120 pcs
Bisk Club Kheer Cookies Biscuit - 300 g
Diamond Alu Foil Paper - 46.7 cm
Bega Gourmet Cheese Slices - 12 pcs
Marks Young Star (16-40) - 400 g
Himalaya Purifying Neem Face Wash - 150 ml
Sensodyne Fresh Mint Paste (Buy 2 Get 1 Free)
Chitol Fish (MNK)
Dettol Original Liquid Hand Wash Refill - 170 ml
Frypan With Glass Lid - 26 cm
China Egg Beater (11/21)
China Foam Ball (4/23)
Prodentalb Junior Toothpaste All - 40 g
Hit Mosquitoes Aerosol (90 TK Off) - 800 ml
Chicken Nuggets 10 pcs - 250 g
Golden Harvest Plain Paratha 10 pcs - 650 g
Eel Fish (Bayng) Big (HB)
Malka Chaat Masala - 50 g
Pran Chatpati Masala - 50 g
Ultra Yogurt Drink Aloe Vera - 200 ml
ACI Aroma Chinigura Rice - 1 kg
Quaker Oats Jar - 1 kg
Ponds White Beauty India - 35 g
Grass Carp Fish - kg
Dairy Milk Chocolate - 24 g
Tata Tea Gold - 400 g
Nescafe Classic Coffee Pack - 200 g
China Punch Ball Toy (4/23)
Cadbury Dairy Milk Silk Plain - 150 g
Savlon Liquid - 250 ml
CP Egg Red Box - 12 pcs
Queen Jelly Stick Candy - 20 g
Oral-B Cavity Defense Toothbrush
Sunsilk Conditioner - 300 ml
Baoma Mosquito Coil (Micro-Smoke)
Funtastic Premium Toast - 210 g
Lux Soft Glow Rose Soap - 75 g
Nestle Classic B-2 (15 TK Off) - 34 g
Gopal Indian Like Me Bra (4/23)
Sajeeb Corn Flakes - 200 g
Remia Salad Dressing All Flavors - 512
Coffee Mate Jar - 400 g
Children 1-3 Toothbrush
Mentos Pure Fresh Gum - 29.7 g
Savlon Mild Soap - 125 g
Fresh Toilet Tissue Gold
Saad Monosodium Glutamate Foil Pack - 100 g
Rooh Afza Sharbat - 750 ml
Bisk Club Orange Biscuits - 70 g
China Punch Clip (05/23)
Frozen Whole Chicken Skinless - 1100 g+
Banana (Sagor)
Morium Super Dates Loose - kg
Tresemme Keratin Smooth Shampoo - 340 ml
Rin Power Bright - 2 kg
Himalaya Neem Face Wash - 50 ml
Memory Chanachur - 200 g
Sunsilk Perfect Straight Shampoo - 330 ml
Muslim Fuchka - 100 g
Air Wick Lavender Meadows - 50 g
Tresemme Hair Fall Conditioner - 190 ml
Muslim Muri Moya - 12 pcs
Alif Tester - 02
Dettol Antiseptic Liquid - 750 ml
Bisk Club Butter Cookies Biscuit - 300 g
Himalaya Neem & Turmeric Soap - 125 g
Polar Butterscotch Ice Cream
Kamal Mushroom Whole - 420 g
Radhuni Chicken Fry Mix - 75 g
Zipbag 20 Bag 260mm X 280mm
Latina Apple Juice - 1 L
Houston Can - 250 ml
Dove Pampering Beauty Bar - 135 g
Shammi Mix Fruit - 330 ml
Aarong Cheese - 100 g
Marshmallow - 100 g
Diabetic Salt Biscuits - 300 g
Horlicks Classic Malt - 450 g
Faysal Candle 25cm - 72
Mama Shrimp Flavor Noodles - 248 g
Harpic Aloe Powder (Buy 2 Get 200 g Free)
Delight High Fiber Toast - 250 g
Ispahani Butterful Biscuit - 200 g
Belleame Cremo Vanilla Biscuit - 320 g
Taaza Black Tea Bags - 100 g
Aarong Cheese Cubes Classic - 120 g
Garnier Men Turbo Bright Face Wash - 100 ml
Dove Gentle Exfoliating Beauty Bar - 135 g
Chatpati Masala - 35 g
Act II Classic Salted - 50 g
Fair & Lovely Advanced Multivitamin - 100 g
Doodles Chicken Curry Noodles - 8 Packs
Remia Garlic Sauce All - 250 ml
Clove (Lobongo) - 100 g
MP Small Kakra Clip - 09
Premio Food Cover - 32 cm
Pepsodent Baby Toothpaste All - 50 g
Lafz Kaveh Body Spray - 100 g
Pastiano Vermicelli Semai - 150 g
Hide & Seek Choco Rolls Biscuits - 75 g
Oral-B Gum Protect Toothbrush
Dove Soap All - 135 g
Lays American Style Cream & Onion - 18 g
Kishwan Fiore Cookies - 300 g
SKB Ladle
China Hanger 5 pcs Set (4/23)
Rupchanda Fish
Poa Fish Sea
Golden Girl Lighting Rubber Ball - 56
China Jelly Toy 7705 (4/23)
Tang Orange Flavored Instant Drink Powder - 75 g
Pringles 134 g 2 pcs Combo (70 TK Off)
Golden Girl Velvet Matte Lipstick - 1.0 g
THL Nano Toothbrush
Parachute Beli Lite - 200 ml
Alif Screw Driver - 4" / 100 mm
Ifad Plain Toast - 350 g
Rafian Check 3-Quarter Pant - 072
Chicken Mini Samosa Premium - 250 g
Bismillah Toothpicks - 03
Water Game
Jordan Toothbrush 6-9
Horlicks - 400 g
Clear Men Cool Sport Menthol Shampoo - 450 ml
Nido Full Cream Milk Powder BIB - 350 g
Fry Chickpeas - 200 g
Ramisa Khaja Pitha Box - 6 pcs
Memory Jhinuk Pitha Packet - 200 g
Owl Taro Root (Kanchu)
Tang Orange Drink Powder Packet - 500 g
Dream Food Color Orange - 28 ml
Meril Baby Lotion - 100 ml
Head & Shoulders Anti Hair Fall 2-in-1 Shampoo - 180 ml
Pacific Gold Tuna - 165 g
Chamak Fabric Brightener - 50 ml
Haribo Worms Candy - 80 g
Fresh Coriander Powder - 100 g
Remia Salad Dressing All - 250 ml
Dream Food Color Red - 28 ml
Dabur Honey - 250 g
RFL Trim Square Container Transparent - 200 ml
Cadbury Dairy Milk Chocolate - 55 g
Parodontax Ultra Clean Toothpaste - 75 g
Godrej Good Knight Gold Flash Refill (Buy 2 Get 1 Off) - 45 ml
Pran Turmeric Powder - 200 g
Pastavilla Penne Rigate Pasta - 500 g
Pran Chicken Sausages - 340 g
Lizol Floral Cleaner - 500 ml
China Birthday Balloon (4/23)
Sensodyne Deep Clean Toothpaste - 70 g
Hand Fan
Johnson's Baby Lotion - 100 ml
Fanta Grape - 320 ml
China Chapati (5/23)
Safe Hand Liquid Hand Wash - 200 ml
Gurer Monakka - 200 g
Beef Burger Patty - 240 g
Pran Potato Biscuit (Buy 3 Get 1) - 100 g
CBL Munchee Chunky Choco - 216 g
Sunsilk Perfect Straight Shampoo - 340 ml
Asli Anmol Kali Mehndi
China Air Bou Toy (05/23)
Rupchanda Premium Najirshail Rice - 5 kg
Speed Pet - 250 ml
Golda Shrimp Big
Dr. Rhazes Orchid Hand Wash Bottle - 300 ml
Everest Kashmirilal Chili Powder - 50 g
Doreo Black Choco Sandwich - 320 g
Seylon Gold Blend Tea Bags - 100 g
Kellogg's Corn Flakes - 100 g
KitKat 2 Finger Chocolate - 15.4 g
Kazi Filo Pastry - 400 g
Latina Orange Juice - 1 L
Maggi Healthy Vegetable Soup 4 pcs Combo
RFL Monia Lock Rectangular Container - 825 ml
Angelic Sparkling Air Freshener - 300 ml
Baoma Mosquito Coil
Bakemans Butter Cookies - 250 g
Beef Italian Sausage - 280 g
RFL 4" Square Container - 919340
Golden Harvest Mini Chicken Spring Roll - 250 g
Shan Bombay Biryani - 60 g
Discovery Yellow Mustard - 227 g
Oreo Original Biscuits - 133 g
Pepsodent Germi Check Toothpaste - 200 g
Kazi & Kazi Black Tea - 80 g
Clorox - 500 ml
Shezan Mango Fruit Drinks - 250 ml
Lemon Bright Liquid Dishwash (Buy 2 Get 1) - 500 g
Safe Hand Wash Value Pack Green - 350 ml
Chef's Choice Mushroom Whole - 425 g
Ifad Toast 3 pcs Combo (Save 40)
Marks Gold (40+) - 400 g
Pran Turmeric Powder - 100 g
Italiano Apple Cider Vinegar - 500 ml
Batashi Fish (HB)
Eon Chicken Spring Roll - 300 g
Moong Dal Papor
China Single Khunti 09 (5/23)
Mr. Brasso Refill - 500 ml
Belleame Dry Cake Biscuit - 325 g
ACI Pure Najirshail Rice - 5 kg
Mama Masala Noodles - 496 g
Parachute Extra Care Oil - 150 ml
Vivo Whip Topping With Vegetable Fat - 1100 g
MP Hair Clip - 02
Minister Bright Wash Detergent - 1000 g
7Up Pet - 1 L
Coconut Craze - 250 g
Ultra Probiotic Yogurt Sugar Free - 200 ml
Parachute Advanced Onion Hair Oil - 200 ml
Haiko Chocolate Chips - 100 g
Milk Marie - 285 g
Parachute Skin Pure Olive Oil 2-in-1 - 200 ml
Queen Jelly Soft Candy - 60 g
Pran Frooto Mango Juice - 250 ml
Golden Rose Pressed Powder - 12.7 g
Discovery Penne Rigate Pasta - 500 g
Nido Full Cream Milk Powder BIB - 700 g
Ispahani Puffy Bite Biscuit - 150 g
Atta Fruit
Walnut (Akrut) - 50 g
Sanjida Wooden Spoon - 52
Septex Total Antiseptic Soap - 100 ml
Vaseline Healthy White Lotion - 300 ml
Saad Black Salt Plastic Jar - 100 g
7Up Can - 250 ml
FCD Castor Oil B.P - 70 ml
Ponds Pure Detox Face Wash - 100 g
Dan Cake Classic Brownies - 180 g
Ajwa Dates Loose - kg
Gahana M. Stone Hair Band - 08
Onion With Leaves - kg
Dettol Antiseptic Disinfectant - 500 ml
Kajoli Fish (HB)
Kazi Aloo Puri - 450 g
Golden Harvest Dal Puri 10 pcs - 454 g
Ramisa Til Papri Box - 220 g
Reggia Penne Mezzane - 500 g
Fair & Lovely BB Cream - 40 g
Vaseline Hair Tonic - 100 ml
Ramisa Jhal Monakka - 200 g
Radhuni Beef Kala Bhuna Masala - 65 g
Junior Cup Noodles - 40 g
Organikare Orange Hand Wash - 250 ml
Senora Confidence - 8 Pads
Sweet Chilli Dipping Sauce - 435 ml
Kishwan Kulfi Bite Biscuit Cookies - 220 g
China Relax Bait - 50 g
Polar Strawberry Ice Cream Tub - 1 L
Nestle KitKat Chunky - 40 g
Sensodyne Whitening Toothpaste - 70 g
Memory Chingri Monakka Packet - 200 g
Faysal Party Spray - 73
Sticker Small - 4
Savory Biryani Masala - 40 g
Cerelac 4 Fruits With Milk - 400 g
Biryani Scent - 28 ml
China Party Balloon 20 pcs (4/23)
Utsho Tri-c Mango Drink Powder - 125 g
Sepnil Stops Germ Marigold - 1000 ml
RFL Lock Round Container 300 ml - 917822
Universe Spiral Khata - 200 Pages
Polo Mint Big - 12 g
Cadbury Celebrations - 132 g
Cadbury Choclairs Packet - 402.5 g
Nescafe Original Drinks All Flavors - 240 ml
Dan Cake Fruit Cake - 220 g
Kishwan Rio & Peanut Cookies - 230 g
Nutrilife Cranberry Fruit Magic - 1 L
Vinut Red Grape Juice - 250 ml
PC Mon Joli Vernis - 10 ml
Fair & Lovely BB Cream - 18 g
China Remote Car (5/23)
Cocola Chicken Masala Noodles Box - 36 pcs
Shan Mutton Biryani Masala - 60 g
Mojo Pet - 2 L
Harpic (Buy 2 Get 40 TK Off)
Fair & Lovely Advanced Multivitamin - 25 g
Purnava Mosquito Repellent - 1s
Colgate Charcoal Clean Toothpaste - 120 g
UM Spoon 303 - 1 pc
Neem Original Olive & Aloe Vera Soap - 100 g
Garnier Color Naturals Shade No 3.16 - 100 ml
Chicken Wings & Drumsticks - 500 g
Sajeeb Hot Tomato Sauce - 1 kg
Godrej No.1 Lime Soap - 100 g
Funtastic Nimok Halal Biscuit (Buy 3 Get 1)
Golden Harvest Wholemeal Paratha 20 pcs - 1600 g
Safety Plus Glass Cleaner Refill - 350 ml
Dreem Jelly Straw - 70 g
Ramisa Chicken Chips - 125 g
Teer Mustard Oil - 250 ml
Choco Delight - 220 g
Nescafe Gold Blend - 100 g
MP Baby Hair Band - 05
Colgate Active Salt Toothpaste - 100 g
Chicken Buffalo Wings - 500 g
Shokti Plain Yogurt - 80 g
Kazi Dal Puri - 450 g
Kazi Vegetable Singara 10 pcs - 300 g
Pastiano Maja Spirali Pasta Poly - 250 g
Horlicks Junior Stage 2 - 500 g
Bombay Sweets Peanut White - 200 g
Good Knight Advance 45N
Pepsi Can - 250 ml
Freedom Heavy Flow Cotton - 8 Pads
Meswak Toothpaste - 200 g
HIT Cockroach Aerosol - 625 ml
Choco Superman Cookies - 150 g
Lacys Aluminium Foil 30cm x 8m - 26.4 sqft
Dolphin Pudding All Flavors - 85 g
Kishwan Nutty Magic - 60 g
Tamarind Gala (Prince) - 250 g
M&M Chocolate - 45 g
Cosby Fun Girls - 20 g
Lays Spanish Tomato Tango - 30.5 g
Plum Big
Kellogg's Corn Flakes - 250 g
Choc Coin Stick - 50 g / 56 g
Savlon Lavender Hand Wash - 1 L
CP Gangsters Bite Chili - 250 g
Munchys Lexus Vegetable Biscuits - 200 g
Golden Harvest Chicken Meatball Spicy - 250 g
Golden Rose Velvet Matte Lipstick - 4.2 g
Tripti Maya Fuchka - 30 pcs
Sharpener - 03
Foster Clarks Whipped Topping Mix - 72 g
Shezan Corn Flakes - 275 g
Head & Shoulders Cool Menthol Shampoo - 340 ml
Godrej Hair Color Easy 5-Min Shampoo Based - 20 ml
C.N Chicken Kofta - 300 g
Koromcha Fruit - kg
Pillow Chocolate Chips
Olympic Hello Cookies Biscuits - 250 g
Trident Sugar Free Gum Artificial Flavor - 14 Sticks
Clean & Clear Aqua Splash - 50 ml
CBL Munchee Butter Fingers - 80 g
Parodontax Fluoride Toothpaste - 75 g
Kings Sunflower Oil - 1 L
Nautilus Lite Tuna - 165 g
Dettol Liquid - 100 ml
Faber-Castell Eraser 51 - 1 pc
Ruchi Garlic Pickle - 200 g
Good Luck Kids Dream Book - 1
SC Biotin Shampoo With Onion Extract (Buy 1 Get 1 Free) - 250 ml
Fast Wash Detergent Powder - 2 kg
American Garden Mustard - 227 g
Muslim Kancha Fuchka - 250 g
CP Crispy Chicken Popcorn - 250 g
Trident Chewing Gum All - 27 g
Sriracha Hot Chilli Sauce - 440 ml
Queens Multigrain Bread - 300 g
Marshmallow Bar-B-Q
Ritzbury Milk Chocolate - 100 g
Thai Johnson's Baby Soap 2 pcs (4/21) - 150 g
MP Thai Hair Kakra - 52
Sajeeb Orange Soft Drink Powder - 250 g
Ma Steel Plate - 001
Bacha Fish Small (MNK)
Mango Banana
Dettol Hand Wash 170 ml (Buy 2 Get Tiffin Box Free)
Jordan Clean Smile Toothbrush
Pran Dal Puri - 10 pcs
Foster Clarks Orange - 450 g
Pusti Soyabean Oil Poly - 1 L
Amul Kool Kesar Flavor - 150 ml
Kellogg's Froot Loops - 285 g
Clean & Clear Morning Energy Berry Face Wash - 100 ml
Ginger Tea Box - 60 g
Garnier Men Oil Clear Face Wash - 100 ml
Al-Ansar Sukkari Dates Loose - kg
Odofresh Toilet Cleaner - 500 ml
Nescafe Classic Coffee Jar - 100 g
Golden Harvest Spicy Chicken Sausages - 340 g
CloseUp Toothpaste - 160 g
Ruchi Chanachur - 500 g
Aarong Cheese Spread Classic - 180 g
Roja Shrimp Samosa - 240 g
Memory White Badam Packet - 200 g
Shing Fish Processed - kg
Top Tasting Salt - 100 g
Mini Singara - 600 g
Cosby Toys Included Chocolate - 5 g
Mentos Fruit Flavor
Dollys Mango Pickle - 200 g
Roja Shrimp Sticks - 240 g
Slim Soft Toothbrush
China Silicone Brush Set 68 (5/23)
Senora Confidence Sanitary Napkin Belt - 15 Pads
Pran Jhatpot Chicken Sausages - 250 g
Calypso Mayolite - 250 ml
Ruchi Tomato Sauce - 1000 g
China USB Charger Fan (5/23)
Zipbag 12 Bags 330mm x 380mm
Savlon Men Soap - 100 g
Foster Clarks Custard Powder - 200 g
Foster Baking Powder - 110 g
Sanjida Wash Net - 54
Savoy Doi Ice Cream - 120 ml
Mortein Insta Combi - 45 ml
Koral Fish Medium (STR)
Domex Lime Fresh Toilet Cleaner - 750 ml
10 Color Pen - 05
Pastry Sheet - 500 g
Lays Spanish Tomato Tango - 18 g
Finis Insect Powder - 40 g
Dolphin Nata de Coco Drink All - 270 g
Pringles Potato Crisps All - 40 g
Pantene Hairfall Control Shampoo - 340 ml
Mixed Fruit Box - 250 g
Radhuni Sorshe Ilish Masala - 65 g
Katol River Fish XL (Kata)
China Ceramic Mug Y11 (4/23)
Dove Nourishing Secrets Shampoo - 340 ml
Lezita Chicken Franks - 340 g
Tuna Fish
Foster Clarks Food Color Red - 28 ml
Prince Olive Pickle - 400 g
Al Shafi Natural Honey Jar - 250 g
Palmolive Shower Gel - 250 ml
Nutella R&B - 750 g
Haldiram White Pepper Powder - 30 g
Horlicks Classic Malt BIB - 500 g
Shanto Tennis Ball 37/5
Lachchha Semai - 200 g
Chupa Chups Bon Bon Mango Lollipop Bubble Gum
Veet Cream Normal Skin - 100 g
Haribo Starmix Candy - 80 g
Pastiano Maja Macaroni Pasta Poly - 250 g
Haiko Custard Powder - 150 g
Ruchi Orange Jam - 480 g
Shipon Ear Dul - 081
Deshi Paratha Mini Family - 650 g
Starship Milk Powder - 1 kg
Safe Hand Ocean Antiseptic Handwash - 1000 ml
Glucon-D Orange Flavor Jar - 400 g
Schick 2 Exacta Sensitive
Masala Badam - 200 g
Golden Harvest Plain Chicken Sausage - 340 g
Star Fruit (Kamranga)
Champion Toothbrush
Lactogen 1 Tin - 400 g
Garnier Naturals Deep Black - 20 ml
Kikka Fish (HB)
Kwality Velvet Band VB
Ponds Pure Bright Facial Foam - 100 ml
Fogg Fine Brazilian Burst - 120 ml
NAN Optipro 4 - 350 g
Gahana Stone Hair Band - 011
Doreo Black Choco Biscuits - 420 g
Kazi & Kazi Black Tea Family Pack - 400 g
ACI Pure Aromatic Rice - 2 kg
Moha Shoal Fish XXL
Mrigel Fish Big (MNK)
Berrak Penne Pasta - 500 g
Honey Dew - kg
Organikare Lemon Hand Wash - 250 ml
Lacys Aluminium Foil 30cm x 7.62m - 25 sqft
Lusni 4 pcs Big
BD White Vinegar - 300 ml
Sensodyne Fresh Mint Toothpaste - 40 g
Shahi Toast - 200 g
Brooke Bond Taaza Tea - 200 g
Kellogg's Chocos Moons & Stars - 375 g
Chocolate Rice Cake Decoration - 50 g
Lurpak Unsalted Butter - 200 g
Sunsilk Long & Healthy Growth Shampoo - 350 ml
BD Food Chanachur - 140 g
Moha Shoal Fish XL - kg
Kazi Chicken Cutlets - 200 g
CP Chicken Patty - 500 g
Casserole With Glass Lid - 26 cm
Creamy Peanut Butter Chunky - 340 g
Domex Ocean Fresh Toilet Cleaner - 500 ml
Whisper Ultra Clean XL - 15 Pads
Golden Girl Sticky Notes - 03
Foster Clarks Orange Packet - 750 g
MP Thai Hair Clip - 52
Galaxy Chocolate All - 40 g / 42 g
Mr. Noodles Chicken - 40 g
Filo Sheet Pastry Sheet - 30 pcs
Binimoy Cartoon File - 051
Barley Arrowroot - 200 mg
Nivea Men Roll-On Fresh Active - 50 ml
Silk Bottle Assorted - 500 ml
Kodomo Baby Soap - 75 g
Parachute Aloe Vera Hair Oil - 150 ml
Master Hair Clip - 021
Good Luck Fast Gel Pen - pcs
Barchun Oyster Sauce - 350 ml
MP Hair Rubber Band - 32
Choco Cup Strawberry Cup - 40 g
Pran Thai Noodles Oriental (Buy 3 Get 1) - 150 g
Ramisa Pata Pitha - 250 g
ZeroCal - 100 tablets
Shokti Misti Doi - 80 g
Pran Thai Hot & Spicy Noodles (Buy 3 Get 1) - 150 g
Cadbury Dairy Milk Silk Chocolate - 123 g
Fresh Masala Tea - 50 g
Pearl Nail Polish - 3 ml
Golden Girl Noris Pencils 1 pc - 5.7
YC Whitening Face Wash Milk - 100 ml
China Tissue Box (4/23)
Muslim Ring Papor - 250 g
China Glass Wiper (4/23)
Harpic Bathroom Cleaner - 500 ml
Pepsi Can Malaysia - 320 ml
Twing Pet - 250 ml
Floor Cleaner Rose - 900 ml
Mentos Mint Flavor - 37 g
China Glass Stand (5/23)
China Tea Filter (5/23)
Engage Woman Pocket Perfume - 17 ml
Cockroach Finish Spray - 250 ml
Shamuk Pitha Packet
Dr. West Brush Original
BD Food Bar-B-Q Chanachur - 140 g
Roja Fish Kievs 12 pcs - 240 g
China Mini Fans CS1308-3 (4/23)
Bisk Club Shero Biscuits (Buy 3 Get 1)
Pran Special Dry Cake Biscuit - 300 g
Ramisa Jhinuk Nimki - 200 g
Wichy Organic Extra Virgin Coconut Oil - 300 ml
Golden Girl Both Side Tape - 52
Shahi Water Game - 57
Suree Soy Sauce - 150 ml
Reggia Penne Ziti Macaroni - 500 g
China Toilet Brush (4/23)
Farm Fresh Full Cream Milk Powder - 500 g
Vivo Topping / Whipping Cream - 1100 g
Harpic 750 ml + Lizol 500 ml
Olivs Oils Land Extra Virgin Olive Oil - 250 ml
Eno Lemon Flavor Combo - 6 pcs
Skittles Giants Chocolate - 45 g
Colgate Max Fresh Toothpaste All - 100 ml
Mountain Dew - 250 ml
Creamy Peanut Butter - 340 g
Rajkonna Olive Oil - 120 ml
Garnier Color Naturals Black 1 - 100 ml
Toren Classic Dark Chocolate - 52 g
Jordan Buddy Toothbrush 1-4
Nona Hilsha Shutki
China Grater (5/23)
Johnson's Baby Shampoo - 100 ml
Cyprina Red Grape Juice - 250 ml
Kodomo Toothpaste - 40 g
Sunsilk Shampoo All - 300 ml / 340 ml
Color Gum
Pusti HT Lexus Biscuit - 45 g
Muslim Til Naru - 1 Box
Ispahani Lexus Biscuit - 150 g
Vegetable Spring Roll - 300 g
Foster Clarks Corn Flour - 400 g
Dabli Bhaja - 200 g
Head & Shoulders Anti Dandruff Shampoo Smooth & Silky - 180 ml
Gillette Foam Sensitive Skin - 98 g
Golden Harvest Red Wheat Roti - 800 g
ACI Fun Chanachur Hot & Spicy - 300 g
Ponds Mineral Clay Face Wash - 90 g
Fay Anti Tobacco Air Freshener - 300 ml
Kishwan Dry Cake - 290 g
China Water Bottle (5/23)
Kazi & Kazi Masala Tea - 50 g
Gatorade Soft Drink All Flavors - 250 ml
Alam Baby Dress - 521
Baking Soda - 250 g
Mama Masala Noodles - 248 g
K. Brothers Black Soap - 50 g
Milk Marie Biscuit - 210 g
Lamia Uno Card Plastic - M-08
Rok Tiles & Bathroom Cleaner - 1000 ml
SKB Water Bowl - 18 cm
China Kitchen Strainer (5/23)
Horlicks Women's Plus Jar - 400 g
Kishwan Lexus Biscuit (Buy 2 Get 1) - 216 g
Dove Environmental Defence Shampoo - 340 ml
Fanta Strawberry - 320 ml
Gahana Bruj - 501
Irish Spring Soap - 104.8 g
Spearmint - 13.5 g
Paper Stem Cotton Swab BCB-006
Sakura Plum Dry - 115 g
Prince Diabetic Orange Jelly - 500 g
China Cloth Clip With Rope (05/23)
Mum Drinking Water - 5 L
Beef Tikka Kabab - 10 pcs
Nescafe Original Drinks - 240 ml
Radhuni Vinegar - 540 g
Super White Powder - 1000 g
Igloo Choco Cheers - 1 L
Pokemon Card
Lacys Zip Bag Quart Size - 25 Bags
China Glass Mug 3716 (4/23)
Igloo Vanilla Ice Cream - 1 L
Duck (Rajhash)
Shipon Ornaments - 051
Hurricane Candle Small
Lavino Neem Face Wash - 120 ml
AS Breast Feeling Milk
Safe Hand Liquid Hand Wash Poly (Buy 2 Get 1) - 200 ml
Aluminium Foil Container 12 pcs - 510 ml
Olympic Multi Grain Cookies - 150 g
Refill Pen Anytime Black - 6 pcs
Happy Cheese Slices 12 pcs - 200 g
Shanto Ludu Small - 52
Magic Toothpowder - 100 g
RFL Diamond Mug Blue - 2 L
Denver Body Spray - 165 ml
Fresh Cup Noodles Masala Flavor - 40 g
Nescafe Classic Full Flavor - 50 g
Explorer Toothbrush
Haiko Topping Mix - 70 g
Glow & Handsome Mens - 100 g
Mama Chicken Cup Noodles - 62 g
Nescafe 3-in-1 Hazelnut Latte 4 pcs Combo - 25 g
Diamond Stretch Wrap - 200/60 ft
Dove Pampering Beauty Bar - 90 g
Masafi Apple Juice - 1 L
Tresemme Hair Fall Defense Shampoo - 580 ml
Dan Cake Danish Butter Cookies - 150 g
SS Pressure Cooker - 5.5 L
W7 Eyeliner - 8 ml
Spicy Chicken Nuggets - 250 g
Memory Til Papri - 40 g
Organikare Multani Face Pack - 80 g
Ponds Bright Beauty Cream - 23 ml
Mama Shrimp Tom Yum Noodles - 60 g
Cocola Anarkali Toast Biscuits - 200 g
Kishwan Cookies Jar - 700 g
Nutkao Cocoa & Hazelnuts Spread - 400 g
Pran Turmeric Powder Jar - 200 g
Finis Powder - 100 g
Hi-Speedy Color Cream - 40 g
Memory Chat Masala - 50 g
Suree Sweet Chilli Sauce - 295 ml
Lindt Excellence Extra Creamy Milk Chocolate - 100 g
Umami Tomato Ketchup - 300 ml
Mortein Vap Refill - 45 ml
Himalaya Baby Cream - 50 ml
CBL Munchee Stix Strawberry Wafer - 200 g
Johnson's Baby Lotion Pink - 200 ml
Safety Plus Orchid Floor Cleaner - 1000 ml
Ruchi Olive Pickle - 400 g
Dabur Hajmola Tablet Bottle All - 100 pcs
Food Container - 930 ml
Binimoy Eraser - 8
C.N Multigrain Roti 20 pcs - 1000 g
Skin Cafe Coconut Oil - 250 ml
Cowhead Organic Rolled Oats - 500 g
Parachute Naturale Nourishing Care Shampoo - 170 ml
Savlon Baby Wipes - 120 pcs
Petra Phoenix Graphite Pencil - 2B
Wonder Kids Chocolate
Ferrero Rocher T3 pcs - 37.5 g
Haldiram Kheer Mix - 35 g
Speed Can - 250 ml
Foster Clarks Chocolate Essence - 28 ml
Radhuni Shaadmishali - 96 g
Faysal Fringe Curtain - 08
Almarai Cheese 8 Portions - 120 g
Mediplus Toothpaste - 70 g
Mr. Egg Noodles - 180 g
Haribo Gold Bears Candy - 160 g
Chashi Chinigura Rice - 5 kg
Kwality Jordana Lip Liner NJLC
2 Fun Instant Noodles Masala Flavor - 496 g
Bengal Classic Black Tea - 200 g
Ramisa Ring Chips Kacha - 200 g
Sprite Can Malay - 320 ml
MP Thai Hair Clip - 03
Safety Plus Rose Fresh Hand Wash - 200 ml
Kazi Chicken Teasers - 250 g
Rok Liquid Dishwash - 750 ml
Chopstick Instant Noodles - 8 Packs
Tresemme Shampoo - 450 ml
Faysal Beyblade Toy - 001
Savory Meat Masala - 100 g
Nivea Soft Light Moisturising Cream - 50 ml
Turmeric Powder - 100 g
RS Olive Oil Extra Virgin Tin - 200 ml
Dr. Ruchi Mixed Fruit Jam - 400 g
Nankhatai Biscuit - 200 g
Monia Lock Rectangular Container - 200 ml
Bombay Sweets Til Naru - 15 pcs
Corn Flakes - 275 g
Mediplus Dental Floss - 40 m
Meril Baby Powder Allantoin - 100 ml
Memory Tila Papri - 10 pcs
Golden Harvest Vegetable Spring Roll - 400 g
Kolson Egg Noodles - 180 g
Zess Cheese Cracker - 153 g
Attitude Toothbrush
Rupa Kidline Vest - 28/30
Ponds Oil Control Face Wash - 100 g
Savlon Aloe Vera Hand Wash - 250 ml
Plain Paratha 10 pcs - 650 g
China Tiffin Box (05/23)
Doodles Chicken Masala Noodles - 4 Packs
Cyprina Apple Juice - 250 ml
Fresh Chilli Powder - 100 g
Radhuni Falooda Mix Mango - 250 g
Lemon Bright Liquid Dish Wash - 500 ml
Fast Mac Sweet Corn - 425 g
Pran Mango Bar 30 pcs - 420 g
Beef Pepperoni - 200 g
Savlon Baby Wipes Jar - 160 pcs
China Chakni (5/23)
Buter Dal Bhaja Packet - 200 g
AMA Coffee 3-in-1 10 Sticks - 150 g
Nutella (Buy 2 Get 75 TK Save) - 180 g
Dabur Herbal Toothpaste - 160 g
Sunny Kids Fun Toothbrush
China Lunch Box 3 pcs Set (05/23)
Savory Kala Bhuna Masala - 80 g
Fair & Lovely Ayurvedic Care Cream - 50 g
Cosby Fun Boys - 20 g
Dreem Jelly Orange - 70 g
Trix Liquid Dish Wash - 1000 ml
Calypso Mayonnaise - 500 ml
RFL Globe Jar - 200 ml
Frooto Mango Juice - 500 ml
All Time Sandwich Bread - 600 g
Shan Pilau Biryani - 50 g
Shuvro Detergent Powder (Buy 3 Get 1 Free) - 500 g
Johnson's Baby Soap - 75 g
Ramisa Peara Sandesh Packet - 150 g
Coca-Cola Classic Can - 320 ml
Borges Green Sliced Olives (Buy 1 Get 1) - 230 g
Ruchi Mixed Chutney - 450 g
Mushroom Whole - 425 g
Mirinda Can - 250 ml
Kodomo Head To Toe Wash - 100 ml
Ripon Umbrella - 045
Nivea Soft Moisturiser Cream - 100 ml
Ruchi Alu-Bukhara Pickle - 200 g
CP Chicken Italian Bologna - 340 g
Lifebuoy Care Handwash - 170 ml
Kodomo Baby Bath - 100 ml
MP Hair Kakra Band - 33
Jet Liquid Jar - 1000 ml
China Tiffin Box (5/23)
Pran Fresco Premium Butter - 200 g
Cotton Buds Flat - 120 pcs
RFL Dalia Spice Container - 430 ml
Denim After Shave - 100 ml
Septex Vita+ Antiseptic Soap - 100 g
D. Bhars Palm Candy - 200 g
Pangas Fish River
Minister Lemon One Wash Detergent - 500 g
Dan Cake Chocolate Dry Cake - 300 g
Enchanteur Talc Powder All - 125 g
Skin Cafe Aloe Vera Gel - 240 ml
Spring Air Freshener Flora - 300 ml
Silicone Nipple 6m+ - N-24
Jui Hair Care Oil - 200 ml
Grape Fruit Candy - 100 g
Veet Hair Removal Cream Sensitive - 50 g
PC Matte Chiffon Lipstick
Single Beef Chicken Samosa - 250 g
HIT Cockroaches Aerosol - 400 ml
Fresh Chilli Powder Jar - 200 g
Kishwan Saltine Cookies - 270 g
Gopal Indian Like Me Bra S-34 (4/23)
Beef Salami - 200 g
Davidoff Espresso 57 Coffee - 100 g
Chocolate Vanilla Cup - 40 g
Pusti HT Plain Toast Biscuit - 250 g
Marshmallow Chocolate All - 100 g
Organikare Pink Hand Wash - 250 ml
Surma Fish
Tic Tac Mint - 7.2 g
Gillette Series Moisturizing Gel - 195 g
Butter Delight - 220 g
Aarong Cheese Spread Chili Garlic - 180 g
Italiano Round Lid
Dan Cake Chocolate Swiss Roll - 200 g
Davidoff Dark & Chocolatey Coffee - 100 g
Unik Xtreme Toilet Cleaner - 750 ml
Lux Gardenia Blossom Soap - 170 g
Prodentalb Venus Toothbrush
Ramisa Shapla Chips Kacha - 250 g
BD Kheer Mix - 150 g
Lamia Pokemon Card - 03
Simex Power Mosquito Bat - SP-7718A
Sajeeb Isubgol Bhushi - 40 g
Faysal Party Popper - 08
Garden Fresh Mango Drink - 500 ml
Baby Plate UMC002
Cocola Butter Delight Biscuits - 250 g
Xpert Dishwash Liquid - 250 ml
Pantene Damage Care Shampoo - 340 ml
Nutrilife Pomegranate Fruit Magic - 160 ml
Faysal Foil Balloon Number 02 (4/23)
EM Floral Pasta - 400 g
Faysal Lol Toy Ball 082 (4/23)
China Football (5/23)
Shinrai Japanese Green Tea - 50 g
Faysal Candle 20cm - 22
Zareen Tea - 200 g
RFL Lock Round Container - 600 ml (917819)
Sunsilk Shampoo - 650 ml
Shanto Ludu Big - 05
SK Red Grape Juice - 1 L
Roti Chapati 20 pcs - 1000 g
Vegetable Singara
Haiko Corn Flour - 300 g
Golden Girl Sticker Big - 01
Dal Bhaja - 200 g
Surf Excel Matic Liquid - 1000 ml
Kazi & Kazi Black Tea Family Pack - 200 g
Arla Sandwich Slices - 200 g
Faysal Birthday Cap - 05
Faysal Rainbow Churi - 011
Smart No.1 Gamsa - 05
Red Wheat Roti - 1200 g
Basil Bottle - 10 g
China Pencil Bag (4/23)
Improved Jet Paper Pack - 500 g
Mr. Noodles Magic Masala - 28 pcs
Ruchi Mango Pickle - 400 g
China Water Pot Medium 252 (4/23)
PC Studio Nails - 11.5 ml
Ponds Vanishing Cream - 50 g
Cotton Buds +/- Packet - 50 pcs
Park Choco Wafer - 12 g
Mezban Shahi Borhani - 1500 ml
Oral-B Criss Cross Toothbrush (Buy 2 Get 1)
Milk Pan Bakelite - 16 cm
Clariss Papaya Face Wash - 100 ml
Pusti HT M. Marie Biscuit - 200 g
Fa Vanilla Honey Soap - 125 g
Pran Chicken Roast Masala - 35 g
Margo Soap - 100 g (2 pcs)
Brut Deo Spray Original - 200 ml
Vivo UHT Topping Cake Cream - 1100 g
Toblerone Chocolate - 35 g
Lafz R M Body Spray - 100 g
Glucon-D - 200 g
China Clay Toy TK5106 (5/23)
Muslim Jhal Monakka - 200 g
M&M Peanut Chocolate - 45 g
Tayas Nougat & Caramel Chocolate - 27 g
Mr. Salty Cookies - 240 g
EM Alfabeto Pasta - 400 g
Kodomo Baby Bath - 200 g
Ramisa Chui Semai - 200 g
Foxs Crystal Clear Fruits - 180 g
Nescafe Gold Coffee - 100 g
Premio Spice Container - 1.4 L
Moyen Gamsa - 69
Maxell AAA Battery - 2 pcs
Pepsodent Tooth Powder - 100 g
Karai With Glass Lid - 24 cm
Coriander Powder Jar - 200 g
Strawberry Kids - 80 g
Magic Toothpowder - 50 g
Kishwan Dry Cake Rusky - 290 g
Pineapple Fruit Candy - 100 g
Pantene Smooth & Silky 2-in-1 - 340 ml
Eggy Super Masala Instant Noodles - 480 g
China Vacuum Flask (4/23)
Wild Brew Can - 250 ml
Fresh Masala Instant Noodles (Buy 2 Get 1) - 248 g
Classic Toothbrush
SH Baby Pant - 325
Sundrop Peanut Butter Crunchy - 200 g
Food Cover - 74110
Marlo Water Bottle - 600 ml
China Water Pot 500 ml (4/23)
Happy Kiddo PC Water Bottle - 450 ml
Whisper Ultra Clean XL - 8 pcs
Alpella Nougat Bar Chocolate - 36 g
Golden Girl Party Whistle 12 pcs - 06
Pran Jhatpot Chicken Lollipop - 400 g
Aftab Chicken Spring Roll - 300 g
Pran Jhatpot Chicken Meatball - 400 g
Radium - 72
Teer Borhani Masala - 50 g
Bajaj Almond Drops - 300 ml
Alfa Mayonnaise - 946 ml
Bashundhara Khata P-M-BDK - 84 Pages
AS FL Breast Milk
Naga Chilli Pickle - 300 g
Golden Harvest Econo Paratha 10 pcs - 600 g
Faysal Beyblade Set Tornado 051 (4/23)
Kazi Chicken Keema Paratha - 250 g
Vision Toothbrush
Jhatpat Wonton 10 pcs - 300 g
Special Quality Cream Biscuit Pot - 400 g
Dove Hair Fall Shampoo - 650 ml
Studio X Clean and Fresh Soap - 125 g
Zahir Cap - 071
Haldiram Kabab Masala - 35 g
Himalaya Fairness Kesar Face Wash - 100 ml
China Toy (05/23)
Radhuni Mixed Fruits Jam - 480 g
RFL Tray 11.5
Ruti Tawa Baco - 28 cm
Ajinomoto Tasting Salt - 100 g
Ponds White Beauty - 35 g
Camay Soap - 170 g
Finlay Green Tea Bags - 50 pcs
Chicken Samosa 25 pcs - 250 g
China Mini Clip Fan 631838-1-544
Golden Girl Staedtler Eraser - 01
1 Pack Silicone Nipple MH-013 BX
Spring Air Freshener Lemon - 300 ml
Spark Water Bottle - 700 ml
Godrej No.1 Sandal Soap - 100 g
RS Pomace Olive Oil - 1 L
Wood Apple Big (Sorbot Bell)
7Up Light Pet - 250 ml
Radhuni Butter Chicken Masala - 45 g
Boal Fish Medium
Roja Shrimp Spring Roll - 240 g
Radhuni Khichuri Mix - 500 g
Aroma Dish Cover - 20 cm
Organic Apple Cider Vinegar - 500 ml
Mirinda - 1 L
Dal Puri
Dabur Glucose-D - 400 g
Flormar Just Magic Kajal Liner - 0.35 g
Aarong Low Fat Yoghurt - 500 g
Memory White Chui Packet - 200 g
Rainbow Food Container 4 pcs Set
Emami 7 Oils in One - 100 ml
Pepsodent Toothpaste - 190 g
Junior Horlicks - 500 g
BD White Vinegar - 650 ml
Rubber - 08
Sufia Comb - 04
Tibet Pumpkin Hair Oil - 100 ml
Clear Anti Hair Fall Shampoo - 170 ml
Sanjida Wooden Dabu Spoon - 54
Kishwan Macaroni Pasta - 400 g
Spaghetti Hat Pasta (Buy 1 Get 1) - 400 g
Galaxy Smooth Milk Chocolate - 36 g
China Ceramic Mug 7395 (4/23)
Xpert Dishwash Bar - 300 g
Fresh Egg Masala Stick Noodles - 180 g
Safety Plus Glass Cleaner Spray - 350 ml
Kinder Bueno - 43 g
ACI Puffed Rice - 500 g
Savlon Iris Hand Wash - 1 L
Pencil Cutter - 08
Big Top Lollipop
Grape Black (Jhora)
Chicken Jali Kabab 10 pcs - 300 g
Tema Shrimp - kg
Darlie Toothpaste - 175 g
Harpic Flush Citrus - 2 pcs
Harpic Flushmatic - 50 g
Roja Butterfly Shrimps - 200 g
Vanilla Kids - 80 g
Sensodyne Fresh Gel - 150 g
Kids Chicken Nuggets - 250 g
Foster Clarks Blue Color - 28 ml
Fry Pan Bakelite - 24 cm
Oral-B Ultra Thin Sensitive Green 2+2
Ma SS Glass - 57
Sticker
China Jug (4/23)
Pastiano Lachha Semai (4/23) - 200 g
Coriander Powder - 100 g
Cadbury Fuse Chocolate - 24 g
Faysal UNO Card - 509
Falooda - 150 g
Ponds Pure White Face Wash - 50 g
Cow Brain - kg
ROK Wood Cleaner Spray - 350 ml
Foster Clarks Baking Powder - 225 g
Dr. Rhazes Germ Free Hand Wash - 300 ml
Baby Fatua Set - 08
2.5" Double Color Glass
Reenpark Sweet Corn - 425 g
Islam Water Filter - 5
Monia Lock Rectangular Container - 350 ml
Ramisa Fried Chira Oil Box - 300 g
Teer Mustard Oil - 500 ml
Rubber - 001
Coco Lite Coconut Milk - 400 ml
Puffed Rice Machine - 500 g
AD University Khata 120 Pages - 53
Aer Morning Spray - 220 ml
Pearl Fashion Matte Lipstick - 4 g
Cinthol Health+ Soap - 100 g
Memory Tiler Naru - 200 g
SKB Compartment Plate - 28 cm
Ripon BMW Umbrella - 033
Rubisquib - 522
China Remote Racing Car 548 (5/23)
Supermom Mild Baby Wipes - 120 pcs
Koko Krunch - 330 g
Sunsilk Hijab Refresh Shampoo - 330 ml
Amul Dark Chocolate - 40 g
Prosperity Noodles Tomato
Nestle KitKat Chocolate Packet - 132 g
Sony Alkaline AAA Battery - 2 pcs
Balachaw Dry Chutney - 250 g
Senora Confidence - 10 Pads
Oat Krunch Nutty Choco - 208 g
Polar Caramel Touch Ice Cream - 1 L
Rasna Orange Juice Refill - 400 g
Golap Pitha 5 pcs Packet
Naked Matte Lipstick
Ramisa Nakshi Pitha Box - 4 pcs
One Card Mota - 08
China Magic Slate (5/23)
Sepnil Handwash - 200 ml
Whisper Ultra Night XL+ - 15 Pads
Progrip Toothbrush (Buy 2 Get 1 Free)
Haiko China Grass Milk Halwa
Radish White - kg
ONN Mens Vest - 001
Haiko Tasting Salt - 100 g
Lavino Milk Brightening Face Wash (Buy 1 Get 1) - 100 ml
Ajinomoto Tasting Salt - 42 g
Igloo Cake Ice Cream - 1 L
Teer Firni Mix - 150 g
Safety Plus Lemon Floor Cleaner - 1000 ml
Blue Lady Body Spray - 200 ml
Oral-B Precision Clean Sensitive 2+2
Nivea Soft Cream - 100 ml
Johnson's Baby Cream Milk - 100 ml
Veet Cream Normal - 25 g
Pantene Total Damage Care Shampoo - 180 ml
China Silicone Spoon 58 (5/23)
Lifebuoy Hand Wash - 200 ml
10" Crazy Rice Plate
Savlon Lemon Burst Soap - 100 g
Gahana P. Hair Band - 06
Pran Mustard Oil - 500 ml
Nescafe 3-in-1 Iced Frappe Coffee - 30 g
Dabur Hajmola Borhani Bottle - 100 Tab
Haldiram Mezbani Beef Masala - 50 g
Head & Shoulders Anti Dandruff Lemon Fresh Shampoo - 180 ml
China Milk Pan 12 cm (05/23)
Safe Hand Wash Value Pack Purple - 350 ml
ACI Pure Halim Mix (Buy 2 Get 1) - 100 g
Kazi Paratha 5 pcs - 325 g
China Military Toy Set (05/23)
Memory Misti Monakka Packet - 200 g
Zess Sandwich Lemon Cracker - 180 g
Himalaya Men Intense Oil Clear Face Wash - 100 ml
Organic Baby Oats - 500 g
Julie's Butter Waffles - 100 g
Jussvina Mango Juice With Pulp - 240 ml
Labs Mosquito Repellent Cream - 100 g
Organikare Moharani Face Mask - 80 g
Kill Hit Spray
Tic Tac Red Apple - 7.2 g
Dream Food Color Yellow - 28 ml
SC Banana Shampoo With Egg Protein (Buy 1 Get 1 Free) - 250 ml
Senora Sanitary Napkin Panty - 15 Pads
Faysal Party Spray - 06
Odonil Orchid Dew Air Freshener - 48 g
Lipton Lemon Green Tea 50 Bags - 70 g
Diamond Special Nazir Rice - kg
Clariss Tea Tree Face Wash - 100 ml
Jasmine Green Tea - 60 g
Himalaya Peach Face Wash - 100 ml
Shipon Tops - 011
Head & Shoulders Smooth & Silky Shampoo - 340 ml
Dan Cake Dry Cake - 300 g
Jordan Baby Toothbrush 0-2
Bele Fish Medium (MNK)
Flexible Toothbrush Super Tip
Enchanteur Powder - 125 g
Ramisa Badam Papri - 220 g
China City Bus Toy Pata 793 (6/23)
Aarong Ghee - 900 g
Harpic 750 ml + Harpic Bathroom Cleaner 500 ml (20 TK Off)
SCL Rose Water - 120 ml
Gillette Regular Foam - 98 g
China Fry Pan 26 cm (5/23)
Coca-Cola Zero Sugar - 320 ml
Rupchanda Premium Miniket Rice - 5 kg
Universe Spiral Khata - 300 Pages
MP Hair Kakra Band - 52
MP Safety Pin - 2
MP Safety Pin - 01
AS PP Feeding Bottle - 150 ml
Junior Horlicks Stage 1 - 500 g
Jordan Alfa Soft Toothbrush
Chowa Gamsa - 521
Pusti HT Horlicks Biscuit - 270 g
Mojo Cola Pet - 1 L
Brown Sugar - 800 g
Cerelac Mixed Fruits & Wheat - 400 g
Pran Jhatpot Tandoori Pizza - 210 g
Milk Vita Instant Full Cream Milk Powder - 500 g
Shapla Papor - 150 g
China Hook (05/23)
Dreem Jelly Mango - 70 g
Dove Intense Repair Rescue Shampoo - 330 ml
Partex Cleen Super Soft Toilet Tissue - 4 Rolls
Kodomo 2-in-1 Paste & Toothbrush
Jordan Individual Clean Toothbrush
China Fry Pan 30 cm (05/23)
Prodental B Fish Kid Brush
Golden Girl Party Whistle Small - 05
Fish Fingers - 200 g
ROK Pine Floor Cleaner - 1000 ml
Italiano 5.1" Crazy Soup Bowl
Dove Shea Butter Soap - 100 g
Prince Grape Jelly - 250 g
Alif Tailors Tape - 5 m
Parachute Extra Care Coconut Oil - 300 ml
Harpoon Bathroom Cleaning Powder - 400 ml
China Dress Set (05/23)
Tic Tac Seeds Elachi & Ginger Flavor - 7.2 g
AD Fevicol Super Glue - 3 g
Mezban Shahi Matha - 500 ml
Sajeeb Mr. Chocoz - 200 g
Coca-Cola Rasa Asli - 320 ml
China Ceramic Mug T10 (4/23)
Arpi Deluxe Chocolate Box - 250 g
Nestle Classic Bar Chocolate - 34 g
Ramisa Til Naru - 200 g
Pran Firni Mix - 150 g
Oat Krunch Dark Choco - 208 g
Lactima Cheddar Slice Cheese - 200 g
Fresh Butter Cookies Biscuits Jar - 900 g
Savoy Chocolate Ice Cream - 1/2 L
China Lunch Box (5/23)
Cassava Chips Spicy - 150 g
Alam Baby Fatua Set - 061
Aarong Pasteurized Low Fat Milk - 500 ml
Ramisa Sweet Monakka - 200 g
Garnier Color Naturals Black - 40 ml
China Jug 1.3 L (5/23)
China Shark Toy (4/23)
China Pizza Cutter (5/23)
Queen Jelly Soft Chocolate - 32 g
Sajeeb TS Noodles - 180 g
Act II Golden Sizzle (Buy 5 Get 1) - 50 g
Kabab Sugar - 50 g
Haiko Ice Cream Powder - 160 g
B.T Sunlite AA Battery - 51
Queen Marie Biscuit - 270 g
HIT Mosquitoes Aerosol (50 TK Off) - 475 ml
Safe Hand Liquid Hand Wash 200 ml (Buy 2 Get 1 80 ml)
Premio Spice Container - 500 ml
Chicken Shami Kabab - 250 g
BD Biryani Masala - 40 g
Chocolate Coated Wafer - 24 g
Johnson's Baby Shampoo - 200 ml
Patanjali Aloe Vera Gel - 150 ml
Kool Shaving Foam Ultra Glide - 200 ml
Vivel Soap - 100 g
Fortune Mustard Oil - 1 L
Ginnou Wrapper Chocolate Box - 44 g
Oral B Brush
Cocola Cook Masala Noodles - 360 g
China Grater (5/23)
Whisper Ultra Soft XL - 7 pcs
Fresh Chilli Powder - 200 g
Sunsilk Thick & Long Shampoo - 170 ml
Jamai Pitha - 4 pcs
Parachute Beliful Coconut Oil - 400 ml
China Animal Set (05/23)
Radhuni Orange Flavoring Essence - 28 ml
Johnson's Face Wash - 150 ml
Haiko Bread Crumbs - 200 g
Madina Azoa VIP Dates Loose - kg
Foster Clarks Mango Essence - 28 ml
UM Curry Spoon UM-300
Pran Biryani Masala - 40 g
Vinut Red Grape Juice - 250 ml
Pearl Fashion Nail Polish - 10 ml
Farlin Silicon Nipple in Jar Size L
Agar Agar - 20 mg
Coffee Mug M1 1 pc - 57
Gahana M. Hair Band - 05
Head & Shoulders Cool Menthol 2-in-1 - 340 ml
Cadbury Dairy Milk Silk - 60 g
Hazrat Ali Cheese Bakorkhani - 400 g
Angelic Anti Tobacco Air Freshener - 300 ml
Iced Coffee Latte - 250 ml
Iced Coffee Mocha - 250 ml
4.5" Wave Bowl 78525
Whisper Maxi Fit - 15 pcs
Pure Chilli Powder - 200 g
Aaram Apple Juice - 1 L
Prodental Junior Toothpaste All - 75 g
Fanta Orange Can - 320 ml
Aarong Cheese Spread Pizza - 180 g
Kazi Chicken Pop Corn - 250 g
Lafz Hanan Body Spray Without Alcohol - 90 g
Mouse Glue Trap
Chupa Chups Can - 345 ml
Clariss Milk Face Wash - 100 ml
Golden Girl White Board Marker - 71
Tiffin Box
Suree Oyster Sauce - 295 ml
Golden Girl Mouth Organ - 053
K Grisbi Biscuit - 250 g
DH American E Underwire - 285
Colgate Total Toothpaste - 120 g
Lafz Suk Body Spray - 120 ml
Pran Jhatpat Pizza - 260 g
China Rope 20M 1568 (11/21)
Senora Confidence Teen Sanitary Napkin - 8 Pads
Pastiano Spirali Pasta Poly Pack - 500 g
Clean & Clear Lemon Fresh Face Wash - 100 ml
Polar Cake - 1 L
Shan Chaat Masala - 100 g
Ispahani Premium Green Tea - 70 g
Petra Black Matt Pencil - 2B
Trisa Kid 3-6 Brush 90002
Radhuni Jorda Mix - 400 g
EM Sedano Cut Pasta - 400 g
AD University Khata - 160 Pages
Ripon Baby Umbrella - 521
T. Tuna Flakes Oil - 120 g
Fay Lime Air Freshener - 300 ml
Baby Cotton Buds L-22568 - 200 pcs
ACI Pure Biryani Masala (Buy 2 Get 1) - 40 g
YC Aloe Vera Face Wash - 100 ml
Odonil Air Freshener All - 48 g
Mama Creamy Tom Yum Shrimp Noodles - 60 g
China Scissors FZ01 (5/23)
Fresh Spicy Chicken Flavor Cup Noodles - 40 g
Mini Singara - 300 g
Bombay Sweets Chatur Naru Big Box - 12 pcs
Kopiko Cappuccino Candy - 140 g
Sunsilk Thai Shampoo - 300 ml
Disco Hari No-9
BFL Foil Container 12 pcs - 260 ml
Al-Ansar Sugai Dates Loose - kg
China Kitchen Scissors HY-5003B-021 (5/23)
Nata De Coco Drink All Flavors - 350 ml
Swan Brand Coconut Oil - 350 ml
City Gold Indian Bangles (4/23)
C.N Whole Wheat Roti 20 pcs - 900 g
Cerelac 5 Fruits & Multi Grains - 350 g
Kishwan Sweet Toast - 250 g
Flush Bathroom Cleaning Liquid - 500 ml
China Kata Ball (05/23)
EBY Milk Facewash - 100 ml
Golden Girl Water Game - 09
One Card Patla - 06
Nana Champignons Whole Mushrooms - 425 g
Faysal Sign Pen Set - 521
Nawar Margarine - 250 g
Arla Burger Slices - 200 g
American Harvest Coconut Milk Drink - 290 ml
Europe Grape Chewing Gum - 8 g
Dove Nourishing Shampoo - 330 ml
Pinch Melamine Bowl
Torkari Hari No-8
Curry + Dabu Spoon 301/300
Discovery Sweet Chilli Sauce - 290 g
Trisa Flexible Toothbrush 90008
Dove Nourished Radiance Lotion - 250 ml
C Lamis Pure Black Body Spray - 200 ml
RFL Atta Mesh Small
Muslim Phuljhuri Pitha
Saad Corn Flour - 150 g
Apple Pink Lady - kg
Memory Chira Mowa Packet
Golden Girl Hing - 10 g
Paprika Powder Pot - 15 g
Colour Me Purple Body Spray - 150 ml
Navratna Oil - 100 ml
Zess Biscuits All Flavors - 127 g
Discovery Pitted Green Olives - 330 g
Jaali Khopa Net JKNN
Vanilla Kids 4 pcs Combo - 320 g
9" Juicy Sandwich Tray
Clear Men Deep Cleanse Shampoo - 330 ml
Mama Vagina Gamsa - 701
Premio Colors Kitchen Container All - 5 L
Garden Fresh Mango Drink - 1 L
Prabhuji Soan Papdi - 450 g
Mr. Magic Handwash Powder - 9 g
Lindt Excellence Sea Salt Dark Chocolate - 100 g
Sanjida Spoon - 06
Pran Nazirshail Rice - 5 kg
Golden Rose Velvet Matte Lipstick - 3.5 g
Kaju Delight Biscuits - 220 g
China Salt Pot (5/23)
Sanjida Spoon - 54
Choco Coin Chocolate - 50 g
Chirar Moya - 20 pcs
7Up Can Malay - 320 ml
Chicken Wings & Drumlets - 300 g
Ponds Facial Scrub - 100 g
Clear Complete Active Care Shampoo - 330 ml
Dan Vanilla Layer Cake 12 pcs - 324 g
Monakka Misti - 200 g
Chicken Meatball 30 pcs - 300 g
Coin Strike Chocolate - 20 pcs
MP Hijab Pin - 61
Teer Advance Fortified Oil - 8 L
La Oliva Olive Oil - 150 ml
Pran Butter - 200 g
Calci-Pro Milk Powder - 400 g
Chicken Shami Kabab 10 pcs - 250 g
Dove Hijab Natural Shampoo - 330 ml
Emoji Eraser - 04
Pantene Long Black Shampoo - 340 ml
ROK Citrus Floor Cleaner - 1000 ml
Pori Pakha
Ponds BB Powder - 50 ml
Chunk Pacific Gold Tuna - 165 g
Roja Kids Shrimp Nugget 16 pcs - 240 g
RFL Monia Lock Rectangular Container - 1450 ml (917828)
Savlon Ocean Hand Wash - 250 ml
Almas Prayer Tupi - 011
Marks Active School Chocolate 2-in-1 - 400 g
Golden Girl Eraser - 02
Clear Complete Active Care Shampoo - 170 ml
Savlon Marigold Hand Wash - 200 ml
Birthday Candle - 53
Vegetable Samosa
Aloo Puri
Organikare Petroleum Jelly - 50 ml
Cumin Powder - 50 g
Lamia Rubber - 54
Prodentalb Precision Toothbrush
Trisa Fresh Toothbrush 90011
Ponds BB Cream SPF30 PA++ - 18 g
Bombay Sweets Coffee Nut - 200 g
Milky Bar Mould B-2 (15 TK Off) - 24.5 g
Golden Girl Washi Tape - 50
Gillette Vector+ Razor
Harpic Flushmatic - 2 pcs
Fa Roll-On - 50 ml
KK Orthodox Green Tea - 100 g
Sanjida Wooden Khunti - 04
Hazrat Ali Sweet Bakorkhani - 400 g
Cerelac Rice & Carrot With Chicken - 350 g
Oral-B Neem Extract Toothbrush (Buy 3 Get 1)
PC Cleansing Gel All Skin - 150 ml
PC Porcelain Compact Powder - 12 g
China Spoon Stand (5/23)
BD Mango Pickle - 400 g
Ma Steel Mug - 031
Shakti Strawberry Doi - 80 g
Dan Cake Marble Dry Cake - 280 g
Lizol Jasmine - 500 ml
9" Classic Plate
Ayir Fish Big (HB)
Dove Conditioner - 200 ml
Faysal Back Drop - 08
M&M Peanut Chocolate - 45 g
Kazi & Kazi Masala Tea - 50 g
American BBQ Sauce - 510 g
Pumper - 57
Ma Steel Bati - 05
Dan Cake Cup Cake Delight - 23 g
Polar Premium Doi - 1 L
Cadbury Dairy Milk Silk Hazelnut - 58 g
Italiano Salt Pot 92385
Shan Shami Kabab - 50 g
Quality Smoked Chicken Salami - 200 g
Prosperity Noodles Carrot
Gillette Fusion5 Proglide 2 Cartridges
Golden Harvest Real Mayonnaise - 473 ml
Tea Cup 110 ml 50 pcs Set
Crazy Premium Coup Spoon
Happy Valley Cheese Slices - 200 g
Ponds Mineral Clay Face Scrub - 90 g
Layer Wottagirl SC Body Spray - 135 ml
Glow & Lovely Ayurvedic Care - 25 g
BD Mango Bar 10 pcs Packet - 140 g
Organikare Coffee Scrub - 100 g
Rectangular Air Tight Container - 800 ml
Nior Matt Lipstick - 2.8 g
Savlon Aloe Vera Active Hand Wash - 1 L
Cerelac Apple+Cherry Pack - 400 g
Himalaya Prickly Heat Baby Powder - 100 g
Himalaya Anti-Dandruff Smooth & Silky Shampoo - 375 ml
Kodomo Head To Toe Wash - 200 ml
Pure Square Container - 450 ml
Ligion Henna Powder Box - 100 g
Alpella Crispy Wafer Chocolate - 32 g
Miniket Rice - 25 kg
KS Pearl Moti Mala - 07
B.T Sunlite AAA Battery 2 pcs - 52
Food Cover - 14 cm
Haribo Happy Cola Zourr Candy - 80 g
Jordan Clean Between Toothbrush
Fay Sea Mine Air Freshener - 300 ml
Savory Oregano Plastic Jar - 15 g
Salad Bowl Medium 72702
Savory Nihari Masala - 50 g
Milo Ball - 330 g
Clariss Neem Face Wash - 100 ml
Chilli Powder Jar - 200 g
RFL Monia Lock Square Container 600 ml - 917817
BD Roast Masala - 35 g
N.W. Wasabi Green Peas - 100 g
AD Drawing Book - 05
Kodomo Baby Powder - 400 g
Nivea Men Refreshing Facial Foam - 100 g
Lamia Flashing Light Up Ball - 56
Wave Jasmine Air Freshener - 300 ml
Tata Tea Tetley Premium - 200 g
Garlic Spice Powder - 150 g
Cadbury Cocoa Powder - 250 g
China Floor Mop (5/23)
Ponds Men Acne Face Wash - 100 g
Fresh Whole Wheat Atta - 2 kg
BTC Foil Birthday Banner - 031
Nivea Female Roll-On Fresh Natural - 50 ml
China Kitchen Toy Set (05/23)
China Tawa 28 cm (05/23)
Pran Mango Pickle - 400 g
Doodles Masala Noodles 12 pcs - 744 g
Farmland Suji - 500 g
Atlas Umbrella - 063
Moccona Select Instant Coffee - 100 g
Nido Full Cream Milk Powder - 1 kg
Trix Liquid Dish Wash 500 ml + 250 ml (30 TK Off)
Fresh Chicken Masala Stick Noodles - 180 g
Faysal Note Book - 56
Shan Sindhi Biryani - 60 g
Nescafe Classic Coffee Tin - 100 g
Golden Garden Jelly Bean Jar - 190 g
Nescafe 3-in-1 Hazelnut Latte Coffee - 25 g
RFL Lily Washing Net 18 cm
Bon Appetit Cheese Slices - 170 g
Roja Shrimp Singara - 225 g
Neem Original Honey & Turmeric Soap - 100 g
Naples Mayonnaise - 475 ml
UHT Mango Milk - 125 ml
Foster Clarks Custard Powder Tin - 300 g
Ever Glow Men Intense White Face Wash - 100 ml
Ramisa Athaya - 100 g
Atlas Umbrella - 055
Cadbury Dairy Milk 12 pcs Combo
Tresemme Color Revitalise Shampoo - 580 ml
RFL Lock & Like Container - 700 ml
Pran Aromatic Rice - 5 kg
China Hanger 5 pcs Set 6106 (4/23)
Food Cover Porous/Oval 74109
China Lead Knob 2 pcs Set (5/23)
Pran Frooto Mango Juice - 1 L
Savlon Deep Clean Soap - 100 g
Foster Clarks Banana Jelly - 85 g
Garnier Color Naturals Cream Riche 5.32 - 110 ml
Golden Rose Lipliner Pencil
Dan Cake Pound Cake - 280 g
Universe Spiral Khata Margin - 200 Pages
Pure Turmeric Powder - 100 g
Gillette Shaving Foam - 196 g
Chui Pitha - 200 g
China Glass Mug (5/23)
Savoy Vanilla Ice Cream - 1/2 L
Memory Diabetics Murali - 300 g
Johnson's Top To Toe Baby Bath - 100 ml
Pencil Cutter - 001
Vanilla Adult - 80 g
Golden Harvest Jumbo Vegetable Singara 10 pcs - 400 g
China Slime Toy 1 pc (5/23)
Lakme Eyeconic Kajal - 0.35 g
Fresh Green Tea Bags - 25 pcs
China Socks (5/23)
Clear Anti Hairfall Shampoo - 330 ml
Marble Water Bottle - 650 ml
Bashundhara Khata P-M-SPI - 204 Pages
Farmland Full Cream Milk - 1 kg
Head & Shoulders Shampoo & Conditioner 2-in-1 - 340 ml
Ispahani Dry Cake - 300 g
AD Kippy Clear Glue - 21
Plastic Straws - 100 pcs
Bon Bon Bum Lollipop
Faysal LED Candle - 52
Foster Clarks Cherry Jelly - 85 g
Jet Detergent - 200 g
Septex Everyday Antiseptic Bar Soap - 100 g
M&M - 45 g
Mama Oriental Style Chicken Masala Flavor Noodles - 60 g
Shokti Strawberry Yogurt - 80 g
Lafz Strawberry Lip Balm - 4.5 g
Comfort Fabric Conditioner - 2 L
Italiano 9" Crazy Bowl
Ramisa Jamai Pitha - 150 g
China Vacuum Bottle (4/23)
SH Baby Maggi T-Shirt - 225
Livon Hair Serum - 100 ml
Cadbury Drinking Chocolate - 250 g
Nutella 180 g (Buy 2 Get 1)
Shahi Hot Wheels Car 50561
Ela Vista Pomace Olive Oil - 250 ml
Santa Lucia Spaghettini - 500 g
Gillette Series Sensitive Skin Gel - 195 g
China Silicone Spoon 57 (6/23)
China Rabbit Toy (4/23)
Nut Walker Cashew Nuts - 135 g
Savlon Marigold Hand Wash - 1 L
Faysal Foil Birthday Banner 021 (4/23)
Ma SS Curry Spoon - 52
Do It Body Spray - 200 ml
London Boy Vest XL 55366
BD Coriander Powder - 200 g
Pastiano Macaroni Pasta Poly Pack - 500 g
Zess Chocolate Biscuits - 110 g
Trisa Fresh Super Clean Brush 90006
Vegetable Roll
Jennifer Liquid Lipstick - 5.5 ml
Floor Cleaner Rajanigandha - 900 ml
Paris C. Sun Cream SPF 50+ Vitamin E - 75 ml
Garnier Skin Naturals Hyaluron Serum Mask - 28 g
Darshan Agarbatti - 1 Packet
Schick 2 Exacta Women 5+2
Kurmure Bar-B-Q Chanachur - 300 g
Lizol Lavender Surface Cleaner - 500 ml
Pinkflash Melting Matte Lip Cream - 2.3 ml
Head & Shoulders Neem Shampoo - 340 ml
Boal Fish Big
MP Hair Round Band - 53
Dream Food Color Blue - 28 ml
Golden Rose Fluid Foundation - 34 ml
Foster Clarks Corn Flour Tin - 400 mg
Thai Park Sakura Plum Cherry - 115 g
Chowa Gamsa - 75
Golden Girl White Glue 40 g - 52
Bio Active Lavello Lip Therapy - 3.2 g
China Burger Set (05/23)
Golden Girl Gift Card - 02
Faber-Castell Highlighter Marker - 52
RFL Lock & Like Container 1200 ml - 917827
Foster Clarks Orange - 750 g
Lafz Cherry Blossom Lip Balm - 4.5 g
Maggi Dipping Sauces - 200 g
China Floor Mat 23 (4/23)
Davidoff Fine Aroma Coffee - 100 g
Sajeeb Pancake Mix - 500 g
Ujala Supreme - 50 ml
Improved Jet Liquid Jar - 1000 ml
Navratna Oil - 200 ml
China Pencil Box (4/23)
Britannia Nutri Sugar Free Cracker - 300 g
Finlay Gold Tea - 200 g
Kool Original After Shave Lotion - 50 ml
Shipon Ornaments - 021
ACI Pure Miniket Rice - 5 kg
Spiderman Doll
Prosperity Noodles Egg
Savlon Freedom - 8 Pads
2 Fun Instant Noodles Chicken Flavor - 496 g
VIP Baby Aloe Vera Wet Wipes - 120 pcs
Rice Hari No-7
Mug Medium - 205
Orna House Tupi - 51
Twix - 50 g
Ma SS Tea Spoon - 52
Aer Violet Air Freshener - 240 ml
Dove Beauty Cream Bar - 135 g
China Flowers (5/23)
Hazrat Ali Mawa - 250 g
Crocodile Sando Vest - 508
Organikare Lavender Hand Wash - 250 ml
Wave Rojonigondha Air Freshener - 300 ml
China Duck Toy (06/23)
Cadbury Dairy Milk Fruit & Nut - 137 g
Colgate Vedshakti Paste - 200 g
Dynamite Free Cap Toothbrush
Jui Pure Coconut Oil - 200 ml
Jhatpat Chicken Burger Patty - 4 pcs
Himalaya Strawberry Face Wash - 100 ml
Shokti Strawberry Delight Adult - 80 g
Sunquick Orange Juice - 330 ml
Aftab Chicken Popcorn - 250 g
Dabur Meswak Paste - 100 g
English Anti-lice Shampoo - 125 ml
Haribo Happy Grapes Candy - 80 g
Coca-Cola Light Can Malay - 320 ml
Grass Carp Fish (MNK)
Koral Fish Red - kg
Ponds Bright Beauty - 50 g
Binimoy Eraser - 01
Garnier Men Acno Fight Face Wash - 50 ml
Orange Fruit Candy - 100 g
Kazi Plain Paratha 10 pcs - 650 g
Ever Glow Lemon Face Wash Gel - 100 ml
Crazy Rice Spoon
Golden Girl Nail Polish - 5 ml
Chatpati Masala - 50 g
Kishwan Macaroni Pasta (Buy 2 Get 1) - 900 g
Cadbury 5 Star 3D Chocolate - 42 g
Naturalis Apple-Grape Juice - 1 L
Figaro Green Olives - 340 g
Watermelon Eraser - 01
Royal Elegance Choice Whole Mushrooms - 001
Golden Girl Eye Mask Toy - 52
Shokti Misti Doi - 80 g
Hat Fusilli Pasta - 500 g
Dabur Amla Hair Oil - 200 ml
Haiko Baking Powder Pot - 160 g
Angelic G/V Air Freshener - 300 ml
Hi-Speedy Pro Color Cream - 150 g
Cadbury 5 Star - 24 g
Sun Empire Sunflower Oil - 5 L
Bigen Hair Color Conditioner - 881
Schweppes Dry Ginger Ale - 320 ml
SCL Rajkonna Hair Pack - 50 g
Savory Chicken Masala - 100 g
Enchanteur Deo Roll-On - 50 ml
Dan Cake Cup Cake Delight Vanilla Gift Box - 23 g
Sepnil Skin Color Facemask - 50 pcs
Golden Girl Haunted Face Mask - 53
Pantene Hair Fall Control 2-in-1 - 340 ml
Vanish Toilet Cleaner - 500 ml
Himalaya Oil Control Lemon Face Wash - 100 ml
Oral B (Buy 2 Get 1)
Pran Sweetened Yogurt - 100 g
Sunlite AAA Battery - 2 pcs
Disposable Floss Tooth Picks BB-50006
Push Eraser - 22
Kodomo Baby Shampoo All - 100 ml
Ramisa Antasa Pitha Box - 12 pcs
Chicken Burger Patty 8 pcs - 400 g
Faysal Water Game 001 (4/23)
Groome Green Clay Face Wash - 100 ml
Eagle Max Hit Jumbo Mosquito Coil
Ripon Umbrella - 083
Beef Classic Sausage - 280 g
Taslima Towel - 001
Jussvina Mixed Fruit Juice With Pulp - 240 ml
Technic Mega Lash - 14 ml
Unique Dust Pen SM Blue
10" Soup Plate
University Khata - 200 Pages
Golden Girl Acrylic Color 12 pcs - 022
Lux Velvet Jasmine Soap - 170 g
Miss & Mrs Kajal - 0.35 g
Armaf Tag-Him For Man Body Spray - 200 ml
Elenoor LED Bulb B22 - 12W
Studio X Clean And Fresh Mens Soap 125 g (Buy 2 Get 20 TK OFF)
Ramisa Jhinuk Pitha Box - 250 g
ACI Pure Kabab Masala (Buy 2 Get 1) - 50 g
Masafi Mango Juice - 1 L
J.P Side Clip JSC10
He Pocket Perfume - 18 ml
Roseberry Plum Dry - 115 g
Black Pepper Powder - 150 g
Cekters Doll
Hazrat Ali Salted Bakorkhani - 400 g
Neem Original Pure Neem Soap - 100 g
Ripon Baby Raincoat - 004
Water Lighting Ball Small - 05
Shan Tikka BBQ Masala - 50 g
Cyprina Tropical Juice - 250 ml
Bitter Stick (Cherota) - 50 g
RFL Lock & Like Container 1650 ml - 917826
Latina Mango Juice - 1 L
BD Chilli Sauce - 1000 g
Bactrol Liquid Hand Wash - 200 ml
Chupa Chups Bubble Gum - 12 g
Kellogg's Muesli Fruit And Nut - 500 g
White Tone Face Cream - 25 g
6.5" Star Bowl
Ariel Matic Liquid - 1 L
Memory Badam Papri - 100 g
Kishwan Choco Cookies Jar - 700 g
Memory Sweet Pitha Box
Tresemme Revitalise Shampoo - 580 ml
Cowhead Pure Milk - 1 L
Ayesha Gas Lighter - 56
China Vacuum Cup (4/23)
China Mug BJB525 (4/23)
Reenpark Green Peas - 425 g
Tibet Detergent Powder - 500 g
Cinthol Lime Soap - 100 g
D-Nee Organic Hair & Body Baby Wash - 125 ml
Pears Blue Soap - 125 g
Shuvro Detergent Powder 1 kg (Buy 2 Get 1 Free)
Kellogg's Special Original - 455 g
Elenoor LED Lamp 5W All
Kadom Phool Ball
Lifebuoy Lemon Hand Wash - 200 ml
Haiko Cardamom Powder - 35 g
Organikare Henna Hair Pack - 80 g
Mountain Dew Pet - 1 L
Nescafe Original Coffee Jar - 200 g
Trisa Focus Toothbrush 90009
3 Section Baby Plate 10"
Ulker Laviva Sutlu Dolgulu Chocolate - 35 g
Golden Rose Matte Lipstick Crayon - 3.5 g
Almas Crystal 33 Dana Tasbih
Dettol Original Odour Wipes - 10 pcs
7" Classic Plate
Pepsi Diet - 1 L
Petra Resin Adhesive - 50 g
Mr. Potato Original - 160 g
Fay Magnolia Air Freshener - 300 ml
Suree Low Fat Coconut Milk - 400 ml
Dettol Re-energize Hand Wash - 200 ml
SKB Rice Spoon 01
Suree Premium Soy Sauce - 342 g
Arbella Penne Rigate Pasta - 500 g
Golden Rose CC Cream Primer - 30 ml
Parachute Just For Baby Shampoo - 200 ml
Pastiano Shell Pasta Poly Pack - 500 g
Everest Tea Masala - 50 g
Lakme Blush & Glow Face Wash - 50 g
Nut Walker Mixed Nuts - 135 g
SKB Soup Bowl SS Lid - 14 cm
Pran Dry Cake
EM Sedano Full Pasta - 400 g
Pran Tamarind Sauce - 340 g
Nihar Lovely Castor Hair Oil - 150 ml
Meiji Yan Yan Choco - 50 g
Nescafe Original Can - 240 ml
NAN Optipro 2 Tin - 400 g
Ramisa Phul Pitha Box - 3 pcs
Garlic Pickles - 200 g
Act II Butter Lovers - 33 g
YC Men Extra Whitening Face Wash - 100 ml
Parachute Naturale Nourishing Care Shampoo 340 ml (Plus PNS 170 ml Free)
Panam Deep Clean Facial Wash (Buy 1 Get 1 Free)
Pusti HT Mustard Oil - 500 ml
Golden Girl Taj - 35
Farmland Atta - 2 kg
Birthday Banner - 021
Gahana Bruj - 22
China Silicone Brush & Spoon 221 (5/23)
HIT Mosquitoes Aerosol - 625 ml
YC Men Oil Control Face Wash - 100 ml
Golden Girl Ever Fine Refill Ball Pen
Amigo T-Shirt - 425
Discovery Pitted Black Olives - 330 g
Septex Everyday Antiseptic Soap - 170 ml
Meril Petroleum Jelly - 50 ml
Korma Roast Masala - 35 g
Senora Sanitary Napkin Panty - 10 Pads
Fair & Lovely Multani Mati Soap - 100 g
Strawberry Kids 4 pcs Combo - 320 g
China Apple Cutter (5/23)
Koral Fish XL - kg
Cute Water Bottle - 400 ml
Miss & Mrs Dip Liner - 4.5 ml
Nestle KitKat Chunky Chocolate - 40 g
Vaseline Blue Seal - 50 ml
Tawa
HIT Mosquitoes Lime Aerosol (50 TK Off) - 475 ml
China Staples Set (5/23)
Sunsilk Freshness Shampoo - 375 ml
Fay Lavender Air Freshener - 300 ml
Foster Mixed Fruit Jam - 450 g
Aftab Vegetable Samosa - 300 g
Kaikka River Fish - kg
Twinings Green Tea - 40 g
Ayir Fish XL (HB)
Radhuni Spices Chilli - 1000 g
Ring Papor - 200 g
ACI Pure Mustard Oil - 1000 ml
Savlon Cream - 60 g
Studio X Anti Dandruff Shampoo - 355 ml
Faysal Box Eraser - 05
Fan Apple Juice - 1 L
Clariss Natural Honey - 125 g
Everest Sabji Masala - 50 g
Medle (B+G+T+G) Box - 65 g
Andalus Grease Remover - 500 ml
Clemon - 1 L
Taslima Towel - 79
China Measurement Cup (5/23)
Savoy Wafer Cone Biscuit Box - 12 pcs
Faysal Birthday Balloon - 031
Igloo Butter Scotch - 1 L
Kumarika Hair Oil - 200 ml
AD Fevicol Clear Glue - 81
Pran Mix Fruit Jam - 500 g
One Stone Teep - OST
Chaad Fabric Brightener - 50 ml
Glass 160 ml 50 pcs Set
Glucon-D Tangy Orange Flavor Packet - 450 g
Golden Girl Glue Stick 9 g - 51
Jeba Koral Fish (MNK)
Senora Confidence Folded - 5 Pads
Rafian Check Trouser 4X-5X - 003
China Animals Toy (4/23)
Trisa Young Toothbrush 6+ Years 90003
Bombay Sweets Chatur Naru Small Box - 12 pcs
Party Balloon - 021
Shan Korma Curry Mix - 50 g
Dan Cake Oats Cookies - 150 g
Emami 7 Oils in One Hair Oil Almond - 200 ml
Safety Plus Lemon Fresh Hand Wash - 200 ml
Roja Shrimp Puri - 320 g
Kodomo Baby Lotion - 180 ml
Pangas Fish (THR)
Mezban Shahi Labang - 500 ml
Partex Cleen Kitchen Towel 2-Ply
Foster Clarks Mango - 500 g
Crocodile Underwear - 07
Queen Snow Jelly Choco Packet - 100 g
MP Baby Hair Clip - 001
Wet N Wild Matte Lipstick - 6 g
Vitamin E Hair Capsule - 60 pcs
Kellogg's Chocos Fills - 250 g
Opal 10.5" Plate 6 pcs - P1
Kishwan Jhal Chanachur - 250 g
Fresh Dessert Instant Milk Powder - 500 g
Orkide Sunflower Oil - 5 L
Shokti Plain Doi - 80 g
Munch Chocolate - 21 g
Red King Mens Cooling Oil - 100 ml
Pears Amber Plant Soap - 125 g
Ispahani Green Tea - 25 pcs
Ulker Yupo Meyve Bahcesi - 80 g
Ripon Umbrella - 053
Funtastic Butter Cookies Jar - 800 g
China Cutting Board (05/23)
Alif Measuring Tape - 5 m
Fair & Handsome Cream - 60 g
Savlon Lemon Burst Hand Wash - 1 L
Marks Active School - 400 g
Nail Color Cream - 15 g
Farmland Brown Atta - 1 kg
Suree Fish Sauce - 295 ml
Milk Vita Full Cream Milk Powder - 1 kg
Coffee Maker - 011
Himalaya Glow Rose Face Wash - 100 ml
Teer Mustard Oil - 1 L
Miss & Mrs Nail Polish - 15 ml
Dreem Choco Food Color / Essence - 28 ml
C.N Egg Box Brown - 12 pcs
AS TF Of Breast Milk
10" Bowl
Red Leaf Marker - 04
RFL Easy Dust Pan
Finlay Gold Tea - 400 g
Meril Splash Milk Soap - 100 g
Party Spray - 0553
5" Square Container Snowdrop
ZM Labs Shaving Foam - 200 ml
Golden Girl Joytiti Pencil 3B - 41
Four Seasons Basmati Rice - 1 kg
YC Facial Fit Expert Face Wash - 100 ml
Olio Orolio Massage Oil - 150 ml
Igloo Mango Melody Ice Cream - 1 L
ROK Glass Cleaner Refill - 350 ml
Golden Girl Ever Fine Refill Ball Pen - pcs
Dove Oxygen Moisture Shampoo - 330 ml
Godrej No.1 Jasmine Soap - 100 g
Bashundhara Drawing Khata D-S-GLU - 48 Pages
10" Star Soup Plate
Pastiano Penne Pasta Poly Pack - 500 g
United Melamine Tray 14.5" UM-111
Pran Mega Lollipop - 1 pc
UM Spoon UM-304
Fresh Suji - 500 g
Kishwan Coconut Cookies - 250 g
Oat Krunch Chunky Hazelnut - 208 g
Ifad Eggy Instant Masala Noodles (Buy 2 Get 1) - 400 g
Aer Morning Spray - 270 ml
China Mug GPZB27T (4/23)
Vaseline Cocoa Glow Lotion - 400 ml
Master Plus Bruj - 001
Golden Rose Purifying Cleansing Gel - 150 ml
Kellogg's Chocos - 715 g
Almarai Triangles Cheese - 120 g
Fresh Premium Tea Bags - 50 pcs
Bee Hives Honey - 500 g
Meril Baby Soap - 75 g
Sun Up Drinks All Flavors - 350 ml
BD Hot Chanachur - 140 g
Nestle Gold Corn Flakes Econo Pack - 500 g
Pusti HT Chanachur - 110 g
Foodex Pure Chilli - 435 ml
Sanjida Wooden Beloon - 057
Himalaya Power Glow Face Wash - 100 ml
Khaleque Potato Mixer - 05
Clariss Whitening Face Wash - 100 ml
Johnson's Baby Wash Top To Toe - 200 ml
Sunsilk Hijab Refresh Shampoo - 340 ml
Trisa Flexible White Toothbrush 90010
YC Green Tea Face Wash - 100 ml
Johnson's 24Hour Moisture Soft Cream - 200 ml
Dabur Honey - 500 g
Fresh Coriander Powder Jar - 200 g
Ponds Bright Beauty Facial Foam - 100 g
Teer Haleem Mix - 200 g
Himalaya Blueberry Face Wash - 100 ml
SH Baby Tops - 275
Gatsby Hair Cream - 125 g
5 Star 3D Chocolate - 42 g / 45 g
Nestle Nan Optipro 6-12 Months
Tresemme Keratin Smooth Conditioner - 170 ml
RFL Spark Water Bottle - 550 ml
RFL Mina Container Small White & Red
Supreme Slim Matte Lipsticks - 1.3 g
Ramisa Athaya Pitha Box - 300 g
EM Corni Pasta - 400 g
Ruchi Orange Jam - 250 g
Santan Coconut Milk Powder - 200 g
Lafz Hair Shampoo - 200 ml
Fair & Lovely Glow & Lovely Hand / Glow - 100 g
Yupi Twin Burger - 14 g
MP Thai Hair Stick - 02
Dubai KitKat 17 g (Buy 3 Get 1 Free)
Aer Monsoon Spray - 220 ml
Godrej No.1 Rose Soap - 100 g
Cocola Big Bazar Deshi Instant Noodles - 700 g
Crown Fruit Basket
Savoy Strawberry Ice Cream - 1/2 L
China Tissue Stand (5/23)
Kishwan Butter Cookies Biscuits - 245 g
TS Mozzarella Cheese - 250 g
Gahana Hair Clip - 23
Shokti Strawberry Delight Kids - 80 g
Fay Jasmine Air Freshener - 300 ml
Smart Glass Crazy
China Plate Stand (5/23)
Vegetable Washing Net Assorted - 34 cm
Oral-B Crisscross
Lamia Exam Board 09 (4/23)
Rice Flour Roti - 400 g
Golden Girl Toy Helmet - 59
Maybelline Colossal Kajal - 0.35 g
Frenchs Yellow Mustard - 170 g
Roja Calamari Bites - 200 g
Red Ruby Mango
Lindt Excellence 99% Ultimate Dark Chocolate - 50 g
Dettol 2-in-1 Wipes - 15 pcs
Spring Fresh Orange Air Freshener - 300 ml
Cerelac Honey & Wheat Milk - 400 g
Ama Full Cream Milk Powder - 500 g
Faysal Board Puzzle - 011
Shammi Mango Juice - 1 L
UM Soup Bati UM-005
Shuvro Detergent Powder 1 kg (Buy 3 Get 1 Free)
Kamrul Vacuum Water Bottle - 200 ml
Lafz Apple Cider Vinegar Foaming Face Wash - 100 ml
IKO Sugar Free Pumpkin Crackers - 178 g
Pantene Silky Smooth Care Shampoo - 340 ml
Al Shafi Natural Honey - 227 g
Mr. Hung Tempura Flour - 500 g
Almas Tupi - 07
Ma SS Glass - 501
Heinz Sandwich Spread - 300 g
Flexible Head Toothbrush 90007
Savlon Active Hand Wash - 250 ml
China Fox Toy (4/23)
Shape Up Non Fat Box - 400 g
Cocola Egg & Chicken Bowl Noodles - 65 g
Kazi & Kazi Green Lemongrass Tea - 60 g
Pran Sweet Toast - 200 g
ZeroCal Box - 200 pcs
Garnier Men Shampoo Color Natural Black 10 ml + 10 ml
Herman Peanut Butter Crunchy - 340 g
JFB Face Cream - 100 ml
Golden Rose Eyes Dream Eyeliner
Fresh Turmeric Powder Jar - 200 g
BTC Birthday Banner - 06
Baby Toothbrush 0-3 90001
Boal Fish XL - kg
Ever Glow Papaya Facial Scrub - 175 ml
7Up Light Can - 250 ml
Royal Garden
Saad Cocoa Powder Tin Can - 100 g
Lacys Zip Wrap 30cm X 30m
Prosperity Noodles Pumpkin
China Pepper Mill (5/23)
Ruchi Tomato Ketchup - 1000 g
Foam Ball Rugby
Post Oreos Cereal - 311 g
Suree Rich Creamy Coconut Milk - 400 ml
Aer Fresh Spray - 270 ml
Munchee Lexus Choco Cream Biscuits - 225 g
Lamia Fruits Bubble - 54
Mr. Potato Barbecue Crisps - 100 g
Nissin Gekikara Ramen Hot Carbo - 81 g
6.5" Winter Bowl Blue-Green
Horina Fish - kg
Rupa Frontline Vest L - 001
Himalaya Purifying Neem Face Wash - 300 ml
Neutrogena Sunblock Cream - 88 ml
Nescafe Classic Coffee - 50 g
NAJ Baby Toast Biscuits - 300 g
Cadbury Dairy Milk Roast Almond - 58 g
Igloo Chocolate Ice Cream - 1 L
Kata Ball - 25
Ramisa Jhal Monakka Box - 350 g
Asif Towel Flowers Baby Pant F, L, M - 521
Savoy Mango Ice Cream - 1/2 L
Modhu Sondes - 150 g
Rasna Mango Juice Refill - 400 g
Kool After Shave Lotion - 100 ml
Banana (Champa)
Seven Choco Bean
Oat Krunch Strawberry & Black - 208 g
Savlon Lavender Herbal Hand Wash - 500 ml
Stretchy Anti-Fur Nipple 2 pcs L - AC21011
Ifad Dry Cake - 300 g
Pepsi - 320 ml
Golden Harvest Spring Roll Pastry - 400 g
Evian Natural Mineral Water - 50 cl
Sunsilk Thick & Long - 650 ml
Head & Shoulders Silky Black Shampoo - 180 ml
Shanto Master Chess Board - 001
Finlay Premium Tea - 500 g
Rice Spoon - 305
Curry Powder BIB - 100 g
Roja Deshi Classic Paratha 10 pcs - 650 g
Longkou Rice Vermicelli - 250 g
Horlicks Mother's Plus - 350 g
Cadbury Crunchie Spread - 400 g
Gatorade Drink All - 600 ml
Borges Black Sliced Olives - 230 ml
Litchi Fruit Candy - 100 g
Ispahani Savoury Cookie Biscuit - 140 g
Ramisa Fried Nut White Box - 200 g
Red Cow Butter Oil - 400 g
Cyber Water Bottle - 450 ml
Foster Clarks Orange Essence - 28 ml
Gillette Mach3 Turbo Razor + 4 pcs
Organic Mix Nuts - 400 g
Figaro Green Olives - 450 g
Godrej Lime Soap (Buy 3 Get 1 Free) - 100 g
Zess Peanut Butter Cracker - 230 g
Aarong Low Fat Milk - 500 ml
Petra Color Pencil Half Length - 13
Dan Cake Strawberry Swiss Roll - 200 g
RFL Monia Lock & Like Container - 1100 ml
Organikare Amla Hairpack - 80 g
Chaad Lemon Flavor Dish Washing Liquid - 250 ml
RFL Mina Container
Ovaltine Power Maxx Jar - 400 g
Star Jus Orange - 1 L
Fiama Lemongrass & Jojoba Gel Bar - 125 g
Meiji Hello Panda Choco - 50 g
Munchy's Lexus Choco Biscuit - 200 g
Cake Topper - 07
Avonee Diaper L (9-14 kg) - 5 pcs
China Beauty Toys Set 251 (5/23)
Ruchi Chalta Pickle - 400 g
Mirzapore BOP - 500 g
China Can Opener (5/23)
Pran Apple Jelly - 350 g
Saffron (Jafran) Original - 3 g
Golden Rose Waterproof Liquid Matte Eyeliner - 2 ml
Golden Rose Dream Lips Lipliner
Al Naz 3-Quarter Pant - 053
Julie's Sugar Crackers - 260 g
Shokti Mango Delight Adult - 80 g
AD Note Book - 05
China Ceramic Mug (4/23)
Olay Natural White Day/Night Cream - 50 g
Chicken Lollipop 14 pcs - 300 g
Malka Biryani Masala - 50 g
Khaleque Chocolate Mold - 07
Radhuni Strawberry Flavoring Essence - 28 ml
Ramisa Jhinuk Nimki Box - 300 g
Revive Hairfall Fight Shampoo - 200 ml
China Air Craft Toy (05/23)
China Duck Car Toy (05/23)
Mama Oriental Style Shrimp Tom Yum Flavor Noodles - 75 g
Atlas Umbrella - 073
Cadbury Silk Oreo - 130 g
Yusera Hand Wash Refill Spain - 170 ml
Almas Nivea Attar - 6 ml
Johnson's Baby Cream Pink Jar - 50 g
Shokti Mango Delight Kids - 80 g
Haiko Baking Powder - 150 g
Garnier Men Shampoo Color Brown 10 ml + 10 ml
Hair Punch Clip HPC
Kamrul Chopsticks 1 pc - 52
Clemon Lemon Drink - 2 L
Kurmure Jhal Chanachur - 300 g
Rupa Frontline Vest 2XL - 501
Polo Mint - 12 g
Saffola Honey - 250 g
Shipon Ornaments - 081
A.K Check Trouser - 052
SCL Moisturizing Conditioner - 120 ml
Fogg Czar Scent - 30 ml
Good Maid Dish Washing Paste - 400 g
Aafi Mango Drink Pet - 1 L
Pran Drinko Float Mango - 250 ml
Glow & Handsome Cream Mens - 25 g
Kellogg's Oats - 900 g
BBQ Marshmallow Chocolate - 90 g
Cowhead Low Fat Pure Milk - 1 L
China 3D Scenery (01/23)
Gillette Simply Venus Razor - 4 pcs
Kwality TikTik Clip LG TTL
Lipton Lemon Green Tea 25 Bags - 35 g
Marks Active School Full Cream Milk - 400 g
Fay Narcissus Air Freshener - 300 ml
China Tawa 30 cm (05/23)
Sesa Ayurvedic Hair Oil - 50 ml
Kellogg's Crunchy Bites Chocos - 375 g
Apple Fruit Candy - 100 g
Complan Chocolate Box - 350 g
Faysal Fairy Wings - 09
Good Knight Fabric Roll-On - 8 ml
Mr. Potato Hot & Spicy Crisps - 100 g
Ariel Front Load - 500 g
Midnight Rat & Mouse Killer - 60 g
Pantene Silky Smooth 2-in-1 Shampoo & Conditioner - 180 ml
Ovaltine Malted Milk Jar - 400 g
Garnier Color Naturals Brown - 40 ml
Organikare Glycerine - 60 g
Wonder Dish Wash Bar - 325 g
China Silicone Brush 09 (5/23)
Frenchs Classic Yellow Mustard - 226 g
Whisper Ultra Nights XL Pads
Cadbury Dairy Milk (Buy 3 Get 10 TK Off) - 24 g
Lacys Zip Bag - 25 Bags + 5
Faysal Birthday Banner 57 (4/23)
Complan Natural Box - 350 g
Baby Raincoat - 003
KitKat 4 Finger Dubai (Buy 3 Get 1)
Mr. Bowl Magic Masala Noodles - 60 g
Balloon Dami - 50 pcs
Remia Garlic Sauce - 512
Bashundhara Khata P-BOU(K) - 124 Pages
Hit Anti Mosquito Racquet
Sondesh - 100 g
Cyprina Mango Orange & Apple Juice - 1 L
Nivea Protect & Care Roll-On - 50 ml
Sepnil Tea Tree Oil Sanitizing Hand Wash - 200 ml
Yum Layer Chocovail Cake - 400 g
Grapes White (Australia) - kg
Ultra Twin Toothbrush
Fogg Paradise Body Spray - 120 ml
Tomyum Mushrooms Whole - 425 g
Vita Orange Juice - 1 L
BD Tomato Sauce - 950 g
China Foam Bra 3189 (04/19)
Clear Anti Dandruff Shampoo - 80 ml
Wave Lemon Air Freshener - 300 ml
Hosen Quality Mushroom Whole - 425 g
Mama Oriental Style Chicken Flavor Noodles - 60 g
RFL Rectangular Spice Tray Red
Kellogg's Chocos - 1.15 kg
Clinic Plus Strong & Long Shampoo - 175 ml
Alam Baby Fatua - 051
China Bowl (4/23)
Studio X Brightening Face Wash - 100 ml
China Food Poly Bag BX-8113 (11/21)
Fiama Blackcurrant & Bearberry Soap - 125 g
RFL Diamond Jar - 2.5 L
Camay Soap - 165 g
10" Curry Bowl Violet
China Insulated Mug 8284 (05/23)
Rok Lemon Fresh Dishwashing Liquid - 500 ml
Haiko Tandoori Chicken - 40 g
Black Stone Agarbatti
FE Birthday Mukut Taj - 021
Plate 16 cm 50 pcs Set
Vixol White Bathroom Cleaner - 900 ml
Nut Walker Roasted Salted Cashew Nuts - 250 g
China Kitchen Chimta (5/23)
Ulker Albeni Viva Chocolate - 36 g
Mango Kids - 80 g
BD Turmeric Powder - 200 g
Saad Monosodium Glutamate Foil Pack - 454 g
Petra Scale - 30 cm
Julie's Finger Lemon Sandwich - 110 g
Elenoor LED Lamp B22 - 9W
Italiano 10" Coup Plate
China Weight Scale (4/23)
Julie's Choco More Sandwich - 132 g
China Lunch Box 6123 (5/23)
BTC Birthday Cap KomDami - 51
Matador Orbit Oil Gel Pen - 21
Faysal Dominoes Toy - 220
Sajeeb Orange Soft Drink Powder - 500 g
Beef Roll
Sufia Comb - 09
BD Garlic Pickle - 400 g
Glucon-D Regular Jar - 500 g
Dove Pink Beauty Cream Bar - 90 g
Assorted Design Scarf Pin PSAP - 1 pc
Meiji Hello Panda Strawberry - 50 g
Shokti Vanilla Delight Adult - 80 g
SCL Groome Eyebrow & Facial Razor
AD 2" Packing Tape - 04
RFL Lock & Like Container 1100 ml - 917830
AS Baby Powder Puff
RFL Ruby Oil Jar - 450 ml
Shan Chili Pickle - 300 g
NAJ Dry Cake Biscuits - 300 g
Golden Girl Plastic Puzzle Toy - 09
Fair & Lovely Body Milk - 200 ml
Avonee Diaper XXL (14-25 kg) - 4 pcs
Cerelac Wheat With Apple & Corn Flakes
Utsho Tri-c Orange Drink Powder - 500 g
Nescafe Gold Coffee - 50 g
SK Special Orange Juice - 1 L
Strainer Red 10 cm - 91332
Chupa Chup Babol 16 pcs - 64 g
Faysal Baby Star Rattle Set 012 (4/23)
Lacys Zipwrap 30cm x 60m - 200 ft
Parachute Just For Baby Soap - 125 g
ACI Freedom Sanitary Napkin - 15 Pads
Coin Box Chocolate - 150 g
Foster Clarks Lemon - 750 g
Pran Garlic Pickle - 300 g
Beautina Neem & Aloe Face Wash - 100 ml
Lifebuoy Care Hand Wash - 200 ml
RFL Merina Jar - 2 L
Grape Red (Jhora)
Bru Pure Instant Coffee - 50 g
Haribo Peaches Candy - 80 g
Johnson's Milk Rice Baby Lotion - 100 ml
Blue Lady Deo Body Spray - 200 ml
Reviva Lemon Green Tea Bags - 45 g
CBL Munchee Next Chocolate Wafer - 225 g
Lindt Excellence Mint Intense Dark Chocolate - 100 g
Kumarika Hair Oil - 415 ml
K. Brothers Black Soap - 120 g
YC Neem Face Wash - 100 ml
Teer Beef Masala - 100 g
Fontana Orange Juice - 1 L
China Car Set (05/23)
Italiano Baby Mug Without Pac
China Scissors KC-010 (5/23)
Golden Girl Correction Pen - 52
Kishwan Ghee Toast - 200 g
Organikare Chandan Ready To Use - 80 g
Sesa Ayurvedic Hair Fall Oil - 100 ml
ZM Vitamin C Foaming Face Wash - 100 ml
Haiko Chilli Flakes - 40 g
Pantene Pro-V Shampoo - 375 ml
Rajkonna Multani Mati - 80 g
Kamrul SS Fridge Handle Cover B-09
Microwave Cover - 08
Studio X Anti Dandruff Shampoo - 175 ml
Shezan Hot Tomato Sauce - 1 kg
Arpi Premium Chocolate Packet - 500 g
Aaram Orange Juice - 1 L
CP Plain Chicken Ball - 500 g
Igloo Ice Cafe Premium Ice Cream - 1 L
Borkot Spoon - 52
Parachute Just For Baby Face Cream - 50 ml
Axe Signature Pocket Perfume
Nut Walker Peanuts Roasted & Salted - 150 g
Golden Girl Cake Topper - 54
Shezan Choco Crunch - 330 g
Salad Bowl Large 72703
Kwality Metal Stone Khopa Kata MSKK
MP Hair Clip - 011
Mozzarella Cheese Puffs - 20 g
Coffee Mug Assorted
Garnier Color Naturals Cherry Reds 6.60 - 100 ml
Cyprina Cranberry Juice Drink - 1 L
Guerniss Everyday P. Lotion - 60 ml
UM Baby Plate 9.25" UM-426
Meiji Yan Yan Choco - 50 g
Farm Fresh Mango Milk - 250 ml
RFL Heavy Rehal
PediaSure Complete 1-10 Years - 900 g
Sufia Comb - 06
Twing Orange - 500 ml
China Charger Fan (4/23)
Studio X Ultimate Hold Hair Gel - 100 ml
Lamia Color Ribbon 03 (4/23)
Barilla N.65 Pasta - 500 g
China Air Tight Jar 8284 (05/23)
RFL Bela Ruti Box Big
China Silicone Khunti 06 (5/23)
Golden Girl Hurricane Candle Small - 032
ACI Pure Cumin Powder Jar - 200 g
Golden Rose Longstay Liquid Matte Lipstick - 5.5 ml
Reggia Tofe Macaroni Pasta - 500 g
Corn Flakes Real Honey - 300 g
Radhuni Spices Turmeric - 1000 g
Beels Almond Oil - 70 ml
Aquafresh Toothpaste - 125 ml
Vaseline Petroleum Jelly - 100 ml
Mr. Potato Original - 40 g
Dr. Rhazes Lavender Hand Wash Bottle - 300 ml
Lamia Metal Car Toy - 501
Savlon Antiseptic Shield Hand Wash - 500 ml
L'Oreal Hard Hair Wax - 150 ml
RS Pomace Olive Oil - 500 ml
PC Long Lasting Lipliner - 310
PC Matte Edition Rouge Lipstick
Havoc Silver Body Spray - 200 ml
Ripon Umbrella - 004
Kool Shaving Foam - 100 ml
Purnava Chini Go - 30 Sachets
Vegetable Samosa - 8 pcs
Faysal Bag Eraser - 02
Sesa Hair Oil - 100 ml
Golden Girl Fishing Game Toy - 490
Saltcheese Crackers Biscuits - 200 g
BD Haleem Mix - 175 g
Bisconi Mughal Cookies - 225 g
China Ultraman Pencil Bag (4/23)
Clean & Clear Tub - 100 ml
Golden Girl Water Color - 57
Emami 7 Oils Coconut Hair Oil - 300 ml
T. Twister Apple Fruit Drink - 1.5 L
Seylon Pyramid Tea Bags - 100 g
Pangas Fish Big River
Godrej Aer Car Freshener Gel Blue
Golden Garden Red Plum - 200 g
Danish Turmeric Powder - 500 g
KS Baby Nima - 05
Kazi & Kazi Tulsi Tea - 60 g
Gillette Vector Plus - 2s
Dan Cake Cappuccino Muffin - 40 g
PC Kissproof Lipcolor
May Juice All - 750 ml
Fontana Juice - 1 L
NAJ Horlicks Biscuits - 300 g
Rupa Kidline Vest - 24/26
Shokti Mango Yogurt - 80 g
Mixed Herbs - 10 g
Amigo Mens Underwire - 175
Capsicum Red - kg
Trisa Swiss Clean Toothbrush 90040
Ruchi Mixed Fruit Jam - 250 g
Lipton Lemon Green Tea 50 Bags - 65 g
Garnier Color Naturals Burgundy 3.16 - 40 ml
Organikare Glycerine - 120 g
Malka Afghani Seekh Kabab - 50 g
Improved Jet Pack - 200 g
Blender's Choice Tea - 100 g
Hup Seng Cream Crackers - 225 g
Golden Harvest Burger Patty - 200 g
Fried Chicken Drumstick - 500 g
Islam Funnel - 51
Sunsilk Hairfall Solution Shampoo - 170 ml
Fogg Tycoon Scent - 30 ml
Clariss Rave Culture Fragrances - 100 ml
RS Almond Oil Glass Jar - 90 ml
Nescafe Original Coffee - 200 g
China Toys Set 294 (5/23)
Faysal Pokemon Trading Card - 061
Kellogg's Muesli Fruit Magic - 500 g
Haldiram Pink Salt - 454 g
YC Papaya Face Wash - 100 ml
Discovery BBQ Sauce - 490 g
Foster Clarks Soda - 150 g
Colgate Plax Gentle Care - 250 ml
Golden Girl Hurricane Candle Big - 003
Clorox All - 1 L
Cerelac Rice With Milk - 400 g
Dan Cake Orange Cake - 220 g
Dabur Hajmola Tablet Bottle All Flavors - 100 Tab
Olila Glass Set 6 pcs W026
BD Aromatic Chinigura Rice - 1 kg
Herman Mayonnaise All Flavors - 500 ml
Nestle Coffee Mate Jar - 400 g
Foster Clarks Banana Essence - 28 ml
Parachute Naturale Anti Hair Fall Shampoo - 340 ml
Shipon Chain - 071
China Nail Cutter Set T6489 (3/22)
Quaker Oats Pouch Pack - 450 g
Potato French Fries - 500 g
China Fry Pan (05/23)
Maggi Cooking Sauce - 200 ml
Complan Chocolate Jar - 200 g
Ovaltine Power 10 Jar - 400 g
YC Bamboo Charcoal Black Mask - 100 ml
Domee Baby Corn Spear - 425 g
Blue Lady Deo Roll-On Rasasi - 50 ml
Golden Rose Facial Cleansing Milk - 200 ml
Johnson's Baby Cream Jar - 50 ml
Oat 25 Ten Grains - 200 g
AD Ring Khata - 300 Pages
Nut Walker Roasted Salted Cashew Nuts - 120 g
USA Beauty Care Face Out Soap - 120 ml
AD Chora Book - 55
AS Liquid Silicone Pacifier
Lafz Summer Twist Deodorant - 50 ml
Xpel Dual Action Aerosol - 475 ml
Moong Dal Bhaja - 200 g
Saffola Active Oil - 5 L
BD Tamarind Chutney - 500 g
Rain Shower Gel - 250 ml
AD Nataraj Pencil - 01
SKB Plate - 26 cm
Kellogg's Muesli No Added Sugar - 500 g
Reggia Fusilli Macaroni - 500 g
Baby Cotton Buds L-22569 - 200 pcs
Colgate Plax Complete Care - 250 ml
Garnier Men Turbo Bright - 50 g
ACI Fun French Fries - 180 g
Alam Baby Frock - 024
Everest Hingraj Powder - 25 g
Baby Bowl Set
Lamia Hand Pakha 061 (4/23)
Petra Color Pencils Round Tin - 13
China Cake Mould (11/21)
Marble Cake - 300 g
Golden Girl Ever Fine Liner Pen
Original Listerine Mouth Wash - 250 ml
Prodentalb Orange Breath Spray
Slimming Tea Herb Bags - 50 g
Hosen Sweet Corn - 400 g
Candle
Armaf High Street Body Spray - 200 ml
Erko Mallow Plus Mallow Stick - 20 g
Eurofresh Kids Sensitive Toothpaste - 50 ml
Fanta Orange - 320 ml
Johnson's Baby Milk Lotion Malaysia - 500 ml
City Gold Indian Ear Ring (4/23)
Nut Walker Thai Lime & Chili Cashew Nuts - 130 g
RFL 12" Rice Bowl 919862
NAJ Special Toast Biscuits - 300 g
9" Snowdrop Bowl
Ivory Oil Jar - 1 L
Cadbury Drinking Chocolate - 250 g
EBY Neem Facewash - 100 ml
J.P Leather Baby Belt - 061
EBY Lemon Facewash - 100 ml
Pran Miniket Rice - 5 kg
Wave Lavender Air Freshener - 300 ml
Alif Custom Padlock - 63 mm
Rui Fish Burmese
Foster Strawberry Jam - 450 g
Dove Deeply Nourishing Body Wash - 1 L
Omuda Silver Nailcutter 3061
Memory Badam Topi Box - 250 g
Gillette Mach 3 Base Cartridge - 2
Johnson's Baby Soap - 150 g
Starship Mango Juice - 1000 ml
Kiam Classic Pressure Cooker - 4.5 L
Lavish Vegetable Biscuits - 200 g
Mister Potato Barbecue Crisps - 40 g
Whisper Ultra Clean Soft XL+ - 30s
Mister Potato Original - 75 g
Chicken Tandoori Nuggets 10 pcs - 250 g
Everest Pani Puri Masala - 50 g
Garnier Pure Active Wash - 100 g
China Egg Beater (5/23)
Armaf Spring Violet Air Freshener - 300 ml
Popular Square Container Set B78807-1
12.6" Rectangular Handle Tray 92419
Unique Dust Pan Red
Atlas BMW Long Umbrella - 083
Farlin Plastic Stem Cotton Buds - 200 Swabs
Golden Rose Full Cover Matte Foundation - 30 ml
Gillette Lemon Shaving Foam - 196 g
Mirzapore Tea - 350 g
Savlon Detergent Powder - 500 g
SCL Groome Cleansing Nose - 6 Strips
Mama Oriental Style Chicken Masala Flavor Noodles - 85 g
Angel Khata English - 120 Pages
Mango Adult - 80 g
Olio Orolio Extra Virgin Olive Oil - 250 ml
Golden Girl Joytiti Oil Pastel - 011
Golden Girl Party Sunglass - 5021
SKB Small Bowl - 9 cm
Calypso Mustard - 200 ml
Johnson's Baby Hair Oil India - 100 ml
Fish Papor - 100 g
China Dinos Trans Gear Toy (4/23)
Chicken Roll
China Fry Pan 22 cm (05/23)
Veet Hair Removal Cream - 100 ml
Faysal Gel Hurricane Candle - 022
Fiama Patchouli & Macadamia Soap - 125 g
Sufia Hair Brush - 052
China Hair Punch Clip (5/23)
Shokti Plain Yogurt 4 pcs Combo - 320 g
Lamia Fruits Bathing Toy - 043
Groome Sweet Sunrise Glow Lip Balm - 30 g
Igloo Strawberry Sparkle - 1 L
Sunsilk Shampoo - 350 ml
BD Tandoori Chicken - 35 g
Fresh Biryani Masala - 40 g
BD Kasundi - 290 ml
Cerelac Wheat & Milk - 400 g
Dream Food Color Pink - 28 ml
Jet Classic Detergent Paper Pack - 500 g
Cusson Baby Daily Shampoo - 200 ml
Coriander Leaves - 100 g
Sunquick Guava & Strawberry Juice - 330 ml
Cerelac Rice Potato+Chicken Pack - 400 g
China Lunch Box 6556 (5/23)
Shipon Stone Ear Top - 031
Merry Go Round
RFL Monia Lock Rectangular Container - 825 ml
Lafz Ocean Crest Deodorant - 50 ml
Thai Park Roseberry Plum - 200 g
One Time Rice Spoon - 6 pcs
SKB Water Bowl - 12 cm Deep
Golden Girl Oil Pastels 18 Color - 031
Masafi Orange Juice - 1 L
Cadbury Choclairs - 192 g
Staedtler Mars Lumograph HB Pencil - 1 pc
Islam Water Filter Tap - 5
China Rabbit Dance Toy (5/23)
Domee Sweet Corn - 425 g
Dr. Rhazes Foot Cream - 50 g
Lindt Excellence 70% Cocoa Mild Dark Chocolate - 100 g
Golden Girl Diamond 12 pcs Color Pencil - 061
Colgate Advanced White Paste - 135 g
Kamrul Pen Holder - 051
3.5" Winter Bowl 75785
Blue For Men Deo Roll-On Rasasi - 50 ml
Pran Kacchi Biryani Masala - 45 g
Swiss Garden Young Baby Corn In Brine - 425 g
Nababi Sweetened Bakorkhani - 400 g
Gift Bag - 06
China Coffee Mug 21582-66 (5/23)
Foster Clarks Banana Jelly Powder - 70 g
Pran Tomato Ketchup - 1 kg
Parachute Skin Pure Brightening Aloe Vera Face Wash - 100 g
Shokti Mango Doi - 80 g
Rubber - 06
Shakti Toilet Cleaner - 750 ml
Golden Girl Color Paper - 5.1
Johnson's Baby Soap Thai - 75 g
Cake Gummy - 9 g
Jordan Ultimate You Toothbrush
Mother's Smile Prima 2 (6-24) - 400 g
RFL 11" Rice Bowl 919863
Hair Code Mehendi Creme Color - 25 g
SCL Lightweight Sunscreen - 60 g
Parachute Just For Baby Wash - 200 ml
Galaxy UB Magnetic Chess Board 4812
BTC Gel Candle B-06
Finlay Green Tea Box - 125 g
Farm Fresh Ghee - 450 g
Boro Plus Neem & Aloe Vera Soap - 100 g
Century Tuna Vegetable Oil - 165 g
MP Hair Small Kakra - 05
Camelia Spice Container Transparent - 750 ml
Premio Queen Soap Case
Mango Juice
Plastic Kula Small
8" Bamboo Sticks 81
Taar Hair Band THB
Sanjida Wooden Dal Ghutni - 56
Whisper Ultra Soft XL - 30 Pads
Salt Pot 3 pcs 007
Clariss Mushroom Whole - 425 g
Jordan Buddy Toothbrush 5-10
Trisa Natural Toothbrush 6+ Years 90043
Yum Dry Cake - 300 g
NAJ Ghee Toast Biscuits - 300 g
Shan Achar Gosht Masala - 50 g
Ripon Umbrella - 033
Lafz Hair Color - 130 ml
Gillette Fusion 5+1 4 Cartridges
RFL Belly Jug - 2.3 L
Sepnil Face Mask - 50 pcs
Queen Jelly Umbrella Chocolate - 32 g
Golden Girl Glue Stick 36 g - 53
Shipon Ear Tops - 021
China Fashion Pencil Bag (4/23)
Faysal Floating Candle - 31
Aroma Air Tight Low Container 3 pcs Set
Dove Shampoo - 250 ml
AD Rong Tuli - 051
Atlas Umbrella - 513
SKB Plate - 24 cm
KitKat 2 Finger - 18.5 g
PC Natural Look Foundation - 30 ml
Flormar Waterproof Eyeliner Pencil All
PediaSure Vanilla Milk - 200 g
China Mermaid Toy (06/23)
Star Jus Orange - 1 L
BD Orange Jelly - 500 g
Vaseline Healthy White - 400 ml
7" Snowdrop Soup Plate
10" Snowdrop Star Bowl
Complan Chocolate Jar - 500 g
Pran Meat Masala Powder - 100 g
Muslim Til Tana - 80 g
Crocodile Underwear - 58
Arla Low Fat Milk - 1 L
Zess Cream Crackers - 380 g
Nababi Coconut Bakorkhani - 400 g
Shanto Monopoly Game - 024
Dabur Red Paste Super Saver Pack - 300 g
Rupa Underwear Size-XL/XXL - 57
Lakme Sun Expert Cream - 100 g
Cotton Buds Round - 140 pcs
Ever Glow Milk Face Foam - 100 ml
Line Ocean Green Gel Pen - 21
Senora Sanitary Napkin Belt - 10 Pads
Bossini TS-100 Telephone - 057
Lakme Eyeconic Kajal
Beef Bacon - 200 g
Prince Diabetic Orange Jelly - 500 g
Bombay Sweets Coconut Naru - 15 pcs
Kwality Queen Makeup Puff
Lifebuoy Total 10 Hand Wash - 5 L
Kamrul Oven Cover - 051
Fiama Peach & Avocado Soap - 125 g
Jordan Target White Toothbrush
Vaseline Deep+Healthy Lotion (Buy 1 Get 1) - 400 ml
Haiko Bar-B-Q Masala - 40 g
Golden Garden Black Prune Plum - 200 g
Anita Indian Captains Vest (4/23)
Everest Dry Mango - 50 g
Golden Girl Hero Attack Face Mask - 091
Bed Bug Killer Spray - 250 ml
Atlas BMW Umbrella - 004
Depa Hijab Tupi - 51
Lavino Milk Brightening Face Wash - 100 ml
Ramisa Bundia - 200 g
Trisa Pearl White Toothbrush 90014
China Soap Case (05/23)
Al Shafi Honey - 125 g
China Silicone Brush Set 88 (5/23)
Moha Shoal Fish XXL Kata
Alam Baby Dress - 062
China Small Rabbit Toy (06/23)
IKO 3 Seeds Biscuits - 178 g
Sufia Comb+Mirror Set - 05
Salted Pistachios Can (Buy 1 Get 1 Free) - 110 g
China Glass Mug 5609 1 pc (4/23)
SCL Rajkonna Light Moisturizer - 50 ml
I.S. Ladies Tights - 051
Khaleque Cake Cup - 54
Aseel Vegetable Ghee - 500 g
5" Spring Bowl 75752
Omuda Nailcutter 300152
Premio Clean Paper Basket All
Hosen Honey - 400 g
Crocodile Underwear XL - 57
Chupa Chups Chew Candy - 45 g
Gillette Menthol Shaving Foam - 196 g
Wild Stone Body Spray - 150 ml
Moha Shoal Fish Small
Sunny Double Pack Toothbrush
SCL Rajkonna Amla Powder - 50 g
Thai Rambutan Fruit
Alam Baby Frock Set - 031
Leggos Alfredo Cheese & Chives Pasta Sauce - 490 g
Fettuccine Pasta - 500 g
Flash Toilet Cleaner - 500 ml
Jhatpat Vegetable Spring Roll - 9 pcs
Kiam Queen White Pressure Cooker - 2.5 L
RFL Lily Washing Net 37 cm
Masafi Mixed Fruits Juice - 1 L
Spaghetti Pasta - 500 g
RFL Ivory Oil Jar - 750 ml
China Water Pot Medium 674 (4/23)
Foster Clarks Quick Cooking White Oats - 500 g
10" Double Color Coup Plate
A.S. Baby Nail Clipper AB-627
Golden Girl Birds Puzzles Toy - 011
Dettol Skincare Body Wash - 250 ml
Lamia Tent Play Toy 50 pcs - 056
Cerelac 5 Vegetable & Multi Grains - 350 g
Faysal Lol Toy 084 (4/23)
Nestle Cerelac Baby Rice - 200 g
Shanto Bagaduli Board Game - 054
Foster Clarks Creme Caramel - 71 g
AD VIP Khata 160 Pages - 52
Golden Rose Matte Liquid Lipstick - 7 ml
Sufia Ayna & Chiruni - 001
Memory Jhuri Bhaja Box
Kiam Wok Pan - 28 cm
Munchee Lexus Choco Cream Biscuits - 200 g
Nut Walker Dry Roasted Salted Pistachios - 35 g
Faysal Yo-Yo 002 (4/23)
Water Color Pen 12 pcs - 06
6.5" Curry Bowl
Lurpak Salted Spreadable Butter - 250 g
La Vache Qui Rit - 16 pcs / 240 g
Kids Coin Saver 87978
Milk Pan With Glass Lid - 14 cm
Vatika Enriched Coconut Hair Oil - 150 ml
Oillina Olive Oil - 100 ml
Studio X Clean And Fresh Mens Soap 75 g (Buy 2 Get 15 TK OFF)
Ginnou Chocolate Box - 176 g
12" Oval Dish
MP Hair Mou Kakra - 001
China Animal Set (05/23)
Cyprina Orange Juice With Vit Cae - 1 L
Sunsilk Thick and Long Shampoo - 450 ml
Coffee Mate - 450 g
Golden Garden Sakura Plum - 200 g
Mango Arshini Juice
Shezan Tomato Ketchup - 1 kg
Creme 21 Cream - 50 ml
Cadbury Choclairs - 196 g
Memory Custard Powder Tin - 150 g
Savory Bhuri Bhuna Masala - 35 g
RFL Kaniz Spice Jar 91010
Thai Jar - 2 L
Ifad Nutri Bite Vegetable Crackers Biscuits
Krystal Klean Glass Cleaner - 350 ml
Ma Steel Bowl - 56
Himalaya Anti-Hair Fall Shampoo - 180 ml
R.E Doublemint Mint Candy 30 g - 511
RFL Monia Lock Rectangular Container - 450 ml (917835)
RS Olive Oil Extra Virgin - 90 ml
Many Multi Plug MTS-148
SKB Milk Pan With Handle - 16 cm
China Pizza Toy Set (05/23)
Avonee Diaper M (7-12 kg) - 5 pcs
Dove Conditioner - 330 ml
Notepad 76x102 mm - 53
Haldiram Fried Chicken Masala - 70 g
China Peeler 5384 (11/21)
Roja Seafood Popcorn - 240 g
Faysal Bubble Toy - 54
China Coffee Mug 1124 (4/23)
Vinut Mango Juice - 250 ml
Cocola Lexus Vegetable Biscuits 12 Pack - 240 g
D-Nee Organic Baby Powder - 180 g
Glitter Color Pen - 10 pcs
Fresco Rainbow Container 4 pcs
Clariss Face Wash 100 ml Combo (81 Taka Off)
Clove Powder - 20 g
Fontana Apple Juice - 1 L
Petra Paper Clip 29 mm
Plastic Straws 4 pcs - 59
China Baby Bath Set (05/23)
Ginnou Cho Love Box T5 - 55 g
Hosen Baby Corn - 425 g
Julie's Love Letters Wafer Cube Choco Hazelnut - 150 g
Crocodile Underwear - 08
Emami 7 Oils in One Hair Oil - 300 ml
Shanto Jhuri Ball - 081
United Melamine Tray 18" UM-112
Cool Fresh Toothbrush 90005
Bio Active Goat Milk Soap - 70 g
Black Stone Agarbatti
Lafz Musky Calm Deodorant - 50 ml
One Time Bati 1000 g - 10 pcs
Clariss Sweet Corn - 425 g
Himalaya Moisturizing Aloe Vera Face Wash - 100 ml
Davidoff Espresso 57 - 100 g
Pepsodent Germi Check+ Toothpaste - 200 g
Vitamin E Facial Capsule - 60 pcs
Shohel Fridge Handle Cover - 09
Atlas Umbrella - 033
Kodomo Baby Lotion - 200 ml
Snickers Minis - 333 g
Standard Silicone Nipple 22216/22217/22218
Strawberry Adult - 80 g
Muslim Fashion Half Shirt - 043
KS Baby Frock Set - 001
Ulker Yupo Kola - 80 g
Faysal Hair Set - 061
Gift Bag - 001
London Boy Vest 31/ - 06
Bengal Service Tray Small
SKB Mug - 08 cm
Golden Girl White Glue 60 g - 53
Imagic Beauty Lipgloss - 8 ml
Dettol Fresh Hand Wash - 200 ml
China P. Hanger (4/23)
Rosemary
Trisa We Care Toothbrush 90019
Kishwan Sweet Toast - 350 g
Ama Full Cream Milk Powder - 1 kg
Zareen Tea Packet - 400 g
Glow & Handsome Men Facewash - 100 g
Whisper Maxi Fit - 8 Pads
Flormar Supermatte Lipstick All - 4.2 g
KitKat Chunky - 40 g
Giggles Wet Towels 72 pcs L-31
Parachute Baby Soap - 75 g
He Power Body Spray - 40 ml
Faysal Sharpener - 51
Muslim Til Papri - 70 g
Milk Vita Skimmed Milk Powder - 1 kg
Dabur Honey - 1000 g
Stone Rubber Band SRB240
Haiko Tutti Frutti - 100 g
Listerine Fresh Burst Mouth Wash - 250 ml
China Grill Pan (05/23)
Omuda Nailcutter 300150
Salted Almond Can (Buy 1 Get 1 Free) - 110 g
Pilot Hi-Tecpoint V5 Pen All
Anita Indian Rumal (4/23)
Sanjida Spoon - 57
Vanilla Muffin - 360 g
Shokti Plain Doi 4 pcs Combo - 320 g
Nutella (Save 300 TK) - 750 g
Alex Car Wash - 500 ml
Julie's Oat 25 Chocolate - 200 g
Dosa Tawa - 26 cm
Chocolate Layer Cake - 324 g
Farlin Bottle & Nipple Brush 602
Fair & Lovely Multivitamin - 100 g
Golden Girl Color Sign Pen 12 pcs - 55
Kazi & Kazi Green Mint Tea 25 Bags - 37.5 g
Ceylan Tuna Sunflower Oil - 165 g
Giggles Orchid Wet Wipes 60 pcs L-81
Lux Magical Orchid Body Wash - 500 ml
PC Waterproof Lipliner Pencil - 36M
Pusti HT Energy Biscuit - 180 g
Hot Ice Body Spray - 200 ml
Bear Training Cup With Straw L-22409
RFL 5" Curry Bowl
Memory Narikel Chira - 100 g
EBY Rose Water Mist Toner - 120 ml
Ariel Complete Powder - 1 kg
Lucy Oliva Olive Oil - 150 g
Baby Silicone Spoon AB-212001BC
Asif Short Socks - 5021
Coca-Cola Zero Sugar Can - 330 ml
Pro-Creative Ultra Soft Toothbrush
RFL Luxury Soap Case - 1 pc
Kamrul Pith Chulkani - 51
Pastadora Penne Rigate - 400 g
Shammi Cranberry Juice - 1 L
Ferrero Rocher 16 pcs - 200 g
Alif Tri-Coikur Padlock - 20 mm
ND-SG Ripcurl T-Shirt - 425
A.K Check Trouser Big - 552
Pacific Gold Tuna - 165 g
Sugar Gold - 500 Pellets
Carnation Milk - 405 g
Nivea Men Whitening Roll-On - 50 ml
Jing He Clips - 05
Navratna Oil - 300 ml
Dabur Hajmola Tablet Imli - 100 Tab
Rafian Check Trouser 2X-3X - 082
Fem Creme Bleach Gold - 24 g
RFL 10 Cube Bowl With Lid
Nestle Milo - 400 g
Savoy Mango Ice Cream - 1 L
Baygon Cockroach Control - 570 ml
Ramisa Til Tana Packet - 100 g
Rupa Kidline S Vest 16/18 - 22
Ramisa Jhal Nimki Box - 250 g
China Medicine Box (5/23)
Poa Fish
Silver Pomfret Fish Sauce - 750 ml
Cyprina Mango Nectar - 1 L
RS Castor Oil Glass Jar - 90 ml
Queen Snow Jelly Packet - 100 g
Mug Big - 202B
MP Hijab Band - 031
SK Podakivo Juice - 1 L
Keema Paratha 5 pcs - 500 g
Pantene Shampoo & Conditioner 2-in-1 - 650 ml
Storage Container - 3 L (91001)
China Sheep Toy (06/23)
Shan Paya Masala - 50 g
Nivea Men Fresh Active Body Spray - 150 ml
Wave Anti Tobacco Air Freshener - 300 ml
Royal Garden All Flavors - 220 g
Sanjida Wooden Lemon Chipper - 07
ROK Wood Cleaner Refill - 350 ml
Savoy Vanilla Ice Cream - 1 L
NAJ Spicy Toast Biscuits - 200 g
UM Mini Bati UM-113
China Plane Toy (05/23)
Lamia Fruits Assembling Toy - 063
L'Oreal Paris Repair 5 Conditioner - 175 ml
Nivea Fresh Natural Body Spray - 150 ml
Shipon Payel - 071
SM Sports Bra - 081
Meril Orange Shampoo - 250 ml
Golden Girl Birthday Heart Candle - 001
Shan Tandoori Chicken Masala - 50 g
Kabir Ear Ring - 021
Glucon-D Regular Packet - 500 g
Johnson's Baby Milk + Rice Cream Jar - 100 g
Everest Chhole Masala - 50 g
Almas Miswak - 41
Sticker - 03
Mentos White Sugarfree Gum - 60 g
Haiko Kewra Essence - 25 ml
Cyprina Grape Juice - 1 L
Comfort Touch of Love - 2 L
5" Star Bowl
London Boy Mens Vest - 201
London Boy Vest S / M / L - 08
Cusson Baby Daily Bath - 200 ml
MP Scarf Pin - 001
Money Bag - 052
Qiamlili Makeup Puff 8438
Saffola Honey - 500 g
Cotton Buds BF-113 - 100 pcs
Pran White Vinegar - 650 ml
Petra Correction Pen - 3 g
Patanjali Aloe Vera Gel - 150 ml
Jergens Soap - 125 g
Emami 7 Oils Black Seed - 100 ml
Borkot Rice Spoon - 86
Matador All-Time Ball Pen - 1 pc
Johnson's Baby Cream Tube - 50 g
Ponds Bright Beauty - 23 g
Super Bowl Printed - 3 L
RS Extra Virgin Olive Oil - 250 ml
Magic Book - 052
Eclipse Mints Gum - 35 g
Sufia Hair Comb (Chiruni) - 07
Nivea Men Face Wash - 100 ml
China Kitchen Scissor (05/23)
Jordan Total Clean Toothbrush
Richastar 14 pcs Balloon - 021
Parachute Baby Wash - 100 ml
Herman Peanut Butter All - 340 g
Microwave Cover - 09
Prince Brown Bread - 400 g
Ramisa Badam Tana Packet - 100 g
Dabur Amla Hair Oil - 100 ml
Vitamin E Skin Care Cream - 113 g
Train Cup With Silicone Nipple L-22408
Aseel Vegetable Ghee - 1 L
Air Tight Container Transparent 3 pcs Set
Dettol Fresh Body Wash - 250 ml
SK Red Guava Juice - 1 L
Mentos Gum - 60 g
RFL Vegetable Washing Net 42
Ifad Cheesy Bites 12 Pack - 240 g
Foster Marmalade Jam - 450 g
Super Clay Foam - 031
Money Bag - 002
Neutrogena Clear & Defend Face Wash - 150 ml
Fish Egg
MP Hair Band - 56
Champignons Mushrooms Whole - 400 g
Salted Cashews (Buy 1 Get 1 Free) - 50 g
Mina Container Medium White & Transparent Orange
Nestle Nido Milk Powder Tin - 2500 g
DOMS Mathematical Box - 521
Ma SS Spoon - 51
Aarong Cheese Spread Garlic - 180 g
Almarai 8 Triangles Cheese 120 g - 751
Vatika Enriched Hair Oil - 300 ml
Almas SXH Digital Ring Tasbih - 55
China Mini Dust Bin (05/23)
Sharpener - 53
Hosen Fruit Cocktail Fiesta - 836 g
SCL Rajkonna White Glow Shower Gel - 330 ml
Dettol Cool Liquid Handwash - 200 ml
Fiama Gel + Creme Soap - 125 g
Sunsilk Thick & Long Shampoo - 650 ml
Zess Assorted Biscuits - 451 g
Alif EHBL Pliers 6"
Meiji Hello Panda Milk - 50 g
Dove Mens Body Spray - 150 ml
Crunchos Cashew Can - 100 g
Golden Rose Rich Color Nail Lacquer - 10.5 ml
SKB Small Bowl - 11 cm
UM Love Bati UM-132B
Camel Mixed Snacks - 40 g
Memory Biryani Masala - 50 g
Mr. Potato - 75 g
China Charger Fan 2028 (4/23)
China Paposh 922 (6/23)
Cheese Ball Cracker - 80 g
Rajkonna Active Charcoal Face Wash - 100 ml
Dettol Lime Fresh - 500 ml
Drawing Board - 011
Fresh Soyabean Oil Poly Bag - 1 L
Mix Nut Indo Herb (Buy 1 Get 1 Free) - 150 g Bag
Perrier Mineral Water - 33 cl
YC Beautiful Day Body Shower Gel - 400 ml
AD Plastic File Bag - 012
Linco Orthodontic Pacifier New Born L-22307
China Brushes Set 09 (5/23)
Sony Alkaline AA Battery - 2 pcs
Harpic Flushmatic - 3 pcs
Almas Sultan Attar 6 ml - 021
BD Mixed Pickle - 400 g
Lakme Perfecting Liquid Foundation - 27 ml
Popular Round Container Set
Everest Chicken 65 Kabab Masala - 50 g
China Dream Kitchen Toy 174 (6/23)
Ossum Romance Perfumed Body Mist - 115 ml
Vichu Garden Veggie Soup 4 Servings - 50 g
China P.V.C Dolphin Toy (4/23)
China Lunch Box 1000 ml (05/23)
Golden Garden Mix Plum - 200 g
Hosen Green Olives - 345 g
YC Acne Facial Gel - 250 ml
Flormar Wet & Dry Compact Powder All - 10 g
Nihar 5 Seeds Anti Hair Fall Oil - 200 ml
Pound Cake Chocolate - 280 g
4.5" Spring Bowl 75751
Reyad B.L Multi Plug - 591
Sufia Comb - 08
Gillette Cool Wave Deo Stick - 48 ml
Brylcreem Hair Cream - 125 ml
Hershey's Cookies N Creme - 40 g
BD Boroi Chutney - 500 g
Dimes Mango Can Juice - 330 ml
Foxs Crystal Clear Fruits Mints - 90 g
Ma Steel Bowl - 04
Pusti HT Sweet Nut Biscuit - 200 g
AD Magic Book - 003
Groome Personal Cotton Pads - 80 pcs
Igloo Mango Ice Cream - 1 L
Igloo Ambrosia - 1 L
Shezan Mango Fruit Drink - 250 ml
China SS Wall Hanger (06/20)
Mama Tom Yum Cup Noodles - 62 g
ROK Extra Power Phenyl - 3 L
Dream Food Color Purple - 28 ml
PC Metallic Nails - 11.5 ml
10" Soup Plate 92435
Shokti Vanilla Delight Kids - 80 g
Quaker Oats Jar - 900 g
YC Acner Herbal Soap - 100 g
Sanjida Wooden Haman Dista - 08
PC Aura Botanica Hair Conditioner - 360 ml
Dolphin Nata de Coco Mango Drink - 350 ml
China Medicine Box (05/23)
Lakme Eyeliner - 9 ml
Shan Tikka Seekh Kabab - 50 g
Groome Bird Nest Brightening Face Wash - 100 ml
SCL Groome Tea Tree Nose - 6 Strips
Lipton Green Tea 25 Tea Bags - 32.5 g
Memory Misti Monakka - 300 g
Golden Rose Beauty Balm BB Cream - 30 ml
Sorputi Fish Deshi (HB)
Kwality TikTik Clip MED TTM
China Scissors K-2117 (5/23)
Quality Baby Cap - 06
Glow & Lovely Deep Glow Face Wash - 100 g
Matador Radiant Ball Pen - 1 pc
YC Cucumber Face Wash - 100 ml
China Cooking Toy (05/23)
Beauty Glazed Lipstick - 2.8 g
Imagic Waterproof Eyeliner Pen - 1.5 g
Haribo Chamallows Candy - 70 g
Calvin Klein Underwear XXL - 011
Johnson's Baby Oil - 50 ml
China Coffee Mug (4/23)
Kellogg's Corn Flakes - 275 g
L'Oreal 6 Oil Nourish Conditioner - 360 ml
Borkot Dal Spoon - 86
Glucon-D Nimbu Pani Flavor Jar - 400 g
Clariss Dark Desire Fragrances - 100 ml
Depa Turkey Fashion Orna - 062
Perfect White Toothbrush 90004
Chupa Chups Bon Bon Strawberry Lollipop Bubble Gum
Johnson's Baby Hair Oil - 100 ml
Nancy Crop Basket Big
Fogg Prince Scent - 30 ml
Nylon Toning Brush B&N-22501
Diamond Aluminum Foil - 37.5 ft
Swiss Garden Young Baby Corn - 425 g
BTC Birthday Cap Dami - 06
Lamia Dough Bubble - 57
Foster Clarks Essence - 28 ml (S)
Quality Cap - 55
Depa Party Hijab - 052
Borkot Cup - 57
Lafz Blue Surfer Deodorant - 50 ml
ISPI Mango Instant Drink Jar - 750 g
W7 Power Puff Latex Free
Ramisa Nimki Box - 250 g
8" Crazy Plate
Orna House Hijab Orna - 082
Lux Soap - 175 g
Ligion Active Gold Mehendi - 30 g
Ayesha Soap Case - 001
10" Soup Plate 919111
Cola Flavor Candy - 100 g
Lacys Zip Bag Gallon - 20 Bags
Foska Glitter Gel Pen 12 pcs - 031
China Garlic Cutter (4/22)
Magur Fish Processed - kg
Award Shoe Polish Tin Black/Brown - 40 g
China Sauce Pan 24 cm (5/23)
ORI Good Times Biscuits Tin - 540 g
Alam Baby Frock Set - 082
Anchor Icing Sugar - 250 g
MP Safety Pin Pata - 001
Kishwan Kheer Milk Cookies - 270 g
Vaseline Men Facewash - 100 g
PediaSure Chocolate Milk - 400 g
Dabur Chyawanprash - 500 g
Mr. Potato Sour/Onion - 40 g
Golden Girl Flash Powder - 561
Golden Girl Color Tray - 02
Kwality Eyebrow Razor 3116
Orion Custard Cake - 138 g
Skin Cafe Face Wash - 140 ml
Foster Clarks Lemon Jelly - 85 g
Malka Peshawari Karahi Gosht - 50 g
Zess O Cream Crackers - 184 g
IKO Sugar Free 9 Grain Cracker - 178 g
Nestle Coffee Mate - 1 kg
Ramisa Fish Papor - 80 g
Nababi Premium Bakorkhani - 400 g
Alam Baby T-Shirt Set - 085
Mama Spicy Seafood Cup Noodles - 65 g
Cute Perfumed Coconut Oil - 330 ml
Salted Cashews Can (Buy 1 Get 1 Free) - 110 g
FE Monty Balloon - 061
Ramisa Sweet Monakka Box - 350 g
Savoy Butterscotch Ice Cream - 1 L
Savoy Chocolate Ice Cream - 1 L
RFL Una Jar - 2 L
Orion Choco Pie - 180 g
Golden Girl Foska Water Color Pen 12 pcs - 07
SCL Brightening Mask - 70 g
Kellogg's Special Original - 290 g
Calvin Klein Underwire - 511
China Flower Vase (5/23)
Golden Girl Radiant Sticker - 53
Jhatpat Chicken Popcorn - 150 g
Clariss Bamboo Charcoal Face Wash - 100 ml
Haldiram Chilly Mayonnaise - 500 ml
Dan Cake Tiramisu Plain Cake - 360 g
Garnier Color Naturals Brown 4 - 100 ml
Drinking Cup Silicone 300 ml MH-012
Star Jus Mango - 1 L
NAJ Milk Biscuits - 300 g
Multi Purpose Container 5 L - 91011
Bottle Brush
Sashi Denim Agarbatti
ACI Pure Beef Curry Masala - 20 g
Faysal Spinner Ball 061 (4/23)
Golden Harvest Chicken Hot Wings - 500 g
Shokti Vanilla Delight Adult 4 pcs Combo - 320 g
Rupa Mens Vest - 511
PediaSure Vanilla Flavor Milk - 400 g
China Vacuum Cup (4/23)
China Vacuum Cup KL140 (4/23)
Alif TPP Padlock - 25 mm
President Vest L - 201
Premio Kitchen Container - 7 L
Savlon Iris Herbal Hand Wash - 500 ml
Uno Card Small - 54
Lafz Refreshing Face Wash - 75 ml
Swan Kids Vest - 14
Godrej Expert Rich Hair Color Burgundy
Petra Crown Oil Pastel - 18
Fried Chicken Wings - 500 g
Olive Miswak - 45
Dettol Disinfectant Spray Trigger - 350 ml
Marks Diabetic Milk - 400 g
Johnson's Baby Oil India - 100 ml
Mothercare Baby Lotion - 300 ml
Hershey's Strawberry Syrup - 623 g
Lacys Baking and Cooking Paper - 30cm x 5m
Almas Prayer Tupi - 5057
Camay Natural Soap - 125 g
Opal Bowl Set 5 pcs B11
Italiano Hot Chilli Sauce - 340 g
Shammi White Grape Juice - 240 ml
Pastiano Penne Pasta Jar - 500 g
Almas Silver Attar 6 ml - 0052
Hot Mug
Anita Indian Kallani Bra (03/23)
Dove Beauty Bar Soap All - 135 g
CP Japanese Bite - 250 g
AS Silicone EN Milk Bottle
Funtastic Glucoz Biscuit - 225 g
Quality Cap - 004
Organikare Miracle Face Wash - 100 g
Cyprina Apple Juice - 1 L
Rubik's Cube - 003
Golden Rose Moisturizing Cream Foundation - 35 ml
Pastavilla Shell Pasta - 500 g
Golden Girl Ribbon - 5
Pantene Anti Dandruff - 180 ml
Golden Rose Automatic Eyeliner Pen
Gift Gallery Mug - 081
NAJ Garlic Toast Biscuits - 200 g
One Card Plastic - 082
Pantene Silky Smooth Care 2-in-1 - 650 ml
Pran Jhatpat Potato Wedges - 500 g
AS Baby Spoon Set
Dove Hair Fall Rescue - 450 ml
Faysal Musical Luminous Candle - 051
Zipbag 25 Bags 200mm x 175mm
Pastavilla Filini Pasta - 500 g
Beauty Powder Case 92681
Crown Mayonnaise
Tropica Black Cumin Honey - 250 g
Golden Rose My Matte Lip Ink Lipstick - 5 ml
Rasna Fruitplus Mango Instant Drink Mix - 750 g
Double Milk Baby Powder - 180 g
L'Oreal CC 316 Burgundy Hair Color - 21 ml + 24 ml
Lays Barbecue Flavor - 184 g
Dove Beauty Cream - 75 ml
Lacys Aluminium Foil 45cm x 7.6m - 37.5 SqFt
Clariss Lemon Face Wash - 100 ml
Alam Baby Dress - 004
Dove Shower Gel - 250 ml
Meringue Marshmallow S Cookie - 65 g
City Gold Indian Chain (4/23)
Munchy's Lexus Cheese Cream - 190 g
Armaf Club De Nuit Women Body Spray - 200 ml
Golden Rose Silky Touch Compact Powder - 12 g
Maya Hair & Scalp Oil - 100 ml
Panam Deep Clean Beetroot Facial Wash (Buy 1 Get 1 Free)
Ponds Age Miracle Face Cream - 100 g
Fogg Napoleon Body Spray - 120 ml
Pran BBQ Sheek Kabab Masala - 50 g
Mini Water Glass - 153
Swiss Garden Kidney Beans - 425 g
Golden Girl Color Sign Pen 24 pcs - 032
China Ginger Chipper (5/23)
Borges Green Sliced Olives - 230 ml
Finlay Gold Tea 50 Bags - 100 g
China Mob Cover (5/23)
Dabur Gold Beliful Hair Oil - 200 ml
Music S.C Feeding Bottle 01101 - 240 ml
Nivea Men Deep Cleaning Face Wash - 100 g
RFL Bela Ruti Box Medium
SKB Tongs 02
Colgate Peppermint Mouth Wash - 500 g
Gatsby Water Gloss Hair Cream - 150 g
Safety Plus Almond & Milk Hand Wash
Nut Walker Cashew Nut - 270 g
11" Snowdrop Crazy Plate
Golden Girl Color Sign Pens - 12 pcs
Golden Girl Gift Box Small - 03
Layer Absolute Craze Body Spray - 135 ml
Pineapple Toy - 001
Vinut Orange Juice - 250 ml
China Egg Beater 401 (6/23)
Discovery Peanut Butter Chunky - 340 g
Fry Pan Without Lid - 16 cm
Sugar Free Green - 100 Pellets
China Tiffin Box 8520 (5/23)
Gillette Mach 3 Sharp Blade Razor
Orkide Olive Oil Pet - 250 ml
Vanish Liquid - 400 ml
China Basketball (5/23)
China Whistle Toy 10 pcs (5/23)
Crunchos Cashew Packet - 100 g
Fazlu Blue Heaven Kajal - 2 g
Johnson's Premium Gift Set - 6 pcs
SKB Mug - 10 cm
Cassava Chips Original - 150 g
Good Luck Design My Bag FC
Petra Highlighter Yellow
Bliss Air Freshener BAF
Groome Pretty Pink Glow Lip Balm - 30 g
Meril Aloe Petroleum Jelly - 50 ml
Shinex Floral Floor Cleaner - 1000 ml
Astral Cream - 50 ml
MacCoffee Original Jar - 50 g
Nuttos Organic Mixed Nuts - 400 g
Sketch Paper - 03
Cyprina Exotic Nectar 8 Fruits With Vit Cae - 1 L
Khaleque Donut Cutter - 05
Faysal Wooden Puzzle - 591
Gambit Body Spray - 200 ml
Spice Black Salt Powder - 100 g
Pran Jhatpot Chicken Meatball - 250 g
Shammi Tamarind Drink - 250 ml
China P.V.C Ball Toy (05/23)
L'Oreal Elseve Shampoo - 280 ml
Binimoy Cartoon File - 022
Anita Indian Guddi Bra (4/23)
Livon Serum - 18 ml
Flormar Silk Matte Liquid Lipstick All - 4.5 ml
10" Snowdrop Coup Plate
16" Rectangular Handle Tray 92331
Rupa Kidline Vest - 20/22
Almas Prayer Tupi - 5011
Golden Girl Joytiti Pencil 2H - 41
Baby Food Bib - 006
Sourav Baby Dress - 081
Matador Pin-Point Ball Pen - 1 pc
Nivea Silver Protect Roll-On - 50 ml
Dust Pen Medium Red
Bright Star LED Light BS-7622
Engage Woman Body Spray - 150 ml
Tayas Almond S. Chocolate Tin - 700 g
O Knock Choco Wafer C Biscuits - 25 g
Fresh Actifit Mustard Oil - 500 ml
Golden Girl Glitter Powder - 021
Bakemans Glaze - 200 g
Golden Rose Liquid Concealer All - 2 ml
Joytiti Drawing Book - 001
China Lemon Squeezer (5/23)
Super White Powder - 500 g
Gatorade Sports Drink - 500 ml
Lacys Baking & Cooking Paper - 30cm x 10m
China Long Jharu (4/23)
Lamia Sharpener - 51
Head & Shoulders Cool Menthol 2-in-1 Shampoo & Conditioner - 650 ml
MacCoffee Gold - 50 g
Benson & Hedges All Flavors - 1 Packet
City Gold Indian Basket (4/23)
China Duck Toy (4/23)
Pantene Silky Smooth Shampoo - 650 ml
Premio TopUp Washing Net All
Schweppes Tonic Water - 320 ml
Calvin Klein Underwear - 001
Karim Baby Hand Watch - 031
Shan Bihari Kabab - 50 g
T3 Baby Glass Feeding Bottle NB-205G
YC Gold Caviar Face Wash - 100 ml
Faysal Birthday Candle - 09
Flormar Lipliner Pencil All - 1.14 g
Parimal Gurudev Agarbatti
Rui Fish XXL
YC Papaya Double White Plus Soap - 100 g
Kwality Stone Nose Pin BNP
Khaleque Cake Board - 501
Colour Me Pop Art Body Spray - 150 ml
Anita Indian Rupa Frontline Vest (4/23)
Gillette Anti Clear Gel - 75 ml
Chaad Orange Dish Washing Liquid - 250 ml
Parachute Just for Baby Oil - 100 ml
Pran Coriander Powder - 200 g
Table Clock - 055
Golden Rose Soft Color Mono Eyeshadow - 2.3 g
Golden Rose Eyebrow Pencil
Marlen Lemon Air Freshener - 300 ml
Farlin Natural Feeding Bottle 9oz - 250 ml
RFL Store Container 87808
Aloe Vera All - 1.5 ml
Memory Jhal Monakka Box - 300 g
Ponds Skin Pure Bright Beauty Face Wash - 100 ml
Tetley Lemon & Honey Green Tea - 25 Bags
Pantene Advanced Hairfall Shampoo - 650 ml
Jet Powder - 1 kg
Parachute Coconut Hair Oil Tin - 200 ml
Golden Rose Lash Plumping Mascara - 13 ml
SME Wooden Bat Hammer - 021
Leggos Tomato & Herbs Sauce - 500 g
Music Training Cup 4 IN 1 6oz - 180 ml
Sundrop Creamy Peanut Spread - 200 g
Shanto Jumping Rope - 05
China Ceramic Mug 0861 (4/23)
China Cup & Saucer Set 6 pcs (5/23)
Halda Valley DW Green Tea - 45 g
L.A Girl Pro Conceal HD - 3 pcs
Faysal Wrapping Paper - 71
China Scissors F-312-241 (6/23)
Digimon Card
Memory Bundia - 200 g
Confume Argan Hair Shampoo - 750 ml
Vegetable Mini Samosa Premium - 250 g
Cadbury Dairy Milk Roast & Almond - 143 g
Fatema Wooden Jewellery Box - 003
Vaseline Aloe Soothe Lotion - 400 ml
EBY Mineral Facewash - 100 ml
Lafz Onion Seed Oil For Hair - 100 ml
Bombay Sweets Peanut White Roast Jar - 250 g
How How Rice Stick - 500 g
Termite Killer Spray - 250 ml
Shanto Cricket Ball - 002
Kishwan Continental Cookies Tin - 900 g
Golden Girl Note Book - 04
PC Roll Act Lashes Mascara - 7 ml
Lattafa Maknoon Air Freshener - 300 ml
Trisa Feel Good Toothbrush 90026
Golden Girl Emoji Eraser 4 pcs - 01
Lacys Chefwrap 30cm x 90m - 90 m
China Water Pot (5/23)
Memory Muruli Box - 300 g
Dove Intense Repair - 450 ml
Halda Valley Black Tea Packet - 200 g
Bosny Rat Glue - 400 ml
Beef Mini Samosa Premium - 250 g
Jhatpat Vegetable Samosa - 10 pcs
AS Baby Milk Bottle - 280 ml
Revlon Shampoo - 20 FL OZ
SKB 3 Layer Ruti Tawa - 26 cm
Gillette Fusion Razor R+1
China Jar (5/23)
Igloo Strawberry Ice Cream - 1 L
AS ARC Automatic Feeding Bottle
China Khunti 77 (5/23)
PC Aura Botanica Shampoo - 360 ml
Sweetened Bakorkhani - 10 pcs
Avena OAT Milk Snacks Bar - 120 g
MSH Ladies Panty - 588
Turmeric Powder - 500 g
SKB BD Pan With SS Lid - 18 cm
Dimes Multi Fruit Can Juice - 330 ml
Cream Crackers Biscuit - 300 g
SCL Rajkonna Hair Repair Powder - 70 g
Angelic Amazonia Air Freshener - 300 ml
Alone MS Cleansing Tissue
Premio Colors Container - 7 pcs
Quality Cap - 553
Kurmure Classic Chanachur - 300 g
Remia Sauce All - 250 ml
Shezan Mango Bar 40 pcs - 360 g
Pastiano Premium Pasta (0% Chol) - 500 g
Barilla Angel Hair N.1 Pasta - 500 g
Fazlu Pocket Comb - 09
Lattafa R-al Musk Air Freshener - 300 ml
Shakti Mango Doi 4 pcs Combo - 320 g
Shinex Floor Fresh Mist Cleaner - 1000 ml
Supermoon RC Light SM9714
Cute Jasmine Bath Soap - 125 g
Miyako Hand Blender HB-7701
PC Shower Gel - 400 ml
Technic Pressed Pigment - 30 g
L'Oreal Total Repair 5 Shampoo - 192.5 ml
AS Baby Brush Comb Set
Aarong Milk Powder - 400 g
Piramid Rubisquib - 532
Lion Zact Toothpaste - 190 g
Faysal Singing Candle - 05
Gulsha Fish Big (HB)
Shokti Mango Yogurt 4 pcs Combo - 320 g
Golden Girl Sharpener 4 pcs - 03
Everest Garam Masala - 50 g
Orkide Olive Oil Glass Bottle - 250 ml
Feeding Bottle Wash BF-200 - 700 ml
Parachute Milky Glow Baby Lotion - 200 ml
Clariss Castor Oil - 100 g
Yardley London L Soap - 100 g
Almas Prayer Tupi - 06
Sunflower Oil 3 L Combo
SCL Lilac Face Wash - 120 ml
London Boy Vest M - 18
Nestle Classic 18 g (Buy 3 Get 10 TK Off)
Tabu Perfumed Powder - 100 ml
Lamia Steel Pumper 022 (4/23)
HIT Dango & Malaria Aerosol - 200 ml
Savlon Mild Anti Soap - 100 g
Jet Liquid Detergent Poly - 600 ml
Parachute Just For Baby Combo Gift Pack
Yoyo Yellow Toy - 05
Camel Green Peas - 130 g
Dr. Ruchi Mango Pickle - 1 kg
Diana SB Cream - 170 ml
Tiffany Break Gift Box
RFL Merry Jug - 2.8 L
Revlon Colorsilk Dark Brown Color 3N
5" Winter Bowl
Cute Neem Bath Soap - 125 g
Sepnil Anti-Bacterial Soap - 100 g
Minister Hexin Hand Sanitizer Spray
Ramisa Jhuri Bhaja - 350 g
ISPI Instant Mango Drink Pack - 125 g
Premio Diamond Bowl - 5 L
YC Gold Caviar Shower Gel - 210 g
Miss & Mrs Volume Mascara - 9 ml
Foodex Chilli Sauce - 435 ml
Lakme White Radiance
"""

# আপনার সম্পূর্ণ Customer লিস্টটি এখানে পেস্ট করবেন
raw_names = """
Tariq Rahman
Tariq Islam
Tariq Uddin
Tariq Ahmed
Tariq Ali
Usman Tariq
Shakib Al Hasan
Mashrafee Mortaza
Tamim Iqbal
Imrul Kayes
Mohammad Ashraful
Nigar Sultana
Nasir Hossain
Rubel Hossain
Mehedi Hasan Miraz



Shafiq Rahman

Shafiq Islam

Shafiq Uddin

Shafiq Ahmed

Shafiq Ali

Rafiq Rahman

Rafiq Islam

Rafiq Uddin

Rafiq Ahmed

Rafiq Ali

Arif Rahman

Arif Islam

Arif Uddin

Arif Ahmed

Arif Ali

Asif Rahman

Asif Islam

Asif Uddin

Asif Ahmed

Asif Ali

Rakib Rahman

Rakib Islam

Rakib Uddin

Rakib Ahmed

Rakib Ali

Sakib Rahman

Sakib Islam

Sakib Uddin

Sakib Ahmed

Sakib Ali

Tamim Rahman

Tamim Islam

Tamim Uddin

Tamim Ahmed

Tamim Ali

Naim Rahman

Naim Islam

Naim Uddin

Naim Ahmed

Naim Ali

Fahim Rahman

Fahim Islam

Fahim Uddin

Fahim Ahmed

Fahim Ali

Farhan Rahman

Farhan Islam

Farhan Uddin

Farhan Ahmed

Farhan Ali

Rayhan Rahman

Rayhan Islam

Rayhan Uddin

Rayhan Ahmed

Rayhan Ali

Imtiaz Rahman

Imtiaz Islam

Imtiaz Uddin

Imtiaz Ahmed

Imtiaz Ali

Imran Rahman

Imran Islam

Imran Uddin

Imran Ahmed

Imran Ali

Mehedi Rahman

Mehedi Islam

Mehedi Uddin

Mehedi Ahmed

Mehedi Ali

Sohel Rahman

Sohel Islam

Sohel Uddin

Sohel Ahmed

Sohel Ali

Rubel Rahman

Rubel Islam

Rubel Uddin

Rubel Ahmed

Rubel Ali

Jewel Rahman

Jewel Islam

Jewel Uddin

Jewel Ahmed

Jewel Ali

Masud Rahman

Masud Islam

Masud Uddin

Masud Ahmed

Masud Ali

Mamun Rahman

Mamun Islam

Mamun Uddin

Mamun Ahmed

Mamun Ali

Hasan Rahman

Hasan Islam

Hasan Uddin

Hasan Ahmed

Hasan Ali

Kamrul Rahman

Kamrul Islam

Kamrul Uddin

Kamrul Ahmed

Kamrul Ali

Nazmul Rahman

Nazmul Islam

Nazmul Uddin

Nazmul Ahmed

Nazmul Ali

Aminul Rahman

Aminul Islam

Aminul Uddin

Aminul Ahmed

Aminul Ali

Ashraful Rahman

Ashraful Islam

Ashraful Uddin

Ashraful Ahmed

Ashraful Ali

Saiful Rahman

Saiful Islam

Saiful Uddin

Saiful Ahmed

Saiful Ali

Anwar Rahman

Anwar Islam

Anwar Uddin

Anwar Ahmed

Anwar Ali

Jamal Rahman

Jamal Islam

Jamal Uddin

Jamal Ahmed

Jamal Ali

Kamal Rahman

Kamal Islam

Kamal Uddin

Kamal Ahmed

Kamal Ali

Rahim Rahman

Rahim Islam

Rahim Uddin

Rahim Ahmed

Rahim Ali

Karim Rahman

Karim Islam

Karim Uddin

Karim Ahmed

Karim Ali

Selim Rahman

Selim Islam

Selim Uddin

Selim Ahmed

Selim Ali

Jashim Rahman

Jashim Islam

Jashim Uddin

Jashim Ahmed

Jashim Ali

Farid Rahman

Farid Islam

Farid Uddin

Farid Ahmed

Farid Ali

Hamid Rahman

Hamid Islam

Hamid Uddin

Hamid Ahmed

Hamid Ali

Mahbub Rahman

Mahbub Islam

Mahbub Uddin

Mahbub Ahmed

Mahbub Ali

Shahin Rahman

Shahin Islam

Shahin Uddin

Shahin Ahmed

Shahin Ali

Alamgir Rahman

Alamgir Islam

Alamgir Uddin

Alamgir Ahmed

Alamgir Ali

Jahangir Rahman

Jahangir Islam

Jahangir Uddin

Jahangir Ahmed

Jahangir Ali

Tofazzal Rahman

Tofazzal Islam

Tofazzal Uddin

Tofazzal Ahmed

Tofazzal Ali

Mofizur Rahman

Mofizur Islam

Mofizur Uddin

Mofizur Ahmed

Mofizur Ali

Habibur Rahman

Habibur Islam

Habibur Uddin

Habibur Ahmed

Habibur Ali

Anisur Rahman

Anisur Islam

Anisur Uddin

Anisur Ahmed

Anisur Ali

Mizanur Rahman

Mizanur Islam

Mizanur Uddin

Mizanur Ahmed

Mizanur Ali

Lutfur Rahman

Lutfur Islam

Lutfur Uddin

Lutfur Ahmed

Lutfur Ali

Atiqur Rahman

Atiqur Islam

Atiqur Uddin

Atiqur Ahmed

Atiqur Ali

Ashiqur Rahman

Ashiqur Islam

Ashiqur Uddin

Ashiqur Ahmed

Ashiqur Ali

Mostafa Rahman

Mostafa Islam

Mostafa Uddin

Mostafa Ahmed

Mostafa Ali

Monir Rahman

Monir Islam

Monir Uddin

Monir Ahmed

Monir Ali

Tushar Rahman

Tushar Islam

Tushar Uddin

Tushar Ahmed

Tushar Ali

Miraj Uddin Chowdhury
Tasnimul Intazam Asif
Bishwaprotap Ray
Tanvir Ahammed Rasel
Emdadul Haque


Nusrat Akter

Nusrat Begum

Nusrat Khatun

Nusrat Jahan

Nusrat Chowdhury

Sadia Akter

Sadia Begum

Sadia Khatun

Sadia Jahan

Sadia Chowdhury

Farhana Akter

Farhana Begum

Farhana Khatun

Farhana Jahan

Farhana Chowdhury

Ayesha Akter

Ayesha Begum

Ayesha Khatun

Ayesha Jahan

Ayesha Chowdhury

Fatema Akter

Fatema Begum

Fatema Khatun

Fatema Jahan

Fatema Chowdhury

Sumaiya Akter

Sumaiya Begum

Sumaiya Khatun

Sumaiya Jahan

Sumaiya Chowdhury

Juthi Akter

Juthi Begum

Juthi Khatun

Juthi Jahan

Juthi Chowdhury

Rumi Akter

Rumi Begum

Rumi Khatun

Rumi Jahan

Rumi Chowdhury

Nipa Akter

Nipa Begum

Nipa Khatun

Nipa Jahan

Nipa Chowdhury

Lipi Akter

Lipi Begum

Lipi Khatun

Lipi Jahan

Lipi Chowdhury

Shimu Akter

Shimu Begum

Shimu Khatun

Shimu Jahan

Shimu Chowdhury

Rima Akter

Rima Begum

Rima Khatun

Rima Jahan

Rima Chowdhury

Lima Akter

Lima Begum

Lima Khatun

Lima Jahan

Lima Chowdhury

Bristi Akter

Bristi Begum

Bristi Khatun

Bristi Jahan

Bristi Chowdhury

Bonna Akter

Bonna Begum

Bonna Khatun

Bonna Jahan

Bonna Chowdhury

Nodi Akter

Nodi Begum

Nodi Khatun

Nodi Jahan

Nodi Chowdhury

Tara Akter

Tara Begum

Tara Khatun

Tara Jahan

Tara Chowdhury

Purnima Akter

Purnima Begum

Purnima Khatun

Purnima Jahan

Purnima Chowdhury

Chadni Akter

Chadni Begum

Chadni Khatun

Chadni Jahan

Chadni Chowdhury

Ritu Akter

Ritu Begum

Ritu Khatun

Ritu Jahan

Ritu Chowdhury

Nitu Akter

Nitu Begum

Nitu Khatun

Nitu Jahan

Nitu Chowdhury

Tuli Akter

Tuli Begum

Tuli Khatun

Tuli Jahan

Tuli Chowdhury

Tumpa Akter

Tumpa Begum

Tumpa Khatun

Tumpa Jahan

Tumpa Chowdhury

Sampa Akter

Sampa Begum

Sampa Khatun

Sampa Jahan

Sampa Chowdhury

Jhumpa Akter

Jhumpa Begum

Jhumpa Khatun

Jhumpa Jahan

Jhumpa Chowdhury

Jhumur Akter

Jhumur Begum

Jhumur Khatun

Jhumur Jahan

Jhumur Chowdhury

Nupur Akter

Nupur Begum

Nupur Khatun

Nupur Jahan

Nupur Chowdhury

Nargis Akter

Nargis Begum

Nargis Khatun

Nargis Jahan

Nargis Chowdhury

Salma Akter

Salma Begum

Salma Khatun

Salma Jahan

Salma Chowdhury

Nazma Akter

Nazma Begum

Nazma Khatun

Nazma Jahan

Nazma Chowdhury

Asma Akter

Asma Begum

Asma Khatun

Asma Jahan

Asma Chowdhury

Reshma Akter

Reshma Begum

Reshma Khatun

Reshma Jahan

Reshma Chowdhury

Shirin Akter

Shirin Begum

Shirin Khatun

Shirin Jahan

Shirin Chowdhury

Nasrin Akter

Nasrin Begum

Nasrin Khatun

Nasrin Jahan

Nasrin Chowdhury

Parvin Akter

Parvin Begum

Parvin Khatun

Parvin Jahan

Parvin Chowdhury

Sharmin Akter

Sharmin Begum

Sharmin Khatun

Sharmin Jahan

Sharmin Chowdhury

Yasmin Akter

Yasmin Begum

Yasmin Khatun

Yasmin Jahan

Yasmin Chowdhury

Tahmina Akter

Tahmina Begum

Tahmina Khatun

Tahmina Jahan

Tahmina Chowdhury

Fahmida Akter

Fahmida Begum

Fahmida Khatun

Fahmida Jahan

Fahmida Chowdhury

Mahmuda Akter

Mahmuda Begum

Mahmuda Khatun

Mahmuda Jahan

Mahmuda Chowdhury

Shahida Akter

Shahida Begum

Shahida Khatun

Shahida Jahan

Shahida Chowdhury

Rashida Akter

Rashida Begum

Rashida Khatun

Rashida Jahan

Rashida Chowdhury

Halima Akter

Halima Begum

Halima Khatun

Halima Jahan

Halima Chowdhury

Amina Akter

Amina Begum

Amina Khatun

Amina Jahan

Amina Chowdhury

Khadija Akter

Khadija Begum

Khadija Khatun

Khadija Jahan

Khadija Chowdhury

Rahima Akter

Rahima Begum

Rahima Khatun

Rahima Jahan

Rahima Chowdhury

Karima Akter

Karima Begum

Karima Khatun

Karima Jahan

Karima Chowdhury

Saleha Akter

Saleha Begum

Saleha Khatun

Saleha Jahan

Saleha Chowdhury

Maleka Akter

Maleka Begum

Maleka Khatun

Maleka Jahan

Maleka Chowdhury

Zobeda Akter

Zobeda Begum

Zobeda Khatun

Zobeda Jahan

Zobeda Chowdhury
"""

# ==========================================
# STEP 2: CONTEXT-AWARE CATEGORIZATION LOGIC
# ==========================================
def get_standard_category(item_name):
    name = str(item_name).lower()
    
    # 1. Most Specific Items (Frozen, Spices, Noodles etc.)
    if any(w in name for w in ['nugget', 'patty', 'burger', 'sausage', 'samosa', 'singara', 'spring roll', 'french fr', 'paratha', 'puri', 'kabab', 'kebab', 'frankfurter', 'bologna', 'popcorn', 'tempura', 'kievs', 'wonton']):
        return 'Grocery & Food', 'Frozen & Ready to Cook'
    elif any(w in name for w in ['Bombay Sweets Peanut White Roast','Biryani Scent','Black Pepper Powder','masala', 'spice', 'powder', 'chilli', 'turmeric', 'coriander', 'cumin', 'garlic', 'ginger', 'onion', 'cardamom', 'clove', 'cinnamon', 'nutmeg', 'oregano', 'methi', 'pach phoron', 'saffron', 'jafran','Tomato Sauce','Agar agar','Nutmeg','biriyani masala']):
        return 'Grocery', 'Spices'
    elif any(w in name for w in ['noodles', 'pasta', 'macaroni', 'spaghetti', 'soup', 'ramen', 'chips']):
        return 'Snacks', 'Instant Food'
        
    # 2. Non-Food Items (Personal Care, Cleaning, Toys, Clothes)
    elif any(w in name for w in ['shampoo', 'lotion', 'face wash', 'cream', 'toothpaste', 'hair oil', 'gel', 'conditioner', 'serum', 'scrub', 'bleach', 'hair color', 'soap', 'hand wash', 'handwash', 'shave', 'razor','Mehendi']):
        return 'Personal Care', 'Skin, Hair & Body Care'
    elif any(w in name for w in ['lipstick', 'eyeliner', 'mascara', 'perfume', 'body spray', 'attar', 'nail polish', 'makeup', 'kajal', 'concealer', 'foundation', 'powder', 'scent','castor','perfumed powder']):
        return 'Personal Care', 'Cosmetics & Fragrance'
    elif any(w in name for w in ['Rin Advanced','cleaner', 'wash', 'detergent', 'tissue', 'napkin', 'phenyl', 'aerosol', 'mosquito', 'dish wash', 'dishwash', 'harpic', 'vim', 'scrubber', 'brush', 'mop', 'coil', 'liquid blue', 'insect', 'cockroach','washing powder','Clean & Fresh Powder']):
        return 'Household', 'Cleaning Supplies'
    elif any(w in name for w in ['khata', 'pencil', 'pen ', 'eraser', 'sharpener', 'marker', 'glue', 'tape', 'stapler', 'color']):
        return 'Stationery & Toys', 'Stationery'
    elif any(w in name for w in ['toy', 'doll', 'balloon', 'game', 'play', 'car ', 'puzzle', 'card', 'board']):
        return 'Stationery & Toys', 'Toys'
    elif any(w in name for w in ['Hair Clip','vest', 'bra', 'panty', 'trouser', 't-shirt', 'lungi', 'orna', 'hijab', 'sock', 'shoe', 'hoodie', 'apron', 'umbrella', 'raincoat', 'bangles', 'ring', 'clip']):
        return 'Fashion & Lifestyle', 'Clothing & Accessories'
    elif any(w in name for w in ['baby', 'diaper', 'pampers', 'huggies', 'nipple', 'pacifier', 'cerelac', 'bib']):
        return 'Baby Care', 'Baby Essentials'
    elif any(w in name for w in ['Scrubber','mug', 'bowl', 'plate', 'glass', 'container', 'pan', 'tawa', 'spoon', 'knife', 'khunti', 'bottle', 'flask', 'cooker', 'jar', 'tray', 'pot', 'box', 'basket', 'rehal', 'mat']):
        return 'Home & Lifestyle', 'Kitchen & Dining'

    # 3. Regular Food, Staples, and Raw Items
    elif any(w in name for w in ['Bun','Kurkure Chilli Chatka','Kurkure Spicy Chicken','Kurkure Masala Munch','Kurkure Naughty Tomato','Mini Burger Bun','Mr. Cookie','Snickers','Oreo','biscuit', 'toast', 'chanachur', 'chips', 'chocolate', 'candy', 'wafer', 'gum', 'cake', 'muri', 'chira', 'chewy','Kolson Vermicelli']):
        return 'Snacks', 'Packaged Snacks'
    elif any(w in name for w in ['juice', 'drink', 'cola', 'sprite', 'pepsi', 'fanta', 'water', 'tea', 'coffee', 'nescafe', 'syrup', 'maltova', 'horlicks', 'complan', 'tang', 'rooh afza','Coca-Cola','paan masala']):
        return 'Beverages', 'Drinks & Mixes'
    elif any(w in name for w in ['milk', 'cheese', 'butter', 'ghee', 'curd', 'yogurt', 'doi', 'egg', 'paneer']):
        return 'Dairy & Egg', 'Dairy Products'
    elif any(w in name for w in ['Sagu','rice', 'dal', 'pulse', 'sugar', 'salt', 'oil', 'atta', 'maida', 'suji', 'flour', 'boot', 'chola', 'besan']):
        return 'Grocery', 'Staples'
    elif any(w in name for w in ['Chicken Breast','beef', 'mutton', 'chicken', 'cock', 'fish', 'shrimp', 'crab', 'shutki', 'pigeon', 'liver', 'meat', 'stomach', 'bone']):
        return 'Meat & Fish', 'Fresh Meat & Fish'
    elif any(w in name for w in ['Lati','tomato', 'potato', 'pumpkin', 'gourd', 'brinjal', 'papaya', 'cabbage', 'cauliflower', 'carrot', 'apple', 'grape', 'mango', 'orange', 'malta', 'guava', 'date', 'plum', 'spinach', 'bean', 'capsicum', 'cucumber', 'lemon', 'fruit', 'chilli']):
        return 'Vegetables & Fruits', 'Fresh Produce'
    else:
        return 'General', 'Others'

def get_occasion_tag(item_name):
    name = str(item_name).lower()
    
    if any(w in name for w in ['nugget', 'patty', 'burger', 'sausage', 'meatball', 'french fr', 'samosa', 'singara', 'spring roll']):
        return 'Fast Food / Quick Snacks'
    elif any(w in name for w in ['bread', 'toast', 'jam', 'jelly', 'egg', 'milk', 'tea', 'coffee', 'oats', 'bun', 'paratha', 'puri', 'cereal', 'corn flakes', 'honey','nestle Nestle Classic']):
        return 'Morning Breakfast'
    elif any(w in name for w in ['chinigura', 'biriyani', 'biryani', 'kewra', 'beef', 'mutton', 'roast', 'ghee', 'alu bukhara', 'kismis', 'polao']):
        return 'Heavy Meal / Biryani Cooking'
    elif any(w in name for w in ['fish', 'shrimp', 'mustard oil', 'turmeric', 'coriander']):
        return 'Regular Meal / Fish Prep'
    elif any(w in name for w in ['cleaner', 'detergent', 'tissue', 'harpic', 'vim', 'aerosol', 'coil', 'dishwash', 'phenyl']):
        return 'Household Chores'
    elif any(w in name for w in ['shampoo', 'lotion', 'face wash', 'perfume', 'lipstick', 'toothpaste', 'soap']):
        return 'Personal Grooming & Care'
    elif any(w in name for w in ['baby', 'diaper', 'cerelac', 'pacifier','Feeding Bottle']):
        return 'Baby Care & Feeding'
    elif any(w in name for w in ['khata', 'pencil', 'pen ', 'eraser', 'toy', 'doll', 'balloon']):
        return 'Kids & School Needs'
    elif any(w in name for w in ['mug', 'bowl', 'plate', 'container', 'pan', 'tawa', 'cooker','Khunti']):
        return 'Kitchen Upgrade'
    elif any(w in name for w in ['vest', 'bra', 'panty', 'trouser', 't-shirt', 'lungi', 'hijab', 'orna']):
        return 'Apparel Shopping'
    elif any(w in name for w in ['biscuit', 'chips', 'chanachur', 'chocolate', 'cake', 'noodles', 'juice', 'cola', 'sprite']):
        return 'Evening Snacks'
    else:
        return 'General Household Needs'

# Prepare Items DataFrame
items = [line.strip() for line in raw_items.split('\n') if line.strip() and line.strip() != 'item_name']
items_df = pd.DataFrame({'item_name': items}).drop_duplicates().reset_index(drop=True)
items_df['item_id'] = ['ITM-' + str(i).zfill(4) for i in range(1, len(items_df) + 1)]
items_df[['Category', 'Sub_Category']] = items_df['item_name'].apply(lambda x: pd.Series(get_standard_category(x)))
items_df['Use_Case_Occasion'] = items_df['item_name'].apply(get_occasion_tag)

# Prepare Customers DataFrame
names = [name.strip() for name in raw_names.split('\n') if name.strip()]
customers_df = pd.DataFrame({'customer_name': names})
customers_df['customer_id'] = ['CUST-' + str(i).zfill(4) for i in range(1, len(customers_df) + 1)]
customers_df['persona'] = np.random.choice(['Regular', 'Weekend_Shopper', 'Holiday_Shopper'], len(customers_df), p=[0.5, 0.3, 0.2])

print(f"Loaded {len(items_df)} items and {len(customers_df)} customers.")

# ==========================================
# STEP 3: SIMULATION WITH TIME PATTERNS
# ==========================================
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 3, 31)
holidays = [datetime(2025, 2, 14), datetime(2025, 2, 21), datetime(2025, 3, 26)]

sales_records = []
print("Generating pattern-based sales dataset. Please wait...")

current_date = start_date
while current_date <= end_date:
    is_weekend = current_date.weekday() in [4, 5]
    is_pre_holiday = (current_date + timedelta(days=1)) in holidays or current_date in holidays
    
    shoppers_today = []
    for _, customer in customers_df.iterrows():
        prob = 0.4 if customer['persona'] == 'Regular' else (0.8 if is_weekend else 0.05) if customer['persona'] == 'Weekend_Shopper' else (0.9 if is_pre_holiday else 0.1)
        if random.random() < prob:
            shoppers_today.append(customer)
            
    for customer in shoppers_today:
        time_slot = random.choices(['Morning', 'Noon', 'Afternoon', 'Evening', 'Night'], weights=[0.25, 0.15, 0.15, 0.30, 0.15])[0]
        
        # ---------------------------------------------------------
        # TIME-BASED BUYING PATTERN LOGIC
        # ---------------------------------------------------------
        if time_slot == 'Morning':
            hour = random.randint(7, 10)
            target_occasions = ['Morning Breakfast', 'Baby Care & Feeding', 'General Household Needs']
            
        elif time_slot == 'Noon':
            hour = random.randint(12, 14)
            target_occasions = ['Regular Meal / Fish Prep', 'General Household Needs']
            
        elif time_slot == 'Afternoon':
            hour = random.randint(15, 17)
            target_occasions = ['Fast Food / Quick Snacks', 'Evening Snacks']
            
        elif time_slot == 'Evening':
            hour = random.randint(18, 20)
            target_occasions = ['Evening Snacks', 'Household Chores', 'Personal Grooming & Care', 'Kids & School Needs', 'Apparel Shopping', 'Kitchen Upgrade']
            
        else: # Night
            hour = random.randint(20, 22)
            # রাতে হেভি মিল বা পরের দিনের রান্না
            target_occasions = ['Heavy Meal / Biryani Cooking', 'Regular Meal / Fish Prep']
            
        order_time = current_date.replace(hour=hour, minute=random.randint(0, 59), second=random.randint(0, 59))
        
        cart = pd.DataFrame()
        
        # Filter items available for this specific time slot
        available_items = items_df[items_df['Use_Case_Occasion'].isin(target_occasions)]
        
        # Fallback to Staples if specific occasion items are missing in dummy list
        if available_items.empty:
            available_items = items_df[items_df['Sub_Category'] == 'Staples']
            
        if not available_items.empty:
            # Basket size is bigger in the Evening and Morning
            num_items = random.randint(3, 8) if time_slot in ['Morning', 'Evening'] else random.randint(2, 5)
            cart = pd.concat([cart, available_items.sample(min(num_items, len(available_items)))])
            
        cart = cart.drop_duplicates(subset=['item_id'])
        if cart.empty: continue
        
        order_id = f"ORD-{order_time.strftime('%Y%m%d')}-{random.randint(1000,9999)}"
        transaction_id = "TXN-" + str(uuid.uuid4()).split('-')[0].upper()
        
        for _, item in cart.iterrows():
            qty = random.randint(1, 4)
            sales_records.append({
                'transaction_id': transaction_id,
                'order_id': order_id,
                'order_datetime': order_time,
                'customer_id': customer['customer_id'],
                'customer_name': customer['customer_name'],
                'customer_persona': customer['persona'],
                'item_id': item['item_id'],
                'item_name': item['item_name'],
                'Category': item['Category'],
                'Sub_Category': item['Sub_Category'],
                'Use_Case_Occasion': item['Use_Case_Occasion'],
                'quantity': qty
            })
            
    current_date += timedelta(days=1)

# ==========================================
# STEP 4: SAVE DATAFRAME
# ==========================================
print("Simulation complete. Converting to DataFrame...")
sales_df = pd.DataFrame(sales_records)
sales_df = sales_df.sort_values(by='order_datetime').reset_index(drop=True)

data_dir = r"C:\D drive\sales_recommendation_system\data"
os.makedirs(data_dir, exist_ok=True)
sales_file_path = os.path.join(data_dir, "merged_sales_history.csv")

print(f"Saving highly realistic dataset to {sales_file_path} ...")
sales_df.to_csv(sales_file_path, index=False)
print("Dataset saved successfully!\n")

print(f"Total Transactions Generated: {sales_df['transaction_id'].nunique()}")
display(sales_df.head(15))
```

```python
import pandas as pd
import os

# ১. ডেটাসেট লোড করা
data_dir = r"C:\D drive\sales_recommendation_system\data"
file_path = os.path.join(data_dir, "merged_sales_history.csv")
df = pd.read_csv(file_path)

# ==========================================
# টাস্ক ১: Time রিমুভ করে শুধু Date রাখা
# ==========================================
# order_datetime কলামটিকে ডেট-টাইমে কনভার্ট করে শুধু date এক্সট্রাক্ট করা
df['order_date'] = pd.to_datetime(df['order_datetime']).dt.date

# পুরনো order_datetime কলামটি ডিলিট করে দেওয়া (ঐচ্ছিক)
df = df.drop(columns=['order_datetime'], errors='ignore')

print("Time removed. Only Date kept in 'order_date' column.")


# ==========================================
# টাস্ক ২: Type A - Item Based Basket (List Format)
# ==========================================
print("Generating Item-Level Basket (List)...")
# প্রতিটি order_id এর আন্ডারে আইটেমগুলোকে ইউনিক লিস্ট হিসেবে গ্রুপ করা
item_basket_list = (df.groupby('order_id')['item_name']
                    .apply(lambda x: list(set(x))) # set() ডুপ্লিকেট আইটেম রিমুভ করবে
                    .reset_index())

# কলামের নাম সুন্দর করা
item_basket_list.rename(columns={'item_name': 'items'}, inplace=True)

# CSV এ সেভ করা
item_basket_list.to_csv(os.path.join(data_dir, "basket_item_level_list.csv"), index=False)
print(f"Item Basket created with shape: {item_basket_list.shape}")


# ==========================================
# টাস্ক ৩: Type B - Sub-Category Based Basket (List Format)
# ==========================================
print("Generating Sub-Category-Level Basket (List)...")
# প্রতিটি order_id এর আন্ডারে সাব-ক্যাটাগরিগুলোকে ইউনিক লিস্ট হিসেবে গ্রুপ করা
subcat_basket_list = (df.groupby('order_id')['Sub_Category']
                      .apply(lambda x: list(set(x)))
                      .reset_index())

# কলামের নাম সুন্দর করা
subcat_basket_list.rename(columns={'Sub_Category': 'sub_categories'}, inplace=True)

# CSV এ সেভ করা
subcat_basket_list.to_csv(os.path.join(data_dir, "basket_subcat_level_list.csv"), index=False)
print(f"Sub-Category Basket created with shape: {subcat_basket_list.shape}")

# মেইন ডেটাসেটটি সেভ করে রাখা
df.to_csv(os.path.join(data_dir, "merged_sales_history_updated.csv"), index=False)
print("All tasks completed successfully!")
```

```python
import pandas as pd
import ast
import os

# ১. ডেটাসেট লোড করা
data_dir = r"C:\D drive\sales_recommendation_system\data"
file_path = os.path.join(data_dir, "merged_sales_history_updated.csv")
df = pd.read_csv(file_path)

print("Calculating Customer Profiles based on Category...")

# ২. প্রতিটি কাস্টমার কোন ক্যাটাগরির কতটি আইটেম কিনেছে তা যোগ করা
customer_cat_counts = df.groupby(['customer_id', 'customer_name', 'Category'])['quantity'].sum().reset_index()

# ৩. কাস্টমার আইডি অনুযায়ী গ্রুপ করে টপ ৫টি ক্যাটাগরি বের করার ফাংশন
def get_top_5_categories(group):
    # quantity অনুযায়ী বড় থেকে ছোট সর্ট করা এবং প্রথম ৫টি নেওয়া
    top_5 = group.sort_values(by='quantity', ascending=False).head(5)
    # শুধু ক্যাটাগরির নামগুলো একটি লিস্ট হিসেবে রিটার্ন করা
    return list(top_5['Category'])

# ৪. কাস্টমার অনুযায়ী ফাংশনটি অ্যাপ্লাই করা
customer_profile_basket = (customer_cat_counts.groupby(['customer_id', 'customer_name'])
                           .apply(get_top_5_categories)
                           .reset_index(name='top_5_categories'))

# ৫. CSV ফাইলে সেভ করা
output_path = os.path.join(data_dir, "customer_profile_basket.csv")
customer_profile_basket.to_csv(output_path, index=False)

print(f"\nCustomer Profile Basket created for {len(customer_profile_basket)} customers!")
print(f"Saved to: {output_path}")

# ৬. প্রথম ৫ জন কাস্টমারের প্রোফাইল দেখা
print("\n=== Sample Output ===")
display(customer_profile_basket.head(5))
```

```python
import pandas as pd
import os

# ১. ডেটাসেট লোড করা
data_dir = r"C:\D drive\sales_recommendation_system\data"
file_path = os.path.join(data_dir, "merged_sales_history_updated.csv")
df = pd.read_csv(file_path)

print("Calculating Occasion Profiles based on Categories & Sub-Categories...")

# ২. প্রতিটি Occasion-এর জন্য Category গুলোর বিক্রির পরিমাণ বের করা
occasion_cat_counts = df.groupby(['Use_Case_Occasion', 'Category'])['quantity'].sum().reset_index()

# টপ ৩ ক্যাটাগরি বের করে লিস্ট বানানো
top_cats = (occasion_cat_counts.sort_values(['Use_Case_Occasion', 'quantity'], ascending=[True, False])
            .groupby('Use_Case_Occasion')
            .head(3)
            .groupby('Use_Case_Occasion')['Category']
            .apply(list)
            .reset_index(name='top_3_categories'))

# ৩. প্রতিটি Occasion-এর জন্য Sub_Category গুলোর বিক্রির পরিমাণ বের করা
occasion_subcat_counts = df.groupby(['Use_Case_Occasion', 'Sub_Category'])['quantity'].sum().reset_index()

# টপ ৩ সাব-ক্যাটাগরি বের করে লিস্ট বানানো
top_subcats = (occasion_subcat_counts.sort_values(['Use_Case_Occasion', 'quantity'], ascending=[True, False])
               .groupby('Use_Case_Occasion')
               .head(3)
               .groupby('Use_Case_Occasion')['Sub_Category']
               .apply(list)
               .reset_index(name='top_3_sub_categories'))

# ৪. Category এবং Sub-Category এর ডেটাফ্রেম দুটো মার্জ (Merge) করা
occasion_profile_basket = pd.merge(top_cats, top_subcats, on='Use_Case_Occasion')

# ৫. CSV ফাইলে সেভ করা
output_path = os.path.join(data_dir, "occasion_profile_basket.csv")
occasion_profile_basket.to_csv(output_path, index=False)

print(f"\nOccasion Profile Basket created and saved to: {output_path}")

# ৬. আউটপুট দেখা
print("\n=== Occasion Profile Mapping ===")
display(occasion_profile_basket)
```

```python
import pandas as pd
import numpy as np
import os
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
from mlxtend.frequent_patterns import fpgrowth, association_rules

# ==========================================
# STEP 0: Data Loading
# ==========================================
data_dir = r"C:\D drive\sales_recommendation_system\data"
file_path = os.path.join(data_dir, "merged_sales_history.csv")

print("Loading main dataset...")
df = pd.read_csv(file_path)

# ==========================================
# STEP 1: Popularity & Basic Features (Fallback)
# ==========================================
print("Calculating Item Popularity...")
# Popularity: কোন আইটেমটি কতবার বিক্রি হয়েছে
popularity_df = df.groupby('item_name')['quantity'].sum().reset_index()
popularity_df = popularity_df.sort_values(by='quantity', ascending=False)
popularity_df.to_csv(os.path.join(data_dir, "step1_item_popularity.csv"), index=False)

# Basket Size: প্রতিটি অর্ডারে মোট কতটি আইটেম ছিল
basket_size_df = df.groupby('order_id')['quantity'].sum().reset_index().rename(columns={'quantity': 'basket_size'})
df = df.merge(basket_size_df, on='order_id', how='left')


# ==========================================
# STEP 2: One-Hot Matrix & FP-Growth (FIXED)
# ==========================================
print("Creating One-Hot Matrix & running FP-Growth...")

# One-hot matrix (Order vs Item)
basket = (df.groupby(['order_id', 'item_name'])['quantity']
          .sum().unstack().reset_index().fillna(0)
          .set_index('order_id'))

def encode_units(x):
    return bool(x >= 1)

one_hot_matrix = basket.map(encode_units)
# Save One-Hot Matrix
one_hot_matrix.to_csv(os.path.join(data_dir, "step2_one_hot_matrix.csv"))

# Run FP-Growth (Support কমানো হয়েছে ০.০৫% এ)
frequent_itemsets = fpgrowth(one_hot_matrix, min_support=0.0005, use_colnames=True)

if not frequent_itemsets.empty:
    # Lift এর পাশাপাশি Confidence দিয়েও ফিল্টার করা হলো
    rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.05)
    rules = rules.sort_values('lift', ascending=False)
    
    # CSV ফাইলে সেভ করা
    rules.to_csv(os.path.join(data_dir, "step2_fpgrowth_rules.csv"), index=False)
    print(f"FP-Growth Success! {len(rules)} rules generated and saved to CSV.")
else:
    print("No frequent itemsets found even with 0.0005 support.")


# ==========================================
# STEP 3: Co-occurrence Matrix (Backup)
# ==========================================
print("Generating Co-occurrence Matrix...")
# Co-occurrence: দুটি আইটেম একসাথে কতবার কেনা হয়েছে
# One-hot matrix এর ডট প্রোডাক্ট নিলেই Co-occurrence পাওয়া যায়
one_hot_numeric = one_hot_matrix.astype(int)
co_occurrence_matrix = one_hot_numeric.T.dot(one_hot_numeric)
np.fill_diagonal(co_occurrence_matrix.values, 0) # নিজের সাথে নিজের পেয়ার 0 করে দেওয়া
co_occurrence_matrix.to_csv(os.path.join(data_dir, "step3_co_occurrence_matrix.csv"))


# ==========================================
# STEP 4: Item-User Matrix & Cosine Similarity (MAIN)
# ==========================================
print("Creating Item-User Matrix & Calculating Collaborative Similarity...")
# Item-User Matrix
item_user_matrix = (df.groupby(['item_name', 'customer_id'])['quantity']
                    .sum().unstack().reset_index().fillna(0)
                    .set_index('item_name'))

item_user_matrix.to_csv(os.path.join(data_dir, "step4_item_user_matrix.csv"))

# Cosine Similarity on Item-User Matrix
item_user_sim = cosine_similarity(item_user_matrix)
item_user_sim_df = pd.DataFrame(item_user_sim, index=item_user_matrix.index, columns=item_user_matrix.index)
item_user_sim_df.to_csv(os.path.join(data_dir, "step4_item_user_similarity.csv"))


# ==========================================
# STEP 5: NLP Content Embedding using HuggingFace MiniLM
# ==========================================
print("Loading HuggingFace MiniLM Model for Content Embedding...")
# Load the pre-trained Sentence Transformer model
model = SentenceTransformer('all-MiniLM-L6-v2')

# Unique আইটেমগুলোর মেটাডেটা আলাদা করা
item_meta = df[['item_name', 'Category', 'Sub_Category', 'Use_Case_Occasion']].drop_duplicates(subset=['item_name']).reset_index(drop=True)

# টেক্সট তৈরি করা যা এম্বেড করা হবে (Item + Category + Occasion)
item_meta['rich_text'] = item_meta['item_name'] + " is a " + item_meta['Sub_Category'] + " in " + item_meta['Category'] + " used for " + item_meta['Use_Case_Occasion']

print("Generating Embeddings (This may take a minute)...")
embeddings = model.encode(item_meta['rich_text'].tolist(), show_progress_bar=True)

# Save Embeddings
embedding_df = pd.DataFrame(embeddings, index=item_meta['item_name'])
embedding_df.to_csv(os.path.join(data_dir, "step5_item_embeddings.csv"))

# Calculate Cosine Similarity on Text Embeddings (Content-Based)
print("Calculating Content-Based Cosine Similarity...")
content_sim = cosine_similarity(embeddings)
content_sim_df = pd.DataFrame(content_sim, index=item_meta['item_name'], columns=item_meta['item_name'])
content_sim_df.to_csv(os.path.join(data_dir, "step5_content_similarity.csv"))

print("\nAll Feature Engineering steps completed and saved successfully to C:\\D drive\\sales_recommendation_system\\data!")
```

```python
import pandas as pd
import os

class HybridRecommender:
    def __init__(self, data_dir):
        print("Loading Engine Components with HARD Diversity Filter...")
        self.data_dir = data_dir
        
        # Load the pre-calculated components
        self.rules = pd.read_csv(os.path.join(data_dir, "step2_fpgrowth_rules.csv"))
        self.collab_sim = pd.read_csv(os.path.join(data_dir, "step4_item_user_similarity.csv"), index_col=0)
        self.content_sim = pd.read_csv(os.path.join(data_dir, "step5_content_similarity.csv"), index_col=0)
        self.popularity = pd.read_csv(os.path.join(data_dir, "step1_item_popularity.csv"))
        
        # Load Item Metadata for Sub-Category matching
        sales_df = pd.read_csv(os.path.join(data_dir, "merged_sales_history.csv"))
        self.item_meta = sales_df.set_index('item_name')[['Sub_Category', 'Use_Case_Occasion']].drop_duplicates().to_dict('index')
        
        print("Hybrid Engine Ready!\n")

    def recommend(self, current_cart, top_n=5):
        recommendations = {}
        
        # Extract the Sub-Categories of items ALREADY in the cart
        cart_subcats = set()
        for item in current_cart:
            if item in self.item_meta:
                cart_subcats.add(self.item_meta[item]['Sub_Category'])

        # 1. FP-Growth (Market Basket) - Weight: 2.0 (High Priority)
        for item in current_cart:
            matched_rules = self.rules[self.rules['antecedents'].apply(lambda x: item in str(x))]
            for _, row in matched_rules.iterrows():
                consequent_item = list(eval(row['consequents']))[0] if isinstance(row['consequents'], str) else list(row['consequents'])[0]
                
                # HARD BAN: Don't add if it's in the cart OR if its sub-category is in the cart
                if consequent_item not in current_cart:
                    if consequent_item in self.item_meta and self.item_meta[consequent_item]['Sub_Category'] not in cart_subcats:
                        recommendations[consequent_item] = recommendations.get(consequent_item, 0) + (row['lift'] * 2.0)

        # 2. Collaborative Filtering (Item-User) - Weight: 1.0
        for item in current_cart:
            if item in self.collab_sim.index:
                # Pull more items initially because we will be throwing many away
                similar_items = self.collab_sim[item].sort_values(ascending=False)[1:30]
                for sim_item, score in similar_items.items():
                    if sim_item not in current_cart:
                        if sim_item in self.item_meta and self.item_meta[sim_item]['Sub_Category'] not in cart_subcats:
                            recommendations[sim_item] = recommendations.get(sim_item, 0) + (score * 1.0)

        # 3. Content-Based (MiniLM) - Weight: 0.5
        for item in current_cart:
            if item in self.content_sim.index:
                content_similar = self.content_sim[item].sort_values(ascending=False)[1:30]
                for sim_item, score in content_similar.items():
                    if sim_item not in current_cart:
                        if sim_item in self.item_meta and self.item_meta[sim_item]['Sub_Category'] not in cart_subcats:
                            recommendations[sim_item] = recommendations.get(sim_item, 0) + (score * 0.5)

        # 4. Bonus: Occasion Synergy
        # Boost items that match the intended occasion of the cart items (e.g., "Biryani & Roast Cooking")
        for item, score in recommendations.items():
            if item in self.item_meta:
                item_occasion = self.item_meta[item]['Use_Case_Occasion']
                for cart_item in current_cart:
                    if cart_item in self.item_meta and item_occasion == self.item_meta[cart_item]['Use_Case_Occasion']:
                        recommendations[item] = score * 1.5

        # 5. Fallback (Popularity)
        if len(recommendations) < top_n:
            top_popular = self.popularity[~self.popularity['item_name'].isin(current_cart)]['item_name'].tolist()
            for pop_item in top_popular:
                if len(recommendations) >= top_n:
                    break
                if pop_item not in recommendations:
                    # Apply the same hard ban to popular items
                    if pop_item in self.item_meta and self.item_meta[pop_item]['Sub_Category'] not in cart_subcats:
                        recommendations[pop_item] = 0.1

        # Sort and return top N
        sorted_recs = sorted(recommendations.items(), key=lambda x: x[1], reverse=True)
        return [item[0] for item in sorted_recs[:top_n]]

# ==========================================
# Testing the Hard Filter Engine
# ==========================================
data_directory = r"C:\D drive\sales_recommendation_system\data"
recommender = HybridRecommender(data_directory)

sample_cart = ["Special Chinigura Rice - 1 kg"]
print(f"Customer Cart: {sample_cart}")

recommended_items = recommender.recommend(current_cart=sample_cart, top_n=5)

print("\nTop 5 Recommendations (HARD Diversity Applied):")
for i, item in enumerate(recommended_items, 1):
    print(f"{i}. {item}")
```
